```python
import sympy as sp

def main():
    print("=== Exact BDI for g=1 Abelian Variety (Test Model) ===\n")

    # 1. 変数の定義
    # 状態変数: x, y (Weierstrassモデルの座標)
    # パラメータ: K (結合強度/分岐パラメータ), w (周波数差/離調パラメータ)
    x, y, K, w = sp.symbols('x y K w', real=True)

    # 2. Weierstrassモデルの係数の定義
    # 論文のサドルノード/ハブの形成を模倣したシンプルなテストモデルを設定します。
    # g2, g3 をパラメータ K, w の関数として定義します。
    g2 = 3 * K
    g3 = w

    # 3. イデアルの生成元の定義 (特異点の条件)
    # 楕円曲線が退化して相境界となるのは、曲線が特異点を持つときです。
    # すなわち、f = 0 かつ df/dx = 0 かつ df/dy = 0 となる (x,y) が存在する条件です。
    f = y**2 - (4 * x**3 - g2 * x - g3)
    fx = sp.diff(f, x)
    fy = sp.diff(f, y)

    print("--- [Process A] Exact BDI (厳密境界イデアル) の抽出 ---")
    print(f"対象の力学系方程式 (f = 0): {f} = 0")

    # 4. グレブナー基底を用いた変数消去 (Algorithm 1)
    # 状態変数 y, x を消去し、パラメータ K, w のみの関係式(境界)を抽出します。
    # 辞書式順序 (lex) で y > x > K > w とします。
    polys = [f, fx, fy]
    gb = sp.groebner(polys, y, x, K, w, order='lex')

    # K, w のみからなる多項式（境界イデアルの生成元）を抽出
    boundary_polys = [p for p in gb if not p.has(x) and not p.has(y)]
    boundary_eq = boundary_polys[0]

    print(f"抽出された Exact BDI の生成元: I_bd = < {boundary_eq} >\n")

    print("--- [Process B] 診断指標 TI と OT の計算 ---")

    # 5. TI (Transversality Index: 横断性指標) の計算
    # 境界イデアルのヤコビアン (勾配ベクトル dφ) を計算します。
    grad_K = sp.diff(boundary_eq, K)
    grad_w = sp.diff(boundary_eq, w)

    print(f"境界の勾配ベクトル dφ = [{grad_K},  {grad_w}]")

    # 特異点ハブの候補である原点 (K=0, w=0) での TI を評価します。
    hub_point = {K: 0, w: 0}
    TI_at_hub = [grad_K.subs(hub_point), grad_w.subs(hub_point)]

    print(f"原点 (K=0, w=0) における TI 評価値: {TI_at_hub}")
    if all(val == 0 for val in TI_at_hub):
        print(" => [判定] TI = 0 です。マクロな横断性が喪失し、相転移が凍結 (不可視化) しています。")

    # 6. OT (Overlap Thickness: 交差重複度) の評価
    # 原点における局所環の次元(代数的な重複度)を、多項式の最低次数として評価します。
    terms = sp.Add.make_args(boundary_eq.expand())
    min_degree = min(sum(sp.degree(term, var) for var in (K, w)) for term in terms)

    print(f"\n原点における OT (交差重複度の最低次数): {min_degree}")
    if min_degree > 1:
        print(" => [判定] OT > 1 です。複数の境界が重なる特異なハブ (Invisibility Hub) が形成されています。")
        print("\n=== [結論] 論文の主定理の確認 ===")
        print("OT > 1 (重複度が2以上) の特異点において、微分の性質により")
        print("TI = 0 (横断性の喪失) が必然的に強制されることが代数的に実証されました。")

if __name__ == "__main__":
    main()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def main():
    print("=== Step 2: Invisibility Hub における Mock 異常項の発生シミュレーション ===\n")

    # 1. パラメータ空間 (K, w) のメッシュ生成
    # Step 1 で特定したハブ (K=0, w=0) の周辺を探索します。
    K_vals = np.linspace(-0.5, 0.5, 100)
    w_vals = np.linspace(-0.5, 0.5, 100)
    K_mesh, w_mesh = np.meshgrid(K_vals, w_vals)

    # 2. 可視的ブロック U_vis の定義
    # 論文の Example 7.31 に倣い、2x2の有限フーリエ変換行列 (固有値は 1 と -1) とします。
    U_vis = np.array([[ 1/np.sqrt(2),  1/np.sqrt(2)],
                      [ 1/np.sqrt(2), -1/np.sqrt(2)]])

    print("可視的ブロック U_vis:")
    print(U_vis)
    eigenvalues, _ = np.linalg.eig(U_vis)
    print(f"U_vis の固有値 (可視的スペクトル): {eigenvalues}\n")

    # 3. 暗黒スカラー chi(K, w) と 生のオフ対角ベクトル B_raw の設定
    # 論文の「共鳴条件」をシミュレートするため、ハブ (0,0) において
    # 暗黒スカラー chi が U_vis の固有値 '1' に一致 (共鳴) するようにモデル化します。
    def calc_chi(k, w):
        # 原点 (Hub) で chi = 1 となり、そこから離れると減衰するモデル
        return 1.0 - (k**2 + w**2)

    # ゲージ変換前の生のオフ対角成分 (適当なノイズや摩擦の初期値)
    B_raw = np.array([0.5, 0.2])

    # 4. Mock異常項 (コホモロジー障害 B_s) の計算
    # ゲージ消去方程式: (U_vis - chi * I) X = -B_raw
    # これを解いて B_raw を消去できるかテストします。
    def calculate_mock_anomaly(chi_val, U, B):
        I = np.eye(2)
        U_chi = U - chi_val * I

        # 擬似逆行列 (pinv) を用いて、消去可能な成分を差し引きます。
        # (U - chi*I) が正則であれば B_gauged はゼロベクトルになります。
        # 特異 (共鳴) になると、像空間に直交する成分が「真の障害」として残ります。
        X = np.linalg.pinv(U_chi) @ (-B)
        B_gauged = B + U_chi @ X

        # 残存した Mock異常項 のノルム (大きさ) を返す
        return np.linalg.norm(B_gauged)

    # メッシュ上の全パラメータで異常項の大きさを計算
    anomaly_surface = np.zeros_like(K_mesh)
    for i in range(K_mesh.shape[0]):
        for j in range(K_mesh.shape[1]):
            chi_val = calc_chi(K_mesh[i, j], w_mesh[i, j])
            anomaly_surface[i, j] = calculate_mock_anomaly(chi_val, U_vis, B_raw)

    print("計算完了。結果をプロットします...")

    # 5. 可視化 (3Dサーフェスプロット)
    fig = plt.figure(figsize=(10, 7))
    ax = fig.add_subplot(111, projection='3d')

    surf = ax.plot_surface(K_mesh, w_mesh, anomaly_surface, cmap='magma',
                           edgecolor='none', alpha=0.9)

    ax.set_title('Emergence of Mock Anomaly at the Invisibility Hub (K=0, w=0)', fontsize=14)
    ax.set_xlabel('Parameter K (Coupling)', fontsize=12)
    ax.set_ylabel('Parameter w (Detuning)', fontsize=12)
    ax.set_zlabel('Mock Anomaly Norm ||B_s|| \n(Shadow Price / Dark Leakage)', fontsize=12)

    # ハブの位置に目印のラインを引く
    ax.plot([0,0], [0,0], [0, np.max(anomaly_surface)], color='cyan', linestyle='--', linewidth=2, label='Invisibility Hub (TI=0)')
    ax.legend()

    fig.colorbar(surf, shrink=0.5, aspect=10, label='Anomaly Magnitude')
    plt.show()

if __name__ == "__main__":
    main()
```
