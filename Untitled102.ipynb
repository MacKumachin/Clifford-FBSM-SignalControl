```python
# ============================================
# Minimal benchmark skeleton for
# A) stagewise pole cancellation
# B) CK-symmetric groupoid tracking
# for Hyperelliptic V8 and Elliptic D8
# ============================================

import math
import cmath
import itertools
from dataclasses import dataclass, field
from typing import List, Dict, Tuple, Any
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# -----------------------------
# Basic data structures
# -----------------------------

@dataclass
class Packet:
    label: Any
    stabiliser_size: int = 1
    lambda_dict: Dict[str, complex] = field(default_factory=dict)

@dataclass
class StageState:
    stage_id: int
    packets: List[Packet]
    divisors: List[str]
    chart_labels: List[str]
    weights: Dict[Any, float] = field(default_factory=dict)
    mult: Dict[Tuple[Any, str], int] = field(default_factory=dict)
    symmetry_edges: List[Tuple[str, str]] = field(default_factory=list)
    cocycle_defect: Dict[Tuple[str, str, str], int] = field(default_factory=dict)
    benchmark_name: str = ""

# -----------------------------
# Utility: safe naming
# -----------------------------

def obj_name(packet_label, divisor, chart):
    return f"{packet_label}|{divisor}|{chart}"

# -----------------------------
# Zeta surrogate and pole proxy
# -----------------------------
# We avoid symbolic Laurent expansion here and use a numerical proxy:
# Evaluate near s0 and use log growth as a pole-intensity proxy.
# This is sufficient for the first benchmark.

def local_factor(packet: Packet, s: complex, state: StageState, q: float = 2.0):
    val = 1.0 + 0j
    for E in state.divisors:
        m = state.mult.get((packet.label, E), 0)
        if m == 0:
            continue
        lam = packet.lambda_dict.get(E, 1.0 + 0j)
        denom = 1.0 - lam * (q ** (-s))
        # avoid exact zero
        if abs(denom) < 1e-12:
            denom += 1e-12
        val *= denom ** (-m)
    return state.weights.get(packet.label, 1.0) * val

def stage_zeta(state: StageState, s: complex, q: float = 2.0):
    return sum(local_factor(p, s, state, q=q) for p in state.packets)

def pole_intensity(state: StageState, s0: float = 1.0, eps: float = 1e-4, q: float = 2.0):
    """
    Numerical proxy for principal-part intensity:
    |Z(s0+eps) - Z(s0+2eps)| scaled in log.
    """
    z1 = stage_zeta(state, s0 + eps, q=q)
    z2 = stage_zeta(state, s0 + 2 * eps, q=q)
    val = abs(z1 - z2)
    return -math.log(max(val, 1e-12))

def pole_order_proxy(state: StageState, s0: float = 1.0, eps_grid=None, q: float = 2.0):
    """
    Crude regression-based proxy for pole order:
    if |Z(s0+eps)| ~ C * eps^{-r}, then log|Z| ~ -r log eps + const.
    """
    if eps_grid is None:
        eps_grid = [1e-1, 5e-2, 1e-2, 5e-3, 1e-3]
    xs, ys = [], []
    for eps in eps_grid:
        z = stage_zeta(state, s0 + eps, q=q)
        xs.append(-math.log(eps))
        ys.append(math.log(max(abs(z), 1e-12)))
    # slope via least squares
    A = np.vstack([xs, np.ones(len(xs))]).T
    slope, intercept = np.linalg.lstsq(A, ys, rcond=None)[0]
    return max(slope, 0.0)

def exceptional_contribution(state: StageState):
    out = {}
    for E in state.divisors:
        total = 0.0
        for p in state.packets:
            total += state.mult.get((p.label, E), 0) * state.weights.get(p.label, 1.0)
        out[E] = total
    return out

# -----------------------------
# Groupoid invariants
# -----------------------------

def build_groupoid_graph(state: StageState):
    G = nx.Graph()
    # Objects
    for p in state.packets:
        for E in state.divisors:
            for ch in state.chart_labels:
                node = obj_name(p.label, E, ch)
                G.add_node(node, packet=p.label, divisor=E, chart=ch)
    # Symmetry edges
    for u, v in state.symmetry_edges:
        if u in G and v in G:
            G.add_edge(u, v, edge_type="symmetry")
    # Chart overlap edges
    for p in state.packets:
        for E in state.divisors:
            chart_nodes = [obj_name(p.label, E, ch) for ch in state.chart_labels]
            for a, b in itertools.combinations(chart_nodes, 2):
                G.add_edge(a, b, edge_type="chart_overlap")
    return G

def orbit_count(state: StageState):
    G = build_groupoid_graph(state)
    return nx.number_connected_components(G)

def packet_isotropy_sizes(state: StageState):
    # Minimal surrogate: isotropy inherited from stabiliser size of packet
    return [p.stabiliser_size for p in state.packets]

def max_isotropy(state: StageState):
    vals = packet_isotropy_sizes(state)
    return max(vals) if vals else 0

def avg_isotropy(state: StageState):
    vals = packet_isotropy_sizes(state)
    return float(np.mean(vals)) if vals else 0.0

def cocycle_score(state: StageState):
    total = len(state.cocycle_defect)
    bad = sum(state.cocycle_defect.values()) if total > 0 else 0
    return bad / total if total > 0 else 0.0

def groupoid_mass(state: StageState):
    # Minimal surrogate:
    # sum packet weight / stabiliser size
    total = 0.0
    for p in state.packets:
        total += state.weights.get(p.label, 1.0) / max(p.stabiliser_size, 1)
    return total

# -----------------------------
# Stage table
# -----------------------------

def stage_summary(state: StageState, s0: float = 1.0):
    return {
        "benchmark": state.benchmark_name,
        "stage": state.stage_id,
        "#packets": len(state.packets),
        "#divisors": len(state.divisors),
        "#charts": len(state.chart_labels),
        "pole_order_proxy": pole_order_proxy(state, s0=s0),
        "pole_intensity": pole_intensity(state, s0=s0),
        "orbit_count": orbit_count(state),
        "max_isotropy": max_isotropy(state),
        "avg_isotropy": avg_isotropy(state),
        "cocycle_score": cocycle_score(state),
        "groupoid_mass": groupoid_mass(state),
    }

# -----------------------------
# Benchmark: Hyperelliptic V8
# -----------------------------

def make_v8_stage(stage_id: int):
    """
    V8 benchmark:
    - 8 packets indexed by 3-bit vectors
    - gradual simplification across stages
    """
    packets = []
    bit_vectors = list(itertools.product([0, 1], repeat=3))

    # Gradual reduction of isotropy
    if stage_id == 0:
        stab = 8
    elif stage_id == 1:
        stab = 4
    elif stage_id == 2:
        stab = 2
    else:
        stab = 1

    divisors = [f"E{j}" for j in range(1, min(stage_id + 2, 4) + 1)]
    charts = [f"U{j}" for j in range(1, 3)]

    for v in bit_vectors:
        lam_dict = {}
        for E in divisors:
            # mild stage-dependent lambda profile
            lam_dict[E] = 1.0 - 0.1 * stage_id + 0.02 * sum(v)
        packets.append(Packet(label=v, stabiliser_size=stab, lambda_dict=lam_dict))

    state = StageState(
        stage_id=stage_id,
        packets=packets,
        divisors=divisors,
        chart_labels=charts,
        benchmark_name="Hyperelliptic V8"
    )

    # weights: gradual redistribution, modest decay
    for p in packets:
        state.weights[p.label] = 1.0 / (1.0 + 0.15 * stage_id)

    # multiplicities: distributed pattern
    for p in packets:
        bit_sum = sum(p.label)
        for E in divisors:
            idx = int(E[1:])
            state.mult[(p.label, E)] = max(0, 2 - abs((bit_sum + stage_id) % 3 - 1))

    # symmetry edges: connect packet translations with one-bit flips
    for p in packets:
        for E in divisors:
            for ch in charts:
                src = obj_name(p.label, E, ch)
                for i in range(3):
                    q = list(p.label)
                    q[i] = 1 - q[i]
                    q = tuple(q)
                    dst = obj_name(q, E, ch)
                    state.symmetry_edges.append((src, dst))

    # cocycle defects: gradually disappear
    triples = list(itertools.combinations([str(v) for v in bit_vectors[:4]], 3))
    for t in triples:
        if stage_id == 0:
            state.cocycle_defect[t] = 1
        elif stage_id == 1:
            state.cocycle_defect[t] = 1 if hash(t) % 2 == 0 else 0
        elif stage_id == 2:
            state.cocycle_defect[t] = 1 if hash(t) % 5 == 0 else 0
        else:
            state.cocycle_defect[t] = 0

    return state

# -----------------------------
# Benchmark: Elliptic D8
# -----------------------------

def make_d8_stage(stage_id: int):
    """
    D8 benchmark:
    - packet size 8 = {0,1,2,3} x {+,-}
    - sharper isotropy collapse than V8
    """
    packet_labels = [(k, sgn) for k in range(4) for sgn in ["+", "-"]]
    packets = []

    # Sharper isotropy drop
    if stage_id == 0:
        stab_reflection = 8
        stab_generic = 4
    elif stage_id == 1:
        stab_reflection = 2
        stab_generic = 2
    else:
        stab_reflection = 1
        stab_generic = 1

    divisors = [f"E{j}" for j in range(1, min(stage_id + 2, 4) + 1)]
    charts = [f"U{j}" for j in range(1, 3)]

    for label in packet_labels:
        k, sgn = label
        is_reflection_like = (k == 0 or k == 2)
        stab = stab_reflection if is_reflection_like else stab_generic
        lam_dict = {}
        for E in divisors:
            idx = int(E[1:])
            lam_dict[E] = 1.0 - 0.18 * stage_id + 0.03 * k + (0.02 if sgn == "+" else -0.02)
        packets.append(Packet(label=label, stabiliser_size=stab, lambda_dict=lam_dict))

    state = StageState(
        stage_id=stage_id,
        packets=packets,
        divisors=divisors,
        chart_labels=charts,
        benchmark_name="Elliptic D8"
    )

    # weights: sharper decay in reflection sector
    for p in packets:
        k, sgn = p.label
        is_reflection_like = (k == 0 or k == 2)
        base = 1.2 if is_reflection_like else 1.0
        decay = 1.0 + (0.35 * stage_id if is_reflection_like else 0.12 * stage_id)
        state.weights[p.label] = base / decay

    # multiplicities: reflection sector initially heavier, then collapses
    for p in packets:
        k, sgn = p.label
        is_reflection_like = (k == 0 or k == 2)
        for E in divisors:
            idx = int(E[1:])
            if is_reflection_like:
                val = max(0, 3 - stage_id - (idx // 2))
            else:
                val = max(0, 1 + (idx == 1) - (stage_id // 2))
            state.mult[(p.label, E)] = val

    # symmetry edges: rotation and reflection edges
    for p in packets:
        k, sgn = p.label
        for E in divisors:
            for ch in charts:
                src = obj_name(p.label, E, ch)

                # rotation
                dst_rot = obj_name(((k + 1) % 4, sgn), E, ch)
                state.symmetry_edges.append((src, dst_rot))

                # reflection
                dst_ref = obj_name(((-k) % 4, "-" if sgn == "+" else "+"), E, ch)
                state.symmetry_edges.append((src, dst_ref))

    # cocycle defects: sharp simplification
    triples = list(itertools.combinations([str(v) for v in packet_labels[:4]], 3))
    for t in triples:
        if stage_id == 0:
            state.cocycle_defect[t] = 1
        elif stage_id == 1:
            state.cocycle_defect[t] = 1 if hash(t) % 7 == 0 else 0
        else:
            state.cocycle_defect[t] = 0

    return state

# -----------------------------
# Run experiment
# -----------------------------

def run_benchmark(make_stage_fn, n_stages=4, s0=1.0):
    states = [make_stage_fn(i) for i in range(n_stages)]
    df = pd.DataFrame([stage_summary(st, s0=s0) for st in states])

    # Add deltas
    for col in ["pole_order_proxy", "orbit_count", "max_isotropy", "cocycle_score"]:
        df[f"delta_{col}"] = df[col].diff()

    return states, df

# -----------------------------
# Plotting helpers
# -----------------------------

def plot_trajectories(df, title_prefix=""):
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["stage"], df["pole_order_proxy"], marker="o", label="pole order proxy")
    ax.plot(df["stage"], df["orbit_count"], marker="s", label="orbit count")
    ax.plot(df["stage"], df["max_isotropy"], marker="^", label="max isotropy")
    ax.set_xlabel("Stage")
    ax.set_ylabel("Value")
    ax.set_title(f"{title_prefix} stage trajectories")
    ax.legend()
    ax.grid(True)
    plt.show()

    fig, ax = plt.subplots(figsize=(8, 4))
    ax.plot(df["stage"], df["cocycle_score"], marker="o", label="cocycle score")
    ax.plot(df["stage"], df["pole_intensity"], marker="s", label="pole intensity")
    ax.set_xlabel("Stage")
    ax.set_ylabel("Value")
    ax.set_title(f"{title_prefix} cocycle / pole-intensity")
    ax.legend()
    ax.grid(True)
    plt.show()

def plot_exceptional_heatmap(states, title_prefix=""):
    all_divs = sorted(set(d for st in states for d in st.divisors), key=lambda x: int(x[1:]))
    data = []
    for E in all_divs:
        row = []
        for st in states:
            row.append(exceptional_contribution(st).get(E, 0.0))
        data.append(row)
    arr = np.array(data)

    fig, ax = plt.subplots(figsize=(8, 4))
    im = ax.imshow(arr, aspect="auto")
    ax.set_xticks(range(len(states)))
    ax.set_xticklabels([st.stage_id for st in states])
    ax.set_yticks(range(len(all_divs)))
    ax.set_yticklabels(all_divs)
    ax.set_xlabel("Stage")
    ax.set_ylabel("Exceptional divisor")
    ax.set_title(f"{title_prefix} exceptional contribution heatmap")
    plt.colorbar(im, ax=ax)
    plt.show()

def draw_groupoid_snapshot(state: StageState, max_nodes=80):
    G = build_groupoid_graph(state)
    nodes = list(G.nodes())[:max_nodes]
    H = G.subgraph(nodes).copy()

    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(H, seed=42)
    nx.draw(H, pos, with_labels=False, node_size=120)
    plt.title(f"{state.benchmark_name} | stage {state.stage_id} | groupoid snapshot")
    plt.show()

# -----------------------------
# Execute both benchmarks
# -----------------------------

states_v8, df_v8 = run_benchmark(make_v8_stage, n_stages=4, s0=1.0)
states_d8, df_d8 = run_benchmark(make_d8_stage, n_stages=4, s0=1.0)

print("=== Hyperelliptic V8 ===")
display(df_v8)

print("=== Elliptic D8 ===")
display(df_d8)

plot_trajectories(df_v8, title_prefix="Hyperelliptic V8")
plot_exceptional_heatmap(states_v8, title_prefix="Hyperelliptic V8")
draw_groupoid_snapshot(states_v8[0])
draw_groupoid_snapshot(states_v8[-1])

plot_trajectories(df_d8, title_prefix="Elliptic D8")
plot_exceptional_heatmap(states_d8, title_prefix="Elliptic D8")
draw_groupoid_snapshot(states_d8[0])
draw_groupoid_snapshot(states_d8[-1])

# -----------------------------
# Minimal comparison table
# -----------------------------

def final_stage_comparison(df_v8, df_d8):
    cols = [
        "benchmark", "stage", "pole_order_proxy", "orbit_count",
        "max_isotropy", "avg_isotropy", "cocycle_score", "groupoid_mass"
    ]
    out = pd.concat([df_v8[cols], df_d8[cols]], axis=0).reset_index(drop=True)
    return out

comparison_df = final_stage_comparison(df_v8, df_d8)
print("=== Comparison table ===")
display(comparison_df)
```

```python
import os
import math
import itertools
from dataclasses import dataclass, field
from typing import List, Dict, Tuple, Any

import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import sympy as sp

# =========================================================
# Output configuration
# =========================================================

OUTDIR = "dark_siegel_stage_outputs_sympy"
os.makedirs(OUTDIR, exist_ok=True)

SAVE_PDF = True
SAVE_PNG = True
DPI = 200

# =========================================================
# Symbolic variables
# =========================================================

s = sp.Symbol("s")
t = sp.Symbol("t")

# =========================================================
# Basic data structures
# =========================================================

@dataclass
class Packet:
    label: Any
    stabiliser_size: int = 1
    lambda_dict: Dict[str, Any] = field(default_factory=dict)   # symbolic/scalar
    exact_pole_divisors: Dict[str, bool] = field(default_factory=dict)

@dataclass
class StageState:
    stage_id: int
    packets: List[Packet]
    divisors: List[str]
    chart_labels: List[str]
    weights: Dict[Any, float] = field(default_factory=dict)
    mult: Dict[Tuple[Any, str], int] = field(default_factory=dict)
    symmetry_edges: List[Tuple[str, str]] = field(default_factory=list)
    cocycle_defect: Dict[Tuple[str, str, str], int] = field(default_factory=dict)
    benchmark_name: str = ""

# =========================================================
# Utilities
# =========================================================

def obj_name(packet_label, divisor, chart):
    return f"{packet_label}|{divisor}|{chart}"

def save_current_figure(stem: str):
    if SAVE_PDF:
        plt.savefig(os.path.join(OUTDIR, f"{stem}.pdf"), bbox_inches="tight")
    if SAVE_PNG:
        plt.savefig(os.path.join(OUTDIR, f"{stem}.png"), dpi=DPI, bbox_inches="tight")

def write_latex_table(df: pd.DataFrame, filename: str, caption: str, label: str, index: bool = False):
    tex = df.to_latex(index=index, escape=False)
    wrapped = (
        "\\begin{table}[t]\n"
        "\\centering\n"
        f"\\caption{{{caption}}}\n"
        f"\\label{{{label}}}\n"
        "\\small\n"
        + tex +
        "\\end{table}\n"
    )
    with open(os.path.join(OUTDIR, filename), "w", encoding="utf-8") as f:
        f.write(wrapped)

# =========================================================
# Symbolic local factor and Laurent analysis
# =========================================================

def local_factor_sym(packet: Packet, state: StageState, s0: float = 1.0, q: int = 2):
    """
    Build a truly meromorphic local factor in s.
    For exact pole divisors, set lambda = q**s0 so that
        1 - lambda*q^{-s0} = 0.
    """
    expr = sp.Integer(1)
    qsym = sp.Integer(q)
    s0sym = sp.Rational(str(s0)) if isinstance(s0, float) else sp.Integer(s0)

    for E in state.divisors:
        m = state.mult.get((packet.label, E), 0)
        if m == 0:
            continue

        exact_pole = packet.exact_pole_divisors.get(E, False)
        if exact_pole:
            lam = qsym**s0sym
        else:
            lam = packet.lambda_dict.get(E, sp.Rational(3, 4) * qsym**s0sym)

        denom = 1 - lam * qsym**(-s)
        expr *= denom**(-m)

    w = sp.Rational(str(state.weights.get(packet.label, 1.0)))
    return sp.simplify(w * expr)

def stage_zeta_sym(state: StageState, s0: float = 1.0, q: int = 2):
    expr = sp.Integer(0)
    for p in state.packets:
        expr += local_factor_sym(p, state, s0=s0, q=q)
    return sp.simplify(expr)

def laurent_series_at_s0(expr, s0: float, order: int = 6):
    """
    Expand in t = s - s0, then re-express in t.
    """
    expr_t = sp.simplify(expr.subs(s, t + s0))
    ser = sp.series(expr_t, t, 0, order)
    return sp.expand(ser.removeO())

def true_pole_order(expr, s0: float, max_order: int = 8):
    """
    Determine pole order from Laurent expansion around s0.
    If lowest power is t^{-r}, return r.
    """
    ser = laurent_series_at_s0(expr, s0=s0, order=max_order)
    if ser == 0:
        return 0

    terms = ser.as_ordered_terms()
    min_pow = None
    for term in terms:
        powers = term.as_powers_dict()
        pow_t = powers.get(t, 0)
        pow_t = int(pow_t)
        if min_pow is None or pow_t < min_pow:
            min_pow = pow_t

    if min_pow is None:
        return 0
    return max(-min_pow, 0)

def principal_part(expr, s0: float, max_order: int = 8):
    ser = laurent_series_at_s0(expr, s0=s0, order=max_order)
    pp = 0
    for term in ser.as_ordered_terms():
        pow_t = int(term.as_powers_dict().get(t, 0))
        if pow_t < 0:
            pp += term
    return sp.simplify(pp)

def principal_part_coefficients(expr, s0: float, max_order: int = 8):
    """
    Returns dict k -> coefficient of t^{-k}
    """
    pp = principal_part(expr, s0=s0, max_order=max_order)
    out = {}
    if pp == 0:
        return out
    for term in sp.expand(pp).as_ordered_terms():
        coeff, pow_t = term.as_coeff_exponent(t)
        pow_t = int(pow_t)
        if pow_t < 0:
            out[-pow_t] = sp.simplify(out.get(-pow_t, 0) + coeff)
    return out

def principal_part_norm(expr, s0: float, max_order: int = 8):
    coeffs = principal_part_coefficients(expr, s0=s0, max_order=max_order)
    if not coeffs:
        return 0.0
    vals = []
    for k, c in coeffs.items():
        cnum = complex(sp.N(c))
        vals.append(abs(cnum))
    return float(sum(vals))

def exceptional_contribution(state: StageState):
    out = {}
    for E in state.divisors:
        total = 0.0
        for p in state.packets:
            total += state.mult.get((p.label, E), 0) * state.weights.get(p.label, 1.0)
        out[E] = total
    return out

# =========================================================
# Groupoid invariants
# =========================================================

def build_groupoid_graph(state: StageState):
    G = nx.Graph()
    for p in state.packets:
        for E in state.divisors:
            for ch in state.chart_labels:
                node = obj_name(p.label, E, ch)
                G.add_node(node, packet=p.label, divisor=E, chart=ch)
    for u, v in state.symmetry_edges:
        if u in G and v in G:
            G.add_edge(u, v, edge_type="symmetry")
    for p in state.packets:
        for E in state.divisors:
            nodes = [obj_name(p.label, E, ch) for ch in state.chart_labels]
            for a, b in itertools.combinations(nodes, 2):
                G.add_edge(a, b, edge_type="chart_overlap")
    return G

def orbit_count(state: StageState):
    G = build_groupoid_graph(state)
    return nx.number_connected_components(G)

def max_isotropy(state: StageState):
    vals = [p.stabiliser_size for p in state.packets]
    return max(vals) if vals else 0

def avg_isotropy(state: StageState):
    vals = [p.stabiliser_size for p in state.packets]
    return float(np.mean(vals)) if vals else 0.0

def cocycle_score(state: StageState):
    total = len(state.cocycle_defect)
    bad = sum(state.cocycle_defect.values()) if total > 0 else 0
    return bad / total if total > 0 else 0.0

def groupoid_mass(state: StageState):
    total = 0.0
    for p in state.packets:
        total += state.weights.get(p.label, 1.0) / max(p.stabiliser_size, 1)
    return total

# =========================================================
# Benchmarks with exact pole design
# =========================================================

def make_v8_stage(stage_id: int):
    packets = []
    bit_vectors = list(itertools.product([0, 1], repeat=3))

    if stage_id == 0:
        stab = 8
    elif stage_id == 1:
        stab = 4
    elif stage_id == 2:
        stab = 2
    else:
        stab = 1

    divisors = [f"E{j}" for j in range(1, min(stage_id + 2, 4) + 1)]
    charts = [f"U{j}" for j in range(1, 3)]

    for v in bit_vectors:
        lam_dict = {}
        exact_poles = {}
        bit_sum = sum(v)

        for E in divisors:
            idx = int(E[1:])
            # exact poles persist gradually in V8
            exact_poles[E] = ((bit_sum + idx + stage_id) % 3 == 0)
            lam_dict[E] = sp.Rational(3, 4) * 2  # off-pole divisor gets lambda != 2^s0 if s0=1

        packets.append(
            Packet(
                label=v,
                stabiliser_size=stab,
                lambda_dict=lam_dict,
                exact_pole_divisors=exact_poles
            )
        )

    state = StageState(
        stage_id=stage_id,
        packets=packets,
        divisors=divisors,
        chart_labels=charts,
        benchmark_name="Hyperelliptic V8"
    )

    for p in packets:
        state.weights[p.label] = 1.0 / (1.0 + 0.15 * stage_id)

    for p in packets:
        bit_sum = sum(p.label)
        for E in divisors:
            state.mult[(p.label, E)] = max(0, 2 - abs((bit_sum + stage_id) % 3 - 1))

    for p in packets:
        for E in divisors:
            for ch in charts:
                src = obj_name(p.label, E, ch)
                for i in range(3):
                    qv = list(p.label)
                    qv[i] = 1 - qv[i]
                    qv = tuple(qv)
                    dst = obj_name(qv, E, ch)
                    state.symmetry_edges.append((src, dst))

    triples = list(itertools.combinations([str(v) for v in bit_vectors[:4]], 3))
    for t3 in triples:
        if stage_id == 0:
            state.cocycle_defect[t3] = 1
        elif stage_id == 1:
            state.cocycle_defect[t3] = 1 if hash(t3) % 2 == 0 else 0
        elif stage_id == 2:
            state.cocycle_defect[t3] = 1 if hash(t3) % 5 == 0 else 0
        else:
            state.cocycle_defect[t3] = 0

    return state

def make_d2n_stage(stage_id: int, n: int = 4):
    packet_labels = [(k, sgn) for k in range(n) for sgn in ["+", "-"]]
    packets = []

    if stage_id == 0:
        stab_reflection = 2 * n
        stab_generic = n
    elif stage_id == 1:
        stab_reflection = 2
        stab_generic = 2
    else:
        stab_reflection = 1
        stab_generic = 1

    divisors = [f"E{j}" for j in range(1, min(stage_id + 2, 4) + 1)]
    charts = [f"U{j}" for j in range(1, 3)]
    reflection_like_positions = {0, n // 2} if n % 2 == 0 else {0}

    for label in packet_labels:
        k, sgn = label
        is_reflection_like = (k in reflection_like_positions)

        lam_dict = {}
        exact_poles = {}
        for E in divisors:
            idx = int(E[1:])
            # sharp pole concentration in reflection sector at early stages
            exact_poles[E] = is_reflection_like and (stage_id <= 1) and (idx <= 2)
            lam_dict[E] = sp.Rational(3, 4) * 2

        packets.append(
            Packet(
                label=label,
                stabiliser_size=(stab_reflection if is_reflection_like else stab_generic),
                lambda_dict=lam_dict,
                exact_pole_divisors=exact_poles
            )
        )

    state = StageState(
        stage_id=stage_id,
        packets=packets,
        divisors=divisors,
        chart_labels=charts,
        benchmark_name=f"Elliptic D{2*n}"
    )

    for p in packets:
        k, sgn = p.label
        is_reflection_like = (k in reflection_like_positions)
        base = 1.2 if is_reflection_like else 1.0
        decay = 1.0 + (0.35 * stage_id if is_reflection_like else 0.12 * stage_id)
        state.weights[p.label] = base / decay

    for p in packets:
        k, sgn = p.label
        is_reflection_like = (k in reflection_like_positions)
        for E in divisors:
            idx = int(E[1:])
            if is_reflection_like:
                val = max(0, 3 - stage_id - (idx // 2))
            else:
                val = max(0, 1 + (idx == 1) - (stage_id // 2))
            state.mult[(p.label, E)] = val

    for p in packets:
        k, sgn = p.label
        for E in divisors:
            for ch in charts:
                src = obj_name(p.label, E, ch)
                dst_rot = obj_name(((k + 1) % n, sgn), E, ch)
                dst_ref = obj_name(((-k) % n, "-" if sgn == "+" else "+"), E, ch)
                state.symmetry_edges.append((src, dst_rot))
                state.symmetry_edges.append((src, dst_ref))

    triples = list(itertools.combinations([str(v) for v in packet_labels[:min(4, len(packet_labels))]], 3))
    for t3 in triples:
        if stage_id == 0:
            state.cocycle_defect[t3] = 1
        elif stage_id == 1:
            state.cocycle_defect[t3] = 1 if hash(t3) % 7 == 0 else 0
        else:
            state.cocycle_defect[t3] = 0

    return state

# =========================================================
# Stage summary with true Laurent data
# =========================================================

def stage_summary(state: StageState, s0: float = 1.0, q: int = 2, max_order: int = 8):
    zeta_expr = stage_zeta_sym(state, s0=s0, q=q)
    pole_order = true_pole_order(zeta_expr, s0=s0, max_order=max_order)
    pp = principal_part(zeta_expr, s0=s0, max_order=max_order)
    pp_coeffs = principal_part_coefficients(zeta_expr, s0=s0, max_order=max_order)
    pp_norm = principal_part_norm(zeta_expr, s0=s0, max_order=max_order)

    row = {
        "benchmark": state.benchmark_name,
        "stage": state.stage_id,
        "#packets": len(state.packets),
        "#divisors": len(state.divisors),
        "#charts": len(state.chart_labels),
        "true_pole_order": pole_order,
        "principal_part_norm": pp_norm,
        "principal_part": sp.latex(pp),
        "orbit_count": orbit_count(state),
        "max_isotropy": max_isotropy(state),
        "avg_isotropy": avg_isotropy(state),
        "cocycle_score": cocycle_score(state),
        "groupoid_mass": groupoid_mass(state),
        "zeta_expr": sp.latex(zeta_expr),
    }

    for k in range(1, 5):
        row[f"pp_coeff_t^(-{k})"] = sp.latex(pp_coeffs.get(k, 0))

    return row

def add_deltas(df: pd.DataFrame, cols=None):
    out = df.copy()
    if cols is None:
        cols = ["true_pole_order", "principal_part_norm", "orbit_count", "max_isotropy", "cocycle_score", "groupoid_mass"]
    for col in cols:
        out[f"delta_{col}"] = out[col].diff()
    return out

# =========================================================
# Runner
# =========================================================

def run_benchmark(make_stage_fn, n_stages=4, s0=1.0, q=2, max_order=8):
    states = [make_stage_fn(i) for i in range(n_stages)]
    rows = [stage_summary(st, s0=s0, q=q, max_order=max_order) for st in states]
    df = pd.DataFrame(rows)
    df = add_deltas(df)
    return states, df

# =========================================================
# Plotting
# =========================================================

def plot_true_pole_trajectories(df: pd.DataFrame, stem: str, title: str):
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["stage"], df["true_pole_order"], marker="o", label="true pole order")
    ax.plot(df["stage"], df["orbit_count"], marker="s", label="orbit count")
    ax.plot(df["stage"], df["max_isotropy"], marker="^", label="max isotropy")
    ax.set_xlabel("Stage")
    ax.set_ylabel("Value")
    ax.set_title(title)
    ax.legend()
    ax.grid(True)
    save_current_figure(stem)
    plt.show()

def plot_principal_part_sync(df: pd.DataFrame, stem: str, title: str):
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["stage"], df["principal_part_norm"], marker="o", label="principal-part norm")
    ax.plot(df["stage"], df["cocycle_score"], marker="s", label="cocycle score")
    ax.plot(df["stage"], df["groupoid_mass"], marker="^", label="groupoid mass")
    ax.set_xlabel("Stage")
    ax.set_ylabel("Value")
    ax.set_title(title)
    ax.legend()
    ax.grid(True)
    save_current_figure(stem)
    plt.show()

def plot_delta_sync(df: pd.DataFrame, stem: str, title: str):
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.plot(df["stage"], df["delta_true_pole_order"], marker="o", label=r"$\Delta$ true pole order")
    ax.plot(df["stage"], df["delta_orbit_count"], marker="s", label=r"$\Delta$ orbit count")
    ax.plot(df["stage"], df["delta_max_isotropy"], marker="^", label=r"$\Delta$ max isotropy")
    ax.plot(df["stage"], df["delta_cocycle_score"], marker="d", label=r"$\Delta$ cocycle score")
    ax.set_xlabel("Stage")
    ax.set_ylabel("Stage difference")
    ax.set_title(title)
    ax.legend()
    ax.grid(True)
    save_current_figure(stem)
    plt.show()

def plot_exceptional_heatmap(states: List[StageState], stem: str, title: str):
    all_divs = sorted(set(d for st in states for d in st.divisors), key=lambda x: int(x[1:]))
    data = []
    for E in all_divs:
        row = []
        for st in states:
            row.append(exceptional_contribution(st).get(E, 0.0))
        data.append(row)
    arr = np.array(data)

    fig, ax = plt.subplots(figsize=(8, 4))
    im = ax.imshow(arr, aspect="auto")
    ax.set_xticks(range(len(states)))
    ax.set_xticklabels([st.stage_id for st in states])
    ax.set_yticks(range(len(all_divs)))
    ax.set_yticklabels(all_divs)
    ax.set_xlabel("Stage")
    ax.set_ylabel("Exceptional divisor")
    ax.set_title(title)
    plt.colorbar(im, ax=ax)
    save_current_figure(stem)
    plt.show()

def draw_groupoid_snapshot(state: StageState, stem: str, title: str, max_nodes: int = 80):
    G = build_groupoid_graph(state)
    nodes = list(G.nodes())[:max_nodes]
    H = G.subgraph(nodes).copy()
    plt.figure(figsize=(8, 6))
    pos = nx.spring_layout(H, seed=42)
    nx.draw(H, pos, with_labels=False, node_size=120)
    plt.title(title)
    save_current_figure(stem)
    plt.show()

# =========================================================
# Export package
# =========================================================

def export_benchmark(states: List[StageState], df: pd.DataFrame, prefix: str):
    df.to_csv(os.path.join(OUTDIR, f"{prefix}_stage_table.csv"), index=False)
    write_latex_table(
        df,
        filename=f"{prefix}_stage_table.tex",
        caption=f"Stage summary with true Laurent diagnostics for the {prefix.replace('_', ' ')} benchmark.",
        label=f"tab:{prefix}_stage_summary",
        index=False
    )

    plot_true_pole_trajectories(
        df,
        stem=f"{prefix}_true_pole_trajectories",
        title=f"{prefix.replace('_', ' ')}: true pole order and groupoid trajectories"
    )
    plot_principal_part_sync(
        df,
        stem=f"{prefix}_principal_part_sync",
        title=f"{prefix.replace('_', ' ')}: principal-part norm and groupoid diagnostics"
    )
    plot_delta_sync(
        df,
        stem=f"{prefix}_delta_sync",
        title=f"{prefix.replace('_', ' ')}: synchronised stage differences"
    )
    plot_exceptional_heatmap(
        states,
        stem=f"{prefix}_exceptional_heatmap",
        title=f"{prefix.replace('_', ' ')}: exceptional contribution heat map"
    )
    draw_groupoid_snapshot(
        states[0],
        stem=f"{prefix}_groupoid_stage0",
        title=f"{prefix.replace('_', ' ')}: CK groupoid snapshot at stage 0"
    )
    draw_groupoid_snapshot(
        states[-1],
        stem=f"{prefix}_groupoid_stage_last",
        title=f"{prefix.replace('_', ' ')}: CK groupoid snapshot at final stage"
    )

# =========================================================
# Run two core benchmarks
# =========================================================

states_v8, df_v8 = run_benchmark(make_v8_stage, n_stages=4, s0=1.0, q=2, max_order=8)
states_d8, df_d8 = run_benchmark(lambda i: make_d2n_stage(i, n=4), n_stages=4, s0=1.0, q=2, max_order=8)

print("=== Hyperelliptic V8 ===")
display(df_v8)

print("=== Elliptic D8 ===")
display(df_d8)

export_benchmark(states_v8, df_v8, prefix="hyperelliptic_v8_sympy")
export_benchmark(states_d8, df_d8, prefix="elliptic_d8_sympy")

# =========================================================
# Comparison table
# =========================================================

comparison_cols = [
    "benchmark", "stage", "true_pole_order", "principal_part_norm",
    "orbit_count", "max_isotropy", "avg_isotropy",
    "cocycle_score", "groupoid_mass",
    "pp_coeff_t^(-1)", "pp_coeff_t^(-2)", "pp_coeff_t^(-3)", "pp_coeff_t^(-4)"
]

comparison_df = pd.concat([df_v8[comparison_cols], df_d8[comparison_cols]], axis=0).reset_index(drop=True)
comparison_df.to_csv(os.path.join(OUTDIR, "comparison_table_sympy.csv"), index=False)

write_latex_table(
    comparison_df,
    filename="comparison_table_sympy.tex",
    caption="Comparison of true Laurent-stage diagnostics for the Hyperelliptic $V_8$ and Elliptic $D_8$ benchmarks.",
    label="tab:benchmark_comparison_sympy",
    index=False
)

display(comparison_df)

print(f"All SymPy outputs written to: {OUTDIR}")
print(sorted(os.listdir(OUTDIR)))
```

```python
# =========================================================
# Dihedral n-sweep extension
# Assumes the previous SymPy benchmark cell has already run
# =========================================================

EPS = 1e-9

def concentration_index(series):
    """
    max |delta| / sum |delta|
    """
    vals = np.array(series.dropna(), dtype=float)
    denom = np.sum(np.abs(vals))
    if denom < EPS:
        return 0.0
    return float(np.max(np.abs(vals)) / denom)

def synchronisation_index(df):
    """
    SI = sum_i min(|Δr_i|, |ΔI_i|+|ΔN_i|+|ΔK_i|) / sum_i |Δr_i|
    """
    dr = np.abs(df["delta_true_pole_order"].fillna(0).to_numpy(dtype=float))
    dI = np.abs(df["delta_max_isotropy"].fillna(0).to_numpy(dtype=float))
    dN = np.abs(df["delta_orbit_count"].fillna(0).to_numpy(dtype=float))
    dK = np.abs(df["delta_cocycle_score"].fillna(0).to_numpy(dtype=float))
    denom = np.sum(dr)
    if denom < EPS:
        return 0.0
    num = np.sum(np.minimum(dr, dI + dN + dK))
    return float(num / denom)

def regime_summary(df):
    """
    Extract regime-level diagnostics from a benchmark dataframe.
    """
    return {
        "benchmark": df["benchmark"].iloc[0],
        "final_stage": int(df["stage"].iloc[-1]),
        "final_true_pole_order": float(df["true_pole_order"].iloc[-1]),
        "final_principal_part_norm": float(df["principal_part_norm"].iloc[-1]),
        "final_orbit_count": float(df["orbit_count"].iloc[-1]),
        "final_max_isotropy": float(df["max_isotropy"].iloc[-1]),
        "final_cocycle_score": float(df["cocycle_score"].iloc[-1]),
        "pole_drop_concentration_PC": concentration_index(df["delta_true_pole_order"]),
        "isotropy_collapse_concentration_IC": concentration_index(df["delta_max_isotropy"]),
        "orbit_reorg_concentration_OC": concentration_index(df["delta_orbit_count"]),
        "cocycle_reorg_concentration_KC": concentration_index(df["delta_cocycle_score"]),
        "synchronisation_index_SI": synchronisation_index(df),
    }

def plot_overlay_metric(dfs_dict, metric, stem, title, ylabel=None):
    plt.figure(figsize=(8, 5))
    for name, df in dfs_dict.items():
        plt.plot(df["stage"], df[metric], marker="o", label=name)
    plt.xlabel("Stage")
    plt.ylabel(ylabel if ylabel else metric)
    plt.title(title)
    plt.grid(True)
    plt.legend()
    save_current_figure(stem)
    plt.show()

def plot_overlay_delta_metric(dfs_dict, metric, stem, title, ylabel=None):
    delta_col = f"delta_{metric}"
    plt.figure(figsize=(8, 5))
    for name, df in dfs_dict.items():
        plt.plot(df["stage"], df[delta_col], marker="o", label=name)
    plt.xlabel("Stage")
    plt.ylabel(ylabel if ylabel else delta_col)
    plt.title(title)
    plt.grid(True)
    plt.legend()
    save_current_figure(stem)
    plt.show()

def plot_regime_indices(regime_df, stem, title):
    x = np.arange(len(regime_df))
    labels = regime_df["benchmark"].tolist()

    plt.figure(figsize=(9, 5))
    plt.plot(x, regime_df["pole_drop_concentration_PC"], marker="o", label="PC")
    plt.plot(x, regime_df["isotropy_collapse_concentration_IC"], marker="s", label="IC")
    plt.plot(x, regime_df["synchronisation_index_SI"], marker="^", label="SI")
    plt.xticks(x, labels, rotation=20)
    plt.ylim(0, 1.05)
    plt.ylabel("Index value")
    plt.title(title)
    plt.grid(True)
    plt.legend()
    save_current_figure(stem)
    plt.show()

def plot_n_dependence(regime_df, stem, title):
    """
    Extract n from benchmark labels of the form 'Elliptic D8', 'Elliptic D12', ...
    V8 baseline is shown separately at x=0.
    """
    xs = []
    pcs = []
    ics = []
    sis = []
    labels = []

    for _, row in regime_df.iterrows():
        bench = row["benchmark"]
        if "Hyperelliptic V8" in bench:
            x = 0
        else:
            # e.g. "Elliptic D8" -> 8
            dnum = int(bench.split("D")[-1])
            x = dnum
        xs.append(x)
        pcs.append(row["pole_drop_concentration_PC"])
        ics.append(row["isotropy_collapse_concentration_IC"])
        sis.append(row["synchronisation_index_SI"])
        labels.append(bench)

    order = np.argsort(xs)
    xs = np.array(xs)[order]
    pcs = np.array(pcs)[order]
    ics = np.array(ics)[order]
    sis = np.array(sis)[order]
    labels = np.array(labels)[order]

    plt.figure(figsize=(9, 5))
    plt.plot(xs, pcs, marker="o", label="PC")
    plt.plot(xs, ics, marker="s", label="IC")
    plt.plot(xs, sis, marker="^", label="SI")
    for x, y, lab in zip(xs, pcs, labels):
        plt.annotate(lab, (x, y), textcoords="offset points", xytext=(0, 6), ha="center", fontsize=8)
    plt.xlabel("Dihedral size parameter (0 = hyperelliptic V8 baseline, otherwise D_{2n} label number)")
    plt.ylabel("Index value")
    plt.ylim(0, 1.05)
    plt.title(title)
    plt.grid(True)
    plt.legend()
    save_current_figure(stem)
    plt.show()

# ---------------------------------------------------------
# Run V8 baseline and D_{2n} sweep
# ---------------------------------------------------------

n_values = [4, 6, 8]

sweep_states = {}
sweep_dfs = {}

# Baseline
states_v8_sweep, df_v8_sweep = run_benchmark(make_v8_stage, n_stages=4, s0=1.0, q=2, max_order=8)
sweep_states["Hyperelliptic V8"] = states_v8_sweep
sweep_dfs["Hyperelliptic V8"] = df_v8_sweep

# Dihedral sweep
for n in n_values:
    name = f"Elliptic D{2*n}"
    states_dn, df_dn = run_benchmark(lambda i, n=n: make_d2n_stage(i, n=n), n_stages=4, s0=1.0, q=2, max_order=8)
    sweep_states[name] = states_dn
    sweep_dfs[name] = df_dn

# ---------------------------------------------------------
# Export full stage tables for each benchmark
# ---------------------------------------------------------

for name, df in sweep_dfs.items():
    prefix = name.lower().replace(" ", "_")
    df.to_csv(os.path.join(OUTDIR, f"{prefix}_sweep_stage_table.csv"), index=False)
    write_latex_table(
        df,
        filename=f"{prefix}_sweep_stage_table.tex",
        caption=f"True Laurent stage diagnostics for the {name} benchmark.",
        label=f"tab:{prefix}_sweep_stage_table",
        index=False
    )

# ---------------------------------------------------------
# Regime summary table
# ---------------------------------------------------------

regime_rows = [regime_summary(df) for df in sweep_dfs.values()]
regime_df = pd.DataFrame(regime_rows)

regime_df.to_csv(os.path.join(OUTDIR, "dihedral_sweep_regime_summary.csv"), index=False)
write_latex_table(
    regime_df,
    filename="dihedral_sweep_regime_summary.tex",
    caption="Regime-level diagnostics for the hyperelliptic $V_8$ baseline and the elliptic dihedral $D_{2n}$ sweep.",
    label="tab:dihedral_sweep_regime_summary",
    index=False
)

display(regime_df)

# ---------------------------------------------------------
# Final-stage comparison table
# ---------------------------------------------------------

final_stage_rows = []
for name, df in sweep_dfs.items():
    row = df.iloc[-1].copy()
    final_stage_rows.append(row)

final_stage_df = pd.DataFrame(final_stage_rows)
final_stage_df.to_csv(os.path.join(OUTDIR, "dihedral_sweep_final_stage.csv"), index=False)
write_latex_table(
    final_stage_df,
    filename="dihedral_sweep_final_stage.tex",
    caption="Final-stage comparison for the hyperelliptic $V_8$ baseline and the elliptic dihedral $D_{2n}$ sweep.",
    label="tab:dihedral_sweep_final_stage",
    index=False
)

display(final_stage_df)

# ---------------------------------------------------------
# Overlay plots
# ---------------------------------------------------------

plot_overlay_metric(
    sweep_dfs,
    metric="true_pole_order",
    stem="dihedral_sweep_true_pole_order_overlay",
    title="True pole-order trajectories: hyperelliptic V8 vs elliptic D_{2n} sweep",
    ylabel="True pole order"
)

plot_overlay_metric(
    sweep_dfs,
    metric="principal_part_norm",
    stem="dihedral_sweep_principal_part_norm_overlay",
    title="Principal-part norm trajectories: hyperelliptic V8 vs elliptic D_{2n} sweep",
    ylabel="Principal-part norm"
)

plot_overlay_metric(
    sweep_dfs,
    metric="max_isotropy",
    stem="dihedral_sweep_max_isotropy_overlay",
    title="Maximal isotropy trajectories: hyperelliptic V8 vs elliptic D_{2n} sweep",
    ylabel="Max isotropy"
)

plot_overlay_metric(
    sweep_dfs,
    metric="orbit_count",
    stem="dihedral_sweep_orbit_count_overlay",
    title="Orbit-count trajectories: hyperelliptic V8 vs elliptic D_{2n} sweep",
    ylabel="Orbit count"
)

plot_overlay_metric(
    sweep_dfs,
    metric="cocycle_score",
    stem="dihedral_sweep_cocycle_score_overlay",
    title="Cocycle-defect trajectories: hyperelliptic V8 vs elliptic D_{2n} sweep",
    ylabel="Cocycle score"
)

# Delta overlays
plot_overlay_delta_metric(
    sweep_dfs,
    metric="true_pole_order",
    stem="dihedral_sweep_delta_true_pole_order_overlay",
    title="Stage differences of true pole order across symmetry regimes",
    ylabel=r"$\Delta$ true pole order"
)

plot_overlay_delta_metric(
    sweep_dfs,
    metric="max_isotropy",
    stem="dihedral_sweep_delta_max_isotropy_overlay",
    title="Stage differences of maximal isotropy across symmetry regimes",
    ylabel=r"$\Delta$ max isotropy"
)

plot_overlay_delta_metric(
    sweep_dfs,
    metric="orbit_count",
    stem="dihedral_sweep_delta_orbit_count_overlay",
    title="Stage differences of orbit count across symmetry regimes",
    ylabel=r"$\Delta$ orbit count"
)

# Regime index plots
plot_regime_indices(
    regime_df,
    stem="dihedral_sweep_regime_indices",
    title="Regime-level concentration and synchronisation indices"
)

plot_n_dependence(
    regime_df,
    stem="dihedral_sweep_n_dependence",
    title="Dependence of regime indices on the dihedral scale parameter"
)

print("Sweep outputs written to:", OUTDIR)
print(sorted(os.listdir(OUTDIR)))
```

```python
from sage.all import *
import pandas as pd

# ============================================================
# M24 benchmark: octad filtration + CK-groupoid-style summary
# ============================================================

# ------------------------------------------------------------
# 0. Basic objects: M24 and the extended binary Golay code
# ------------------------------------------------------------

G = MathieuGroup(24)

# Sage version compatibility
try:
    C = codes.GolayCode(GF(2), extended=True)
except TypeError:
    C = codes.GolayCode(GF(2))

print("M24 order =", G.order())
print("Golay code parameters =", C)

# ------------------------------------------------------------
# 1. Extract octads = supports of weight-8 codewords
# ------------------------------------------------------------

octads = []
for c in C:
    if c.hamming_weight() == 8:
        supp = frozenset(i + 1 for i, b in enumerate(c) if b == 1)
        octads.append(supp)

octads = sorted(set(octads), key=lambda S: tuple(sorted(S)))
print("number of octads =", len(octads))   # should be 759

assert len(octads) == 759, "Unexpected octad count; expected 759."

# ------------------------------------------------------------
# 2. Fix a 5-point flag and determine the terminal octad
# ------------------------------------------------------------

flag = (1, 2, 3, 4, 5)

terminal_candidates = [O for O in octads if set(flag).issubset(O)]
print("number of octads containing the 5-set =", len(terminal_candidates))

assert len(terminal_candidates) == 1, "By S(5,8,24), the 5-set should lie in a unique octad."
O_star = terminal_candidates[0]
print("terminal octad O_* =", sorted(O_star))

# ------------------------------------------------------------
# 3. Stage filtration by flag containment
# ------------------------------------------------------------

stage_octads = {}
for i in range(6):
    Fi = set(flag[:i])
    stage_octads[i] = [O for O in octads if Fi.issubset(O)]
    print(f"stage {i}: {len(stage_octads[i])} octads")

# ------------------------------------------------------------
# 4. Basic parity relative to the terminal octad
# ------------------------------------------------------------

def parity(O):
    return (-1) ** (len(O.intersection(O_star)) % 2)

for i in range(6):
    vals = [parity(O) for O in stage_octads[i]]
    plus = sum(1 for x in vals if x == 1)
    minus = sum(1 for x in vals if x == -1)
    print(f"stage {i}: parity + = {plus}, parity - = {minus}")

# ------------------------------------------------------------
# 5. Pointwise stage stabilizer
#     G_i = stabilizer of p_1,...,p_i pointwise
# ------------------------------------------------------------

def pointwise_stabilizer(group, fixed_points):
    H = group
    for p in fixed_points:
        H = H.stabilizer(p)
    return H

stage_stabilizers = {}
for i in range(6):
    fixed_pts = flag[:i]
    stage_stabilizers[i] = pointwise_stabilizer(G, fixed_pts)
    print(f"stage {i}: stabilizer order = {stage_stabilizers[i].order()}")

# ------------------------------------------------------------
# 6. Group action on octads
# ------------------------------------------------------------

def apply_perm_to_subset(g, S):
    return frozenset(g(x) for x in S)

def orbit_of_subset(H, S):
    """
    Orbit of subset S under the subgroup H, computed by BFS on generators.
    """
    gens = list(H.gens())
    orb = {S}
    queue = [S]

    while queue:
        T = queue.pop()
        for h in gens:
            U = apply_perm_to_subset(h, T)
            if U not in orb:
                orb.add(U)
                queue.append(U)

    return orb

def orbit_partition(H, domain):
    """
    Partition domain into H-orbits.
    """
    pending = set(domain)
    orbits = []

    while pending:
        S = next(iter(pending))
        orb = orbit_of_subset(H, S)
        orbits.append(sorted(orb, key=lambda X: tuple(sorted(X))))
        pending -= orb

    return orbits

# ------------------------------------------------------------
# 7. Stagewise groupoid-style diagnostics
# ------------------------------------------------------------

def parity_defect_score(H, domain):
    """
    Fraction of generator-octad pairs for which parity changes.
    This is a lightweight proxy for parity incompatibility.
    """
    gens = list(H.gens())
    if len(domain) == 0 or len(gens) == 0:
        return 0.0

    total = 0
    bad = 0
    for O in domain:
        pO = parity(O)
        for h in gens:
            total += 1
            if parity(apply_perm_to_subset(h, O)) != pO:
                bad += 1

    return bad / total if total > 0 else 0.0

def stage_summary(i):
    domain = stage_octads[i]
    H = stage_stabilizers[i]
    H_order = Integer(H.order())

    orbits = orbit_partition(H, domain)
    orbit_sizes = [len(orb) for orb in orbits]
    isotropy_sizes = [H_order // sz for sz in orbit_sizes]

    vals = [parity(O) for O in domain]
    plus = sum(1 for x in vals if x == 1)
    minus = sum(1 for x in vals if x == -1)

    return {
        "stage": i,
        "#octads": len(domain),
        "stabilizer_order": int(H_order),
        "orbit_count": len(orbits),
        "max_orbit_size": max(orbit_sizes) if orbit_sizes else 0,
        "min_orbit_size": min(orbit_sizes) if orbit_sizes else 0,
        "max_isotropy": max(isotropy_sizes) if isotropy_sizes else 0,
        "min_isotropy": min(isotropy_sizes) if isotropy_sizes else 0,
        "avg_isotropy": float(sum(isotropy_sizes) / len(isotropy_sizes)) if isotropy_sizes else 0.0,
        "parity_plus": plus,
        "parity_minus": minus,
        "parity_defect_score": parity_defect_score(H, domain),
    }

summary_rows = [stage_summary(i) for i in range(6)]
summary_df = pd.DataFrame(summary_rows)

print("\n=== Stage summary table ===")
print(summary_df.to_string(index=False))

# ------------------------------------------------------------
# 8. Optional: show orbit representatives for each stage
# ------------------------------------------------------------

for i in range(6):
    H = stage_stabilizers[i]
    orbits = orbit_partition(H, stage_octads[i])
    print(f"\n--- Stage {i} orbit representatives ({len(orbits)} orbits) ---")
    for j, orb in enumerate(orbits[:10], start=1):  # first 10 only
        rep = orb[0]
        print(f"orbit {j}: size={len(orb)}, representative={sorted(rep)}")

# ------------------------------------------------------------
# 9. Optional export
# ------------------------------------------------------------

summary_df.to_csv("m24_stage_summary.csv", index=False)
print("\nSaved: m24_stage_summary.csv")
```

```python
import itertools
import math
import csv
from collections import defaultdict, deque
import pandas as pd

# ============================================================
# M24-light benchmark (pure Python, no Sage)
# Exact on the Golay/Steiner side, proxy on the groupoid side.
# ============================================================

# ------------------------------------------------------------
# 0. Build the perfect binary Golay code G23 from the standard
#    generator polynomial:
#    g(x) = x^11 + x^10 + x^6 + x^5 + x^4 + x^2 + 1
#    Then extend by one parity bit to get G24.
# ------------------------------------------------------------

def carryless_mul(a: int, b: int) -> int:
    """Carryless multiplication over GF(2), with integer bit representation."""
    out = 0
    x = a
    y = b
    while y:
        if y & 1:
            out ^= x
        x <<= 1
        y >>= 1
    return out

# bit i corresponds to x^i
g_poly = sum(1 << k for k in [11, 10, 6, 5, 4, 2, 0])

def int_to_support_24(code23: int):
    """
    Convert a 23-bit codeword to a 24-coordinate support by appending
    one parity coordinate (coordinate 24) to make total weight even.
    Coordinates are labeled 1,...,24.
    """
    supp = {i + 1 for i in range(23) if (code23 >> i) & 1}
    if len(supp) % 2 == 1:
        supp.add(24)
    return frozenset(supp)

def build_g24_supports():
    """
    Enumerate all 2^12 messages m(x), encode c(x)=m(x)g(x) in G23,
    then extend to G24 by adding one parity bit.
    """
    codewords = set()
    for m in range(1 << 12):
        c23 = carryless_mul(m, g_poly)
        if c23 >= (1 << 23):
            raise ValueError("Unexpected degree overflow in G23 encoding.")
        codewords.add(int_to_support_24(c23))
    return sorted(codewords, key=lambda S: (len(S), tuple(sorted(S))))

codewords_24 = build_g24_supports()
print("number of G24 codewords =", len(codewords_24))
assert len(codewords_24) == 4096, "Expected 2^12 = 4096 codewords."

# ------------------------------------------------------------
# 1. Extract octads
# ------------------------------------------------------------

octads = [S for S in codewords_24 if len(S) == 8]
print("number of octads =", len(octads))
assert len(octads) == 759, "Expected 759 octads."

# ------------------------------------------------------------
# 2. Choose a 5-point flag
#    Usually (1,2,3,4,5) works, but we search if needed.
# ------------------------------------------------------------

def find_unique_5set(octads):
    trial = (1, 2, 3, 4, 5)
    candidates = [O for O in octads if set(trial).issubset(O)]
    if len(candidates) == 1:
        return trial, candidates[0]

    for F in itertools.combinations(range(1, 25), 5):
        candidates = [O for O in octads if set(F).issubset(O)]
        if len(candidates) == 1:
            return F, candidates[0]

    raise RuntimeError("No unique 5-set found; something is wrong.")

flag, O_star = find_unique_5set(octads)
print("flag =", flag)
print("terminal octad O_* =", sorted(O_star))

# ------------------------------------------------------------
# 3. Stage filtration by flag containment
#    stage i: octads containing p1,...,pi
# ------------------------------------------------------------

stage_octads = {}
for i in range(6):
    Fi = set(flag[:i])
    stage_octads[i] = [O for O in octads if Fi.issubset(O)]
    print(f"stage {i}: {len(stage_octads[i])} octads")

# ------------------------------------------------------------
# 4. Parity relative to terminal octad
# ------------------------------------------------------------

def parity(O):
    return (-1) ** (len(O.intersection(O_star)) % 2)

for i in range(6):
    vals = [parity(O) for O in stage_octads[i]]
    plus = sum(1 for x in vals if x == 1)
    minus = sum(1 for x in vals if x == -1)
    print(f"stage {i}: parity += {plus}, parity -= {minus}")

# ------------------------------------------------------------
# 5. Incidence-based CK-groupoid proxy
#
#    We do NOT compute exact M24 orbits here.
#    Instead, we build a stage graph on octads:
#    edge O--O' if |O ∩ O'| = 4.
#
#    Why intersection 4?
#    Because for octads O,O', symmetric difference has weight
#    8 iff |O ∩ O'| = 4, which is the most local octad-to-octad move.
# ------------------------------------------------------------

def build_stage_graph(stage_octads_list):
    n = len(stage_octads_list)
    adj = [[] for _ in range(n)]
    for i in range(n):
        Oi = stage_octads_list[i]
        for j in range(i + 1, n):
            Oj = stage_octads_list[j]
            if len(Oi.intersection(Oj)) == 4:
                adj[i].append(j)
                adj[j].append(i)
    return adj

def connected_components(adj):
    n = len(adj)
    seen = [False] * n
    comps = []
    for i in range(n):
        if seen[i]:
            continue
        q = deque([i])
        seen[i] = True
        comp = []
        while q:
            u = q.popleft()
            comp.append(u)
            for v in adj[u]:
                if not seen[v]:
                    seen[v] = True
                    q.append(v)
        comps.append(comp)
    return comps

def groupoid_proxy_stats(stage_octads_list):
    adj = build_stage_graph(stage_octads_list)
    comps = connected_components(adj)
    comp_sizes = [len(c) for c in comps]
    degs = [len(nei) for nei in adj]

    # parity-defect proxy: fraction of adjacency edges across opposite parity
    total_edges = 0
    bad_edges = 0
    for i, nei in enumerate(adj):
        for j in nei:
            if j > i:
                total_edges += 1
                if parity(stage_octads_list[i]) != parity(stage_octads_list[j]):
                    bad_edges += 1

    parity_defect_score = bad_edges / total_edges if total_edges > 0 else 0.0

    return {
        "component_count": len(comps),
        "max_component_size": max(comp_sizes) if comp_sizes else 0,
        "min_component_size": min(comp_sizes) if comp_sizes else 0,
        "avg_component_size": sum(comp_sizes) / len(comp_sizes) if comp_sizes else 0.0,
        "max_degree": max(degs) if degs else 0,
        "avg_degree": sum(degs) / len(degs) if degs else 0.0,
        "parity_defect_score": parity_defect_score,
    }

# ------------------------------------------------------------
# 6. Toy analytic surrogate
#
#    This is not the full arithmetic local factor.
#    It is a stagewise defect surrogate designed to be honest,
#    simple, and computable.
#
#    For O in stage i:
#    defect_i(O) = (# of future flag points not yet captured by O)
#                  + 1_{parity(O)=-1}
#
#    Then define a toy Laurent surrogate:
#      Z_i(s) = sum_O (1 - 2^(1-s))^{-defect_i(O)}
#
#    Since coefficients are positive, the pole order at s=1 is
#    exactly max defect_i(O), and the principal mass is the number
#    of octads attaining that maximal defect.
# ------------------------------------------------------------

def defect_at_stage(O, i, flag):
    future_flag = flag[i:]  # points not yet forced into the filtration
    missing_future = sum(1 for p in future_flag if p not in O)
    parity_penalty = 1 if parity(O) == -1 else 0
    return missing_future + parity_penalty

def analytic_proxy_stats(stage_octads_list, i, flag):
    defects = [defect_at_stage(O, i, flag) for O in stage_octads_list]
    pole_order = max(defects) if defects else 0
    principal_mass = sum(1 for d in defects if d == pole_order)

    # average defect is useful as a softer measure
    avg_defect = sum(defects) / len(defects) if defects else 0.0

    return {
        "toy_pole_order": pole_order,
        "toy_principal_mass": principal_mass,
        "avg_defect": avg_defect,
    }

# ------------------------------------------------------------
# 7. Terminal-alignment metrics
# ------------------------------------------------------------

def terminal_alignment_stats(stage_octads_list):
    ints = [len(O.intersection(O_star)) for O in stage_octads_list]
    return {
        "min_intersection_with_Ostar": min(ints) if ints else 0,
        "max_intersection_with_Ostar": max(ints) if ints else 0,
        "avg_intersection_with_Ostar": sum(ints) / len(ints) if ints else 0.0,
    }

# ------------------------------------------------------------
# 8. Stage summary
# ------------------------------------------------------------

rows = []
for i in range(6):
    current = stage_octads[i]

    parity_vals = [parity(O) for O in current]
    parity_plus = sum(1 for x in parity_vals if x == 1)
    parity_minus = sum(1 for x in parity_vals if x == -1)

    row = {
        "stage": i,
        "#octads": len(current),
        "parity_plus": parity_plus,
        "parity_minus": parity_minus,
    }
    row.update(groupoid_proxy_stats(current))
    row.update(analytic_proxy_stats(current, i, flag))
    row.update(terminal_alignment_stats(current))
    rows.append(row)

summary_df = pd.DataFrame(rows)

print("\n=== M24-light stage summary ===")
print(summary_df.to_string(index=False))

# ------------------------------------------------------------
# 9. Deltas for synchronisation reading
# ------------------------------------------------------------

delta_df = summary_df.copy()
for col in [
    "#octads",
    "component_count",
    "parity_defect_score",
    "toy_pole_order",
    "toy_principal_mass",
    "avg_defect",
    "avg_intersection_with_Ostar",
]:
    delta_df[f"delta_{col}"] = delta_df[col].diff()

print("\n=== Stage differences ===")
show_cols = ["stage"] + [c for c in delta_df.columns if c.startswith("delta_")]
print(delta_df[show_cols].to_string(index=False))

# ------------------------------------------------------------
# 10. Export
# ------------------------------------------------------------

summary_df.to_csv("m24_light_stage_summary.csv", index=False)
delta_df.to_csv("m24_light_stage_deltas.csv", index=False)
print("\nSaved: m24_light_stage_summary.csv")
print("Saved: m24_light_stage_deltas.csv")
```

```python
import itertools
import math
import os
from collections import deque
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# Output configuration
# ============================================================

OUTDIR = "m24_light_outputs"
os.makedirs(OUTDIR, exist_ok=True)

SAVE_PDF = True
SAVE_PNG = True
DPI = 200

# ============================================================
# Helpers
# ============================================================

def save_current_figure(stem: str):
    if SAVE_PDF:
        plt.savefig(os.path.join(OUTDIR, f"{stem}.pdf"), bbox_inches="tight")
    if SAVE_PNG:
        plt.savefig(os.path.join(OUTDIR, f"{stem}.png"), dpi=DPI, bbox_inches="tight")

def write_latex_table(df: pd.DataFrame, filename: str, caption: str, label: str, index: bool = False):
    tex = df.to_latex(index=index, escape=False, float_format=lambda x: f"{x:.6f}")
    wrapped = (
        "\\begin{table}[t]\n"
        "\\centering\n"
        f"\\caption{{{caption}}}\n"
        f"\\label{{{label}}}\n"
        "\\small\n"
        + tex +
        "\\end{table}\n"
    )
    with open(os.path.join(OUTDIR, filename), "w", encoding="utf-8") as f:
        f.write(wrapped)

# ============================================================
# 0. Build G24 from the standard G23 generator polynomial
# ============================================================

def carryless_mul(a: int, b: int) -> int:
    out = 0
    x = a
    y = b
    while y:
        if y & 1:
            out ^= x
        x <<= 1
        y >>= 1
    return out

# g(x)=x^11+x^10+x^6+x^5+x^4+x^2+1
g_poly = sum(1 << k for k in [11, 10, 6, 5, 4, 2, 0])

def int_to_support_24(code23: int):
    supp = {i + 1 for i in range(23) if (code23 >> i) & 1}
    if len(supp) % 2 == 1:
        supp.add(24)
    return frozenset(supp)

def build_g24_supports():
    codewords = set()
    for m in range(1 << 12):
        c23 = carryless_mul(m, g_poly)
        if c23 >= (1 << 23):
            raise ValueError("Unexpected degree overflow in G23 encoding.")
        codewords.add(int_to_support_24(c23))
    return sorted(codewords, key=lambda S: (len(S), tuple(sorted(S))))

codewords_24 = build_g24_supports()
assert len(codewords_24) == 4096, "Expected 4096 codewords."

octads = [S for S in codewords_24 if len(S) == 8]
assert len(octads) == 759, "Expected 759 octads."

print("number of G24 codewords =", len(codewords_24))
print("number of octads =", len(octads))

# ============================================================
# 1. Choose a 5-point flag with unique completion octad
# ============================================================

def find_unique_5set(octads):
    trial = (1, 2, 3, 4, 5)
    candidates = [O for O in octads if set(trial).issubset(O)]
    if len(candidates) == 1:
        return trial, candidates[0]

    for F in itertools.combinations(range(1, 25), 5):
        candidates = [O for O in octads if set(F).issubset(O)]
        if len(candidates) == 1:
            return F, candidates[0]

    raise RuntimeError("No unique 5-set found.")

flag, O_star = find_unique_5set(octads)
tail = frozenset(O_star.difference(flag))   # 3-point tail inside the terminal octad

print("flag =", flag)
print("terminal octad O_* =", sorted(O_star))
print("tail T = O_* \\ flag =", sorted(tail))

# ============================================================
# 2. Stage filtration
# ============================================================

stage_octads = {}
for i in range(6):
    Fi = set(flag[:i])
    stage_octads[i] = [O for O in octads if Fi.issubset(O)]
    print(f"stage {i}: {len(stage_octads[i])} octads")

# ============================================================
# 3. Nontrivial parity relative to terminal tail
# ============================================================

def parity(O):
    return (-1) ** (len(O.intersection(tail)) % 2)

for i in range(6):
    vals = [parity(O) for O in stage_octads[i]]
    plus = sum(1 for x in vals if x == 1)
    minus = sum(1 for x in vals if x == -1)
    print(f"stage {i}: parity + = {plus}, parity - = {minus}")

# ============================================================
# 4. Incidence-based CK-groupoid proxy graph
#    Edge O--O' if |O ∩ O'| = 4
# ============================================================

def build_stage_graph(stage_octads_list):
    n = len(stage_octads_list)
    adj = [[] for _ in range(n)]
    for i in range(n):
        Oi = stage_octads_list[i]
        for j in range(i + 1, n):
            Oj = stage_octads_list[j]
            if len(Oi.intersection(Oj)) == 4:
                adj[i].append(j)
                adj[j].append(i)
    return adj

def connected_components(adj):
    n = len(adj)
    seen = [False] * n
    comps = []
    for i in range(n):
        if seen[i]:
            continue
        q = deque([i])
        seen[i] = True
        comp = []
        while q:
            u = q.popleft()
            comp.append(u)
            for v in adj[u]:
                if not seen[v]:
                    seen[v] = True
                    q.append(v)
        comps.append(comp)
    return comps

def groupoid_proxy_stats(stage_octads_list):
    adj = build_stage_graph(stage_octads_list)
    comps = connected_components(adj)
    comp_sizes = [len(c) for c in comps]
    degs = [len(nei) for nei in adj]

    total_edges = 0
    bad_edges = 0
    for i, nei in enumerate(adj):
        for j in nei:
            if j > i:
                total_edges += 1
                if parity(stage_octads_list[i]) != parity(stage_octads_list[j]):
                    bad_edges += 1

    parity_defect_score = bad_edges / total_edges if total_edges > 0 else 0.0

    return {
        "component_count": len(comps),
        "max_component_size": max(comp_sizes) if comp_sizes else 0,
        "min_component_size": min(comp_sizes) if comp_sizes else 0,
        "avg_component_size": sum(comp_sizes) / len(comp_sizes) if comp_sizes else 0.0,
        "max_degree": max(degs) if degs else 0,
        "avg_degree": sum(degs) / len(degs) if degs else 0.0,
        "parity_defect_score": parity_defect_score,
    }

# ============================================================
# 5. Toy analytic surrogate
# ============================================================

def defect_at_stage(O, i, flag):
    future_flag = flag[i:]
    missing_future = sum(1 for p in future_flag if p not in O)
    parity_penalty = 1 if parity(O) == -1 else 0
    return missing_future + parity_penalty

def analytic_proxy_stats(stage_octads_list, i, flag):
    defects = [defect_at_stage(O, i, flag) for O in stage_octads_list]
    pole_order = max(defects) if defects else 0
    principal_mass = sum(1 for d in defects if d == pole_order)
    avg_defect = sum(defects) / len(defects) if defects else 0.0
    return {
        "toy_pole_order": pole_order,
        "toy_principal_mass": principal_mass,
        "avg_defect": avg_defect,
    }

# ============================================================
# 6. Terminal-alignment statistics
# ============================================================

def terminal_alignment_stats(stage_octads_list):
    ints = [len(O.intersection(O_star)) for O in stage_octads_list]
    tail_ints = [len(O.intersection(tail)) for O in stage_octads_list]
    return {
        "min_intersection_with_Ostar": min(ints) if ints else 0,
        "max_intersection_with_Ostar": max(ints) if ints else 0,
        "avg_intersection_with_Ostar": sum(ints) / len(ints) if ints else 0.0,
        "avg_intersection_with_tail": sum(tail_ints) / len(tail_ints) if tail_ints else 0.0,
    }

# ============================================================
# 7. Stage summary
# ============================================================

rows = []
for i in range(6):
    current = stage_octads[i]
    parity_vals = [parity(O) for O in current]
    parity_plus = sum(1 for x in parity_vals if x == 1)
    parity_minus = sum(1 for x in parity_vals if x == -1)

    row = {
        "stage": i,
        "#octads": len(current),
        "parity_plus": parity_plus,
        "parity_minus": parity_minus,
    }
    row.update(groupoid_proxy_stats(current))
    row.update(analytic_proxy_stats(current, i, flag))
    row.update(terminal_alignment_stats(current))
    rows.append(row)

summary_df = pd.DataFrame(rows)

print("\n=== M24-light stage summary ===")
print(summary_df.to_string(index=False))

# ============================================================
# 8. Deltas
# ============================================================

delta_df = summary_df.copy()
for col in [
    "#octads",
    "component_count",
    "parity_defect_score",
    "toy_pole_order",
    "toy_principal_mass",
    "avg_defect",
    "avg_intersection_with_Ostar",
    "avg_intersection_with_tail",
]:
    delta_df[f"delta_{col}"] = delta_df[col].diff()

print("\n=== Stage differences ===")
show_cols = ["stage"] + [c for c in delta_df.columns if c.startswith("delta_")]
print(delta_df[show_cols].to_string(index=False))

# ============================================================
# 9. Export tables
# ============================================================

summary_df.to_csv(os.path.join(OUTDIR, "m24_light_stage_summary.csv"), index=False)
delta_df.to_csv(os.path.join(OUTDIR, "m24_light_stage_deltas.csv"), index=False)

write_latex_table(
    summary_df,
    filename="m24_light_stage_summary.tex",
    caption="Stage summary for the lightweight $M_{24}$ benchmark.",
    label="tab:m24_light_stage_summary",
    index=False
)

write_latex_table(
    delta_df[show_cols],
    filename="m24_light_stage_deltas.tex",
    caption="Stage differences for the lightweight $M_{24}$ benchmark.",
    label="tab:m24_light_stage_deltas",
    index=False
)

# ============================================================
# 10. Figures
# ============================================================

# Fig 1: filtration size
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["#octads"], marker="o")
plt.xlabel("Stage")
plt.ylabel("Number of octads")
plt.title(r"$M_{24}$ benchmark: stage filtration size")
plt.grid(True)
save_current_figure("m24_light_filtration_size")
plt.show()

# Fig 2: analytic diagnostics
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["toy_pole_order"], marker="o", label="toy pole order")
plt.plot(summary_df["stage"], summary_df["avg_defect"], marker="s", label="average defect")
plt.plot(summary_df["stage"], summary_df["toy_principal_mass"], marker="^", label="principal mass")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title(r"$M_{24}$ benchmark: analytic proxy diagnostics")
plt.grid(True)
plt.legend()
save_current_figure("m24_light_analytic_diagnostics")
plt.show()

# Fig 3: groupoid / parity diagnostics
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["component_count"], marker="o", label="component count")
plt.plot(summary_df["stage"], summary_df["avg_degree"], marker="s", label="average degree")
plt.plot(summary_df["stage"], summary_df["parity_defect_score"], marker="^", label="parity defect score")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title(r"$M_{24}$ benchmark: groupoid proxy diagnostics")
plt.grid(True)
plt.legend()
save_current_figure("m24_light_groupoid_diagnostics")
plt.show()

# Fig 4: terminal alignment
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["avg_intersection_with_Ostar"], marker="o", label=r"avg $|O \cap O_*|$")
plt.plot(summary_df["stage"], summary_df["avg_intersection_with_tail"], marker="s", label=r"avg $|O \cap T|$")
plt.xlabel("Stage")
plt.ylabel("Average intersection")
plt.title(r"$M_{24}$ benchmark: terminal alignment")
plt.grid(True)
plt.legend()
save_current_figure("m24_light_terminal_alignment")
plt.show()

# Fig 5: synchronised deltas
plt.figure(figsize=(8, 4.5))
plt.plot(delta_df["stage"], delta_df["delta_toy_pole_order"], marker="o", label=r"$\Delta$ toy pole order")
plt.plot(delta_df["stage"], delta_df["delta_avg_defect"], marker="s", label=r"$\Delta$ avg defect")
plt.plot(delta_df["stage"], delta_df["delta_avg_intersection_with_Ostar"], marker="^", label=r"$\Delta$ avg $|O\cap O_*|$")
plt.plot(delta_df["stage"], delta_df["delta_parity_defect_score"], marker="d", label=r"$\Delta$ parity defect score")
plt.xlabel("Stage")
plt.ylabel("Stage difference")
plt.title(r"$M_{24}$ benchmark: synchronised stage differences")
plt.grid(True)
plt.legend()
save_current_figure("m24_light_synchronised_deltas")
plt.show()

# ============================================================
# 11. LaTeX include snippet
# ============================================================

latex_snippet = r"""
\input{m24_light_stage_summary.tex}

\input{m24_light_stage_deltas.tex}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{m24_light_filtration_size.pdf}
\caption{Stage filtration size for the lightweight \(M_{24}\) benchmark.}
\label{fig:m24-light-filtration}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{m24_light_analytic_diagnostics.pdf}
\caption{Analytic proxy diagnostics for the lightweight \(M_{24}\) benchmark.}
\label{fig:m24-light-analytic}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{m24_light_groupoid_diagnostics.pdf}
\caption{Groupoid proxy diagnostics for the lightweight \(M_{24}\) benchmark.}
\label{fig:m24-light-groupoid}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{m24_light_terminal_alignment.pdf}
\caption{Terminal alignment diagnostics for the lightweight \(M_{24}\) benchmark.}
\label{fig:m24-light-terminal}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.80\textwidth]{m24_light_synchronised_deltas.pdf}
\caption{Synchronised stage differences for the lightweight \(M_{24}\) benchmark.}
\label{fig:m24-light-sync}
\end{figure}
"""

with open(os.path.join(OUTDIR, "m24_light_include_snippet.tex"), "w", encoding="utf-8") as f:
    f.write(latex_snippet)

print("\nSaved files:")
for fn in sorted(os.listdir(OUTDIR)):
    print(" -", fn)
```

```python
import itertools
import math
import os
from collections import deque, defaultdict
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# Leech/Conway-light preview
# Reuses exact M24/Golay/octad data from the previous cell.
# This is a signed-octad shell proxy, not the full Leech lattice.
# ============================================================

# ------------------------------------------------------------
# 0. Preconditions
# ------------------------------------------------------------

required_names = ["octads", "flag", "O_star"]
missing = [name for name in required_names if name not in globals()]
if missing:
    raise RuntimeError(
        f"Missing prerequisite variables: {missing}. "
        "Run the M24-light cell first."
    )

OUTDIR2 = "leech_conway_light_outputs"
os.makedirs(OUTDIR2, exist_ok=True)

SAVE_PDF = True
SAVE_PNG = True
DPI = 200

def save_current_figure(stem: str):
    if SAVE_PDF:
        plt.savefig(os.path.join(OUTDIR2, f"{stem}.pdf"), bbox_inches="tight")
    if SAVE_PNG:
        plt.savefig(os.path.join(OUTDIR2, f"{stem}.png"), dpi=DPI, bbox_inches="tight")

def write_latex_table(df: pd.DataFrame, filename: str, caption: str, label: str, index: bool = False):
    tex = df.to_latex(index=index, escape=False, float_format=lambda x: f"{x:.6f}")
    wrapped = (
        "\\begin{table}[t]\n"
        "\\centering\n"
        f"\\caption{{{caption}}}\n"
        f"\\label{{{label}}}\n"
        "\\small\n"
        + tex +
        "\\end{table}\n"
    )
    with open(os.path.join(OUTDIR2, filename), "w", encoding="utf-8") as f:
        f.write(wrapped)

tail = frozenset(set(O_star).difference(flag))

# ------------------------------------------------------------
# 1. Stage filtration (support level)
# ------------------------------------------------------------

stage_octads_lc = {}
for i in range(6):
    Fi = set(flag[:i])
    stage_octads_lc[i] = [O for O in octads if Fi.issubset(O)]

# ------------------------------------------------------------
# 2. Conway-like signed decorations
#
# For each octad O, let U = O ∩ O_*.
# We decorate O by an even subset S ⊆ U of size 0 or 2.
# This is a lightweight proxy for signed shell data.
# ------------------------------------------------------------

def signed_decorations(O):
    U = sorted(O.intersection(O_star))
    decos = [frozenset()]  # size 0
    for a, b in itertools.combinations(U, 2):
        decos.append(frozenset([a, b]))
    return decos

def stage_states(i):
    """
    States are pairs (O, S) where O is an octad at stage i,
    and S is a small even sign-decoration on O ∩ O_*.
    """
    states = []
    for O in stage_octads_lc[i]:
        for S in signed_decorations(O):
            states.append((O, S))
    return states

# ------------------------------------------------------------
# 3. Signed parity
#
# Support parity term:
#   |O ∩ tail| mod 2
# Decoration parity term:
#   |S|/2 mod 2   (since |S| is 0 or 2)
# ------------------------------------------------------------

def state_parity(state):
    O, S = state
    support_term = len(O.intersection(tail)) % 2
    deco_term = (len(S) // 2) % 2
    return (-1) ** (support_term + deco_term)

# ------------------------------------------------------------
# 4. Signed-octad adjacency graph
#
# Support edge if |O ∩ O'| = 4.
# Then signed states are adjacent if:
#   (a) same support and decorations differ by exactly 2, or
#   (b) support-adjacent and decoration symmetric difference has size 0 or 2.
# ------------------------------------------------------------

def support_adjacent(O, O2):
    return len(O.intersection(O2)) == 4

def build_state_graph(states):
    n = len(states)
    adj = [[] for _ in range(n)]

    # group states by support
    support_to_indices = defaultdict(list)
    for idx, (O, S) in enumerate(states):
        support_to_indices[O].append(idx)

    supports = list(support_to_indices.keys())

    # (a) internal sign-flip edges
    for O, idxs in support_to_indices.items():
        for a in range(len(idxs)):
            for b in range(a + 1, len(idxs)):
                ia, ib = idxs[a], idxs[b]
                Sa = states[ia][1]
                Sb = states[ib][1]
                if len(Sa.symmetric_difference(Sb)) == 2:
                    adj[ia].append(ib)
                    adj[ib].append(ia)

    # (b) support-level edges
    for i in range(len(supports)):
        Oi = supports[i]
        for j in range(i + 1, len(supports)):
            Oj = supports[j]
            if support_adjacent(Oi, Oj):
                idxs_i = support_to_indices[Oi]
                idxs_j = support_to_indices[Oj]
                for ia in idxs_i:
                    for ib in idxs_j:
                        Sa = states[ia][1]
                        Sb = states[ib][1]
                        if len(Sa.symmetric_difference(Sb)) in (0, 2):
                            adj[ia].append(ib)
                            adj[ib].append(ia)

    return adj

def connected_components(adj):
    n = len(adj)
    seen = [False] * n
    comps = []
    for i in range(n):
        if seen[i]:
            continue
        q = deque([i])
        seen[i] = True
        comp = []
        while q:
            u = q.popleft()
            comp.append(u)
            for v in adj[u]:
                if not seen[v]:
                    seen[v] = True
                    q.append(v)
        comps.append(comp)
    return comps

# ------------------------------------------------------------
# 5. Groupoid-style diagnostics
# ------------------------------------------------------------

def groupoid_stats(states):
    adj = build_state_graph(states)
    comps = connected_components(adj)
    comp_sizes = [len(c) for c in comps]
    degs = [len(nei) for nei in adj]

    total_edges = 0
    bad_edges = 0
    for i, nei in enumerate(adj):
        for j in nei:
            if j > i:
                total_edges += 1
                if state_parity(states[i]) != state_parity(states[j]):
                    bad_edges += 1

    parity_defect_score = bad_edges / total_edges if total_edges > 0 else 0.0

    return {
        "component_count": len(comps),
        "max_component_size": max(comp_sizes) if comp_sizes else 0,
        "min_component_size": min(comp_sizes) if comp_sizes else 0,
        "avg_component_size": sum(comp_sizes) / len(comp_sizes) if comp_sizes else 0.0,
        "max_degree": max(degs) if degs else 0,
        "avg_degree": sum(degs) / len(degs) if degs else 0.0,
        "parity_defect_score": parity_defect_score,
    }

# ------------------------------------------------------------
# 6. Toy analytic surrogate
#
# defect = missing future flag points
#        + parity penalty
#        + sign penalty
# ------------------------------------------------------------

def state_defect(state, i):
    O, S = state
    future_flag = flag[i:]
    missing_future = sum(1 for p in future_flag if p not in O)
    parity_penalty = 1 if state_parity(state) == -1 else 0
    sign_penalty = 1 if len(S) == 2 else 0
    return missing_future + parity_penalty + sign_penalty

def analytic_stats(states, i):
    defects = [state_defect(st, i) for st in states]
    pole_order = max(defects) if defects else 0
    principal_mass = sum(1 for d in defects if d == pole_order)
    avg_defect = sum(defects) / len(defects) if defects else 0.0
    return {
        "toy_pole_order": pole_order,
        "toy_principal_mass": principal_mass,
        "avg_defect": avg_defect,
    }

# ------------------------------------------------------------
# 7. Terminal alignment
# ------------------------------------------------------------

def alignment_stats(states):
    ints_Ostar = [len(O.intersection(O_star)) for (O, S) in states]
    ints_tail = [len(O.intersection(tail)) for (O, S) in states]
    avg_sign_size = sum(len(S) for (O, S) in states) / len(states) if states else 0.0
    return {
        "avg_intersection_with_Ostar": sum(ints_Ostar) / len(ints_Ostar) if ints_Ostar else 0.0,
        "avg_intersection_with_tail": sum(ints_tail) / len(ints_tail) if ints_tail else 0.0,
        "avg_sign_size": avg_sign_size,
    }

# ------------------------------------------------------------
# 8. Stage summary
# ------------------------------------------------------------

rows = []
stage_state_store = {}

for i in range(6):
    states = stage_states(i)
    stage_state_store[i] = states

    parities = [state_parity(st) for st in states]
    plus = sum(1 for x in parities if x == 1)
    minus = sum(1 for x in parities if x == -1)

    row = {
        "stage": i,
        "#states": len(states),
        "parity_plus": plus,
        "parity_minus": minus,
    }
    row.update(groupoid_stats(states))
    row.update(analytic_stats(states, i))
    row.update(alignment_stats(states))
    rows.append(row)

lc_summary_df = pd.DataFrame(rows)

print("\n=== Leech/Conway-light stage summary ===")
print(lc_summary_df.to_string(index=False))

# ------------------------------------------------------------
# 9. Deltas
# ------------------------------------------------------------

lc_delta_df = lc_summary_df.copy()
for col in [
    "#states",
    "component_count",
    "parity_defect_score",
    "toy_pole_order",
    "toy_principal_mass",
    "avg_defect",
    "avg_intersection_with_Ostar",
    "avg_intersection_with_tail",
    "avg_sign_size",
]:
    lc_delta_df[f"delta_{col}"] = lc_delta_df[col].diff()

print("\n=== Leech/Conway-light stage differences ===")
show_cols = ["stage"] + [c for c in lc_delta_df.columns if c.startswith("delta_")]
print(lc_delta_df[show_cols].to_string(index=False))

# ------------------------------------------------------------
# 10. Export tables
# ------------------------------------------------------------

lc_summary_df.to_csv(os.path.join(OUTDIR2, "leech_conway_light_stage_summary.csv"), index=False)
lc_delta_df.to_csv(os.path.join(OUTDIR2, "leech_conway_light_stage_deltas.csv"), index=False)

write_latex_table(
    lc_summary_df,
    filename="leech_conway_light_stage_summary.tex",
    caption="Stage summary for the Leech/Conway-light signed-octad benchmark.",
    label="tab:leech_conway_light_stage_summary",
    index=False
)

write_latex_table(
    lc_delta_df[show_cols],
    filename="leech_conway_light_stage_deltas.tex",
    caption="Stage differences for the Leech/Conway-light signed-octad benchmark.",
    label="tab:leech_conway_light_stage_deltas",
    index=False
)

# ------------------------------------------------------------
# 11. Figures
# ------------------------------------------------------------

plt.figure(figsize=(7, 4))
plt.plot(lc_summary_df["stage"], lc_summary_df["#states"], marker="o")
plt.xlabel("Stage")
plt.ylabel("Number of signed states")
plt.title("Leech/Conway-light: state count")
plt.grid(True)
save_current_figure("leech_conway_light_state_count")
plt.show()

plt.figure(figsize=(7, 4))
plt.plot(lc_summary_df["stage"], lc_summary_df["toy_pole_order"], marker="o", label="toy pole order")
plt.plot(lc_summary_df["stage"], lc_summary_df["avg_defect"], marker="s", label="average defect")
plt.plot(lc_summary_df["stage"], lc_summary_df["toy_principal_mass"], marker="^", label="principal mass")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title("Leech/Conway-light: analytic proxy diagnostics")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_analytic")
plt.show()

plt.figure(figsize=(7, 4))
plt.plot(lc_summary_df["stage"], lc_summary_df["component_count"], marker="o", label="component count")
plt.plot(lc_summary_df["stage"], lc_summary_df["avg_degree"], marker="s", label="average degree")
plt.plot(lc_summary_df["stage"], lc_summary_df["parity_defect_score"], marker="^", label="parity defect score")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title("Leech/Conway-light: groupoid proxy diagnostics")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_groupoid")
plt.show()

plt.figure(figsize=(7, 4))
plt.plot(lc_summary_df["stage"], lc_summary_df["parity_plus"], marker="o", label="parity +")
plt.plot(lc_summary_df["stage"], lc_summary_df["parity_minus"], marker="s", label="parity -")
plt.xlabel("Stage")
plt.ylabel("Count")
plt.title("Leech/Conway-light: parity split")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_parity_split")
plt.show()

plt.figure(figsize=(8, 4.5))
plt.plot(lc_delta_df["stage"], lc_delta_df["delta_toy_pole_order"], marker="o", label=r"$\Delta$ toy pole order")
plt.plot(lc_delta_df["stage"], lc_delta_df["delta_avg_defect"], marker="s", label=r"$\Delta$ avg defect")
plt.plot(lc_delta_df["stage"], lc_delta_df["delta_avg_intersection_with_Ostar"], marker="^", label=r"$\Delta$ avg $|O\cap O_*|$")
plt.plot(lc_delta_df["stage"], lc_delta_df["delta_parity_defect_score"], marker="d", label=r"$\Delta$ parity defect score")
plt.xlabel("Stage")
plt.ylabel("Stage difference")
plt.title("Leech/Conway-light: synchronised stage differences")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_sync")
plt.show()

print("\nSaved files:")
for fn in sorted(os.listdir(OUTDIR2)):
    print(" -", fn)
```

```python
import itertools
import math
import os
from collections import deque, defaultdict
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# CONFIGURATION
# ============================================================

OUTDIR = "leech_conway_light_tightened_outputs"
os.makedirs(OUTDIR, exist_ok=True)

SAVE_PDF = True
SAVE_PNG = True
DPI = 200

# Tightening controls
SUPPORT_INTERSECTION_THRESHOLD = 6   # use 6; if still too connected, try 7
REQUIRE_EQUAL_SIGN_SIZE_ACROSS_SUPPORT = True
REQUIRE_EQUAL_TAIL_PARITY_ACROSS_SUPPORT = True
FINAL_STAGE_COLLAPSE = True
USE_STAGE_DEPENDENT_DECORATION_RESTRICTION = True

# ============================================================
# Utilities
# ============================================================

def save_current_figure(stem: str):
    if SAVE_PDF:
        plt.savefig(os.path.join(OUTDIR, f"{stem}.pdf"), bbox_inches="tight")
    if SAVE_PNG:
        plt.savefig(os.path.join(OUTDIR, f"{stem}.png"), dpi=DPI, bbox_inches="tight")

def write_latex_table(df: pd.DataFrame, filename: str, caption: str, label: str, index: bool = False):
    tex = df.to_latex(index=index, escape=False, float_format=lambda x: f"{x:.6f}")
    wrapped = (
        "\\begin{table}[t]\n"
        "\\centering\n"
        f"\\caption{{{caption}}}\n"
        f"\\label{{{label}}}\n"
        "\\small\n"
        + tex +
        "\\end{table}\n"
    )
    with open(os.path.join(OUTDIR, filename), "w", encoding="utf-8") as f:
        f.write(wrapped)

# ============================================================
# 0. Exact G24 construction (pure Python)
#    Build G23 from the standard generator polynomial, then extend.
# ============================================================

def carryless_mul(a: int, b: int) -> int:
    out = 0
    x = a
    y = b
    while y:
        if y & 1:
            out ^= x
        x <<= 1
        y >>= 1
    return out

# g(x)=x^11+x^10+x^6+x^5+x^4+x^2+1
g_poly = sum(1 << k for k in [11, 10, 6, 5, 4, 2, 0])

def int_to_support_24(code23: int):
    supp = {i + 1 for i in range(23) if (code23 >> i) & 1}
    if len(supp) % 2 == 1:
        supp.add(24)
    return frozenset(supp)

def build_g24_supports():
    codewords = set()
    for m in range(1 << 12):
        c23 = carryless_mul(m, g_poly)
        if c23 >= (1 << 23):
            raise ValueError("Unexpected degree overflow in G23 encoding.")
        codewords.add(int_to_support_24(c23))
    return sorted(codewords, key=lambda S: (len(S), tuple(sorted(S))))

codewords_24 = build_g24_supports()
assert len(codewords_24) == 4096, "Expected 4096 codewords."

octads = [S for S in codewords_24 if len(S) == 8]
assert len(octads) == 759, "Expected 759 octads."

print("number of G24 codewords =", len(codewords_24))
print("number of octads =", len(octads))

# ============================================================
# 1. Choose a 5-point flag with unique completion octad
# ============================================================

def find_unique_5set(octads):
    trial = (1, 2, 3, 4, 5)
    candidates = [O for O in octads if set(trial).issubset(O)]
    if len(candidates) == 1:
        return trial, candidates[0]

    for F in itertools.combinations(range(1, 25), 5):
        candidates = [O for O in octads if set(F).issubset(O)]
        if len(candidates) == 1:
            return F, candidates[0]

    raise RuntimeError("No unique 5-set found.")

flag, O_star = find_unique_5set(octads)
tail = frozenset(set(O_star).difference(flag))

print("flag =", flag)
print("terminal octad O_* =", sorted(O_star))
print("tail T = O_* \\ flag =", sorted(tail))

# ============================================================
# 2. Stage filtration
# ============================================================

stage_octads = {}
for i in range(6):
    Fi = set(flag[:i])
    stage_octads[i] = [O for O in octads if Fi.issubset(O)]
    print(f"stage {i}: {len(stage_octads[i])} octads")

# ============================================================
# 3. Signed decorations (tightened)
#
# State = (O, S) where O is an octad and S is a small sign decoration.
#
# Tightening strategy:
# - stage 5: only empty decoration (terminal collapse)
# - stage 4: only empty decoration or 2-subsets inside O∩tail
# - stages 0..3: empty decoration plus 2-subsets of O∩O_* that touch tail
# ============================================================

def decoration_candidates(O, i):
    U = sorted(O.intersection(O_star))
    U_tail = sorted(O.intersection(tail))

    if FINAL_STAGE_COLLAPSE and i == 5:
        return [frozenset()]

    decos = [frozenset()]

    if USE_STAGE_DEPENDENT_DECORATION_RESTRICTION:
        if i == 4:
            # very tight: only tail-supported 2-subsets
            for a, b in itertools.combinations(U_tail, 2):
                decos.append(frozenset([a, b]))
            return decos
        else:
            # stages 0..3: allow 2-subsets in U that touch tail
            for a, b in itertools.combinations(U, 2):
                pair = frozenset([a, b])
                if len(pair.intersection(tail)) >= 1:
                    decos.append(pair)
            return decos
    else:
        for a, b in itertools.combinations(U, 2):
            decos.append(frozenset([a, b]))
        return decos

def stage_states(i):
    states = []
    for O in stage_octads[i]:
        for S in decoration_candidates(O, i):
            states.append((O, S))
    return states

# ============================================================
# 4. State parity
#
# support term: |O ∩ tail| mod 2
# sign term: 1 if |S|=2 else 0
# ============================================================

def state_parity(state):
    O, S = state
    support_term = len(O.intersection(tail)) % 2
    sign_term = 1 if len(S) == 2 else 0
    return (-1) ** (support_term + sign_term)

def state_tail_parity(state):
    O, S = state
    return len(O.intersection(tail)) % 2

# ============================================================
# 5. Tightened signed-octad graph
#
# Internal edges:
#   same support, sign decorations differ by exactly 2
#
# Cross-support edges:
#   |O ∩ O'| = SUPPORT_INTERSECTION_THRESHOLD
#   and (optionally) same sign size
#   and (optionally) same tail parity
#   and same full state parity
#
# This is deliberately stricter than the preview version.
# ============================================================

def support_adjacent(O, O2):
    return len(O.intersection(O2)) == SUPPORT_INTERSECTION_THRESHOLD

def build_state_graph(states):
    n = len(states)
    adj = [[] for _ in range(n)]

    support_to_indices = defaultdict(list)
    for idx, (O, S) in enumerate(states):
        support_to_indices[O].append(idx)

    supports = list(support_to_indices.keys())

    # (a) internal sign-flip edges
    for O, idxs in support_to_indices.items():
        for a in range(len(idxs)):
            for b in range(a + 1, len(idxs)):
                ia, ib = idxs[a], idxs[b]
                Sa = states[ia][1]
                Sb = states[ib][1]
                if len(Sa.symmetric_difference(Sb)) == 2:
                    adj[ia].append(ib)
                    adj[ib].append(ia)

    # (b) tightened cross-support edges
    for i in range(len(supports)):
        Oi = supports[i]
        for j in range(i + 1, len(supports)):
            Oj = supports[j]
            if not support_adjacent(Oi, Oj):
                continue

            idxs_i = support_to_indices[Oi]
            idxs_j = support_to_indices[Oj]

            for ia in idxs_i:
                for ib in idxs_j:
                    Oa, Sa = states[ia]
                    Ob, Sb = states[ib]

                    if REQUIRE_EQUAL_SIGN_SIZE_ACROSS_SUPPORT and len(Sa) != len(Sb):
                        continue
                    if REQUIRE_EQUAL_TAIL_PARITY_ACROSS_SUPPORT and state_tail_parity(states[ia]) != state_tail_parity(states[ib]):
                        continue
                    if state_parity(states[ia]) != state_parity(states[ib]):
                        continue

                    adj[ia].append(ib)
                    adj[ib].append(ia)

    return adj

def connected_components(adj):
    n = len(adj)
    seen = [False] * n
    comps = []
    for i in range(n):
        if seen[i]:
            continue
        q = deque([i])
        seen[i] = True
        comp = []
        while q:
            u = q.popleft()
            comp.append(u)
            for v in adj[u]:
                if not seen[v]:
                    seen[v] = True
                    q.append(v)
        comps.append(comp)
    return comps

# ============================================================
# 6. Groupoid proxy stats
# ============================================================

def groupoid_stats(states):
    adj = build_state_graph(states)
    comps = connected_components(adj)
    comp_sizes = [len(c) for c in comps]
    degs = [len(nei) for nei in adj]

    total_edges = 0
    bad_edges = 0
    for i, nei in enumerate(adj):
        for j in nei:
            if j > i:
                total_edges += 1
                if state_parity(states[i]) != state_parity(states[j]):
                    bad_edges += 1

    parity_defect_score = bad_edges / total_edges if total_edges > 0 else 0.0

    return {
        "component_count": len(comps),
        "max_component_size": max(comp_sizes) if comp_sizes else 0,
        "min_component_size": min(comp_sizes) if comp_sizes else 0,
        "avg_component_size": sum(comp_sizes) / len(comp_sizes) if comp_sizes else 0.0,
        "max_degree": max(degs) if degs else 0,
        "avg_degree": sum(degs) / len(degs) if degs else 0.0,
        "parity_defect_score": parity_defect_score,
    }

# ============================================================
# 7. Toy analytic surrogate (tightened)
#
# defect = missing future flag points
#        + parity penalty
#        + sign penalty
#
# weighted principal mass:
#   sum_{d = d_max} 1/(1+|S|)
# ============================================================

def state_defect(state, i):
    O, S = state
    future_flag = flag[i:]
    missing_future = sum(1 for p in future_flag if p not in O)
    parity_penalty = 1 if state_parity(state) == -1 else 0
    sign_penalty = 1 if len(S) == 2 else 0
    return missing_future + parity_penalty + sign_penalty

def weighted_state_mass(state):
    O, S = state
    return 1.0 / (1.0 + len(S))

def analytic_stats(states, i):
    defects = [state_defect(st, i) for st in states]
    dmax = max(defects) if defects else 0
    weighted_principal_mass = sum(
        weighted_state_mass(st) for st, d in zip(states, defects) if d == dmax
    )
    avg_defect = sum(defects) / len(defects) if defects else 0.0
    return {
        "toy_pole_order": dmax,
        "weighted_principal_mass": weighted_principal_mass,
        "avg_defect": avg_defect,
    }

# ============================================================
# 8. Terminal alignment
# ============================================================

def alignment_stats(states):
    ints_Ostar = [len(O.intersection(O_star)) for (O, S) in states]
    ints_tail = [len(O.intersection(tail)) for (O, S) in states]
    avg_sign_size = sum(len(S) for (O, S) in states) / len(states) if states else 0.0
    return {
        "avg_intersection_with_Ostar": sum(ints_Ostar) / len(ints_Ostar) if ints_Ostar else 0.0,
        "avg_intersection_with_tail": sum(ints_tail) / len(ints_tail) if ints_tail else 0.0,
        "avg_sign_size": avg_sign_size,
    }

# ============================================================
# 9. Stage summary
# ============================================================

rows = []
stage_state_store = {}

for i in range(6):
    states = stage_states(i)
    stage_state_store[i] = states

    parities = [state_parity(st) for st in states]
    plus = sum(1 for x in parities if x == 1)
    minus = sum(1 for x in parities if x == -1)

    row = {
        "stage": i,
        "#states": len(states),
        "parity_plus": plus,
        "parity_minus": minus,
    }
    row.update(groupoid_stats(states))
    row.update(analytic_stats(states, i))
    row.update(alignment_stats(states))
    rows.append(row)

summary_df = pd.DataFrame(rows)

print("\n=== Leech/Conway-light tightened stage summary ===")
print(summary_df.to_string(index=False))

# ============================================================
# 10. Deltas
# ============================================================

delta_df = summary_df.copy()
for col in [
    "#states",
    "component_count",
    "parity_defect_score",
    "toy_pole_order",
    "weighted_principal_mass",
    "avg_defect",
    "avg_intersection_with_Ostar",
    "avg_intersection_with_tail",
    "avg_sign_size",
]:
    delta_df[f"delta_{col}"] = delta_df[col].diff()

print("\n=== Leech/Conway-light tightened stage differences ===")
show_cols = ["stage"] + [c for c in delta_df.columns if c.startswith("delta_")]
print(delta_df[show_cols].to_string(index=False))

# ============================================================
# 11. Export tables
# ============================================================

summary_df.to_csv(os.path.join(OUTDIR, "leech_conway_light_tightened_stage_summary.csv"), index=False)
delta_df.to_csv(os.path.join(OUTDIR, "leech_conway_light_tightened_stage_deltas.csv"), index=False)

write_latex_table(
    summary_df,
    filename="leech_conway_light_tightened_stage_summary.tex",
    caption="Stage summary for the tightened Leech/Conway-light signed-octad benchmark.",
    label="tab:leech_conway_light_tightened_stage_summary",
    index=False
)

write_latex_table(
    delta_df[show_cols],
    filename="leech_conway_light_tightened_stage_deltas.tex",
    caption="Stage differences for the tightened Leech/Conway-light signed-octad benchmark.",
    label="tab:leech_conway_light_tightened_stage_deltas",
    index=False
)

# ============================================================
# 12. Figures
# ============================================================

# Fig 1: state count
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["#states"], marker="o")
plt.xlabel("Stage")
plt.ylabel("Number of signed states")
plt.title("Leech/Conway-light tightened: state count")
plt.grid(True)
save_current_figure("leech_conway_light_tightened_state_count")
plt.show()

# Fig 2: analytic diagnostics
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["toy_pole_order"], marker="o", label="toy pole order")
plt.plot(summary_df["stage"], summary_df["avg_defect"], marker="s", label="average defect")
plt.plot(summary_df["stage"], summary_df["weighted_principal_mass"], marker="^", label="weighted principal mass")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title("Leech/Conway-light tightened: analytic diagnostics")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_analytic")
plt.show()

# Fig 3: groupoid diagnostics
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["component_count"], marker="o", label="component count")
plt.plot(summary_df["stage"], summary_df["avg_degree"], marker="s", label="average degree")
plt.plot(summary_df["stage"], summary_df["parity_defect_score"], marker="^", label="parity defect score")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title("Leech/Conway-light tightened: groupoid diagnostics")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_groupoid")
plt.show()

# Fig 4: parity split
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["parity_plus"], marker="o", label="parity +")
plt.plot(summary_df["stage"], summary_df["parity_minus"], marker="s", label="parity -")
plt.xlabel("Stage")
plt.ylabel("Count")
plt.title("Leech/Conway-light tightened: parity split")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_parity_split")
plt.show()

# Fig 5: terminal alignment
plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["avg_intersection_with_Ostar"], marker="o", label=r"avg $|O \cap O_*|$")
plt.plot(summary_df["stage"], summary_df["avg_intersection_with_tail"], marker="s", label=r"avg $|O \cap T|$")
plt.plot(summary_df["stage"], summary_df["avg_sign_size"], marker="^", label="avg sign size")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title("Leech/Conway-light tightened: terminal alignment")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_terminal_alignment")
plt.show()

# Fig 6: synchronised deltas
plt.figure(figsize=(8, 4.5))
plt.plot(delta_df["stage"], delta_df["delta_toy_pole_order"], marker="o", label=r"$\Delta$ toy pole order")
plt.plot(delta_df["stage"], delta_df["delta_avg_defect"], marker="s", label=r"$\Delta$ avg defect")
plt.plot(delta_df["stage"], delta_df["delta_avg_intersection_with_Ostar"], marker="^", label=r"$\Delta$ avg $|O\cap O_*|$")
plt.plot(delta_df["stage"], delta_df["delta_parity_defect_score"], marker="d", label=r"$\Delta$ parity defect score")
plt.plot(delta_df["stage"], delta_df["delta_component_count"], marker="x", label=r"$\Delta$ component count")
plt.xlabel("Stage")
plt.ylabel("Stage difference")
plt.title("Leech/Conway-light tightened: synchronised stage differences")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_sync")
plt.show()

# ============================================================
# 13. LaTeX include snippet
# ============================================================

latex_snippet = r"""
\input{leech_conway_light_tightened_stage_summary.tex}

\input{leech_conway_light_tightened_stage_deltas.tex}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{leech_conway_light_tightened_state_count.pdf}
\caption{State count for the tightened Leech/Conway-light benchmark.}
\label{fig:lc-tight-state-count}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{leech_conway_light_tightened_analytic.pdf}
\caption{Analytic diagnostics for the tightened Leech/Conway-light benchmark.}
\label{fig:lc-tight-analytic}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{leech_conway_light_tightened_groupoid.pdf}
\caption{Groupoid diagnostics for the tightened Leech/Conway-light benchmark.}
\label{fig:lc-tight-groupoid}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{leech_conway_light_tightened_parity_split.pdf}
\caption{Parity split for the tightened Leech/Conway-light benchmark.}
\label{fig:lc-tight-parity}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.75\textwidth]{leech_conway_light_tightened_terminal_alignment.pdf}
\caption{Terminal alignment for the tightened Leech/Conway-light benchmark.}
\label{fig:lc-tight-terminal}
\end{figure}

\begin{figure}[t]
\centering
\includegraphics[width=.80\textwidth]{leech_conway_light_tightened_sync.pdf}
\caption{Synchronised stage differences for the tightened Leech/Conway-light benchmark.}
\label{fig:lc-tight-sync}
\end{figure}
"""

with open(os.path.join(OUTDIR, "leech_conway_light_tightened_include_snippet.tex"), "w", encoding="utf-8") as f:
    f.write(latex_snippet)

print("\nSaved files:")
for fn in sorted(os.listdir(OUTDIR)):
    print(" -", fn)
```

```python
import itertools
import os
from collections import deque, defaultdict
import pandas as pd
import matplotlib.pyplot as plt

# ============================================================
# CONFIGURATION
# ============================================================

OUTDIR = "leech_conway_light_tightened_v2_outputs"
os.makedirs(OUTDIR, exist_ok=True)

SAVE_PDF = True
SAVE_PNG = True
DPI = 200

# v2: slightly relaxed cross-support rule
SUPPORT_INTERSECTION_THRESHOLD = 6
REQUIRE_EQUAL_SIGN_SIZE_ACROSS_SUPPORT = False
REQUIRE_EQUAL_TAIL_PARITY_ACROSS_SUPPORT = True
FINAL_STAGE_COLLAPSE = True
USE_STAGE_DEPENDENT_DECORATION_RESTRICTION = True

# ============================================================
# Utilities
# ============================================================

def save_current_figure(stem: str):
    if SAVE_PDF:
        plt.savefig(os.path.join(OUTDIR, f"{stem}.pdf"), bbox_inches="tight")
    if SAVE_PNG:
        plt.savefig(os.path.join(OUTDIR, f"{stem}.png"), dpi=DPI, bbox_inches="tight")

def write_latex_table(df: pd.DataFrame, filename: str, caption: str, label: str, index: bool = False):
    tex = df.to_latex(index=index, escape=False, float_format=lambda x: f"{x:.6f}")
    wrapped = (
        "\\begin{table}[t]\n"
        "\\centering\n"
        f"\\caption{{{caption}}}\n"
        f"\\label{{{label}}}\n"
        "\\small\n"
        + tex +
        "\\end{table}\n"
    )
    with open(os.path.join(OUTDIR, filename), "w", encoding="utf-8") as f:
        f.write(wrapped)

# ============================================================
# 0. Exact G24 construction
# ============================================================

def carryless_mul(a: int, b: int) -> int:
    out = 0
    x = a
    y = b
    while y:
        if y & 1:
            out ^= x
        x <<= 1
        y >>= 1
    return out

# g(x)=x^11+x^10+x^6+x^5+x^4+x^2+1
g_poly = sum(1 << k for k in [11, 10, 6, 5, 4, 2, 0])

def int_to_support_24(code23: int):
    supp = {i + 1 for i in range(23) if (code23 >> i) & 1}
    if len(supp) % 2 == 1:
        supp.add(24)
    return frozenset(supp)

def build_g24_supports():
    codewords = set()
    for m in range(1 << 12):
        c23 = carryless_mul(m, g_poly)
        if c23 >= (1 << 23):
            raise ValueError("Unexpected degree overflow in G23 encoding.")
        codewords.add(int_to_support_24(c23))
    return sorted(codewords, key=lambda S: (len(S), tuple(sorted(S))))

codewords_24 = build_g24_supports()
assert len(codewords_24) == 4096, "Expected 4096 codewords."

octads = [S for S in codewords_24 if len(S) == 8]
assert len(octads) == 759, "Expected 759 octads."

print("number of G24 codewords =", len(codewords_24))
print("number of octads =", len(octads))

# ============================================================
# 1. Choose a 5-point flag with unique completion octad
# ============================================================

def find_unique_5set(octads):
    trial = (1, 2, 3, 4, 5)
    candidates = [O for O in octads if set(trial).issubset(O)]
    if len(candidates) == 1:
        return trial, candidates[0]

    for F in itertools.combinations(range(1, 25), 5):
        candidates = [O for O in octads if set(F).issubset(O)]
        if len(candidates) == 1:
            return F, candidates[0]

    raise RuntimeError("No unique 5-set found.")

flag, O_star = find_unique_5set(octads)
tail = frozenset(set(O_star).difference(flag))

print("flag =", flag)
print("terminal octad O_* =", sorted(O_star))
print("tail T = O_* \\ flag =", sorted(tail))

# ============================================================
# 2. Stage filtration
# ============================================================

stage_octads = {}
for i in range(6):
    Fi = set(flag[:i])
    stage_octads[i] = [O for O in octads if Fi.issubset(O)]
    print(f"stage {i}: {len(stage_octads[i])} octads")

# ============================================================
# 3. Signed decorations
# ============================================================

def decoration_candidates(O, i):
    U = sorted(O.intersection(O_star))
    U_tail = sorted(O.intersection(tail))

    if FINAL_STAGE_COLLAPSE and i == 5:
        return [frozenset()]

    decos = [frozenset()]

    if USE_STAGE_DEPENDENT_DECORATION_RESTRICTION:
        if i == 4:
            for a, b in itertools.combinations(U_tail, 2):
                decos.append(frozenset([a, b]))
            return decos
        else:
            for a, b in itertools.combinations(U, 2):
                pair = frozenset([a, b])
                if len(pair.intersection(tail)) >= 1:
                    decos.append(pair)
            return decos
    else:
        for a, b in itertools.combinations(U, 2):
            decos.append(frozenset([a, b]))
        return decos

def stage_states(i):
    states = []
    for O in stage_octads[i]:
        for S in decoration_candidates(O, i):
            states.append((O, S))
    return states

# ============================================================
# 4. State parity
# ============================================================

def state_parity(state):
    O, S = state
    support_term = len(O.intersection(tail)) % 2
    sign_term = 1 if len(S) == 2 else 0
    return (-1) ** (support_term + sign_term)

def state_tail_parity(state):
    O, S = state
    return len(O.intersection(tail)) % 2

# ============================================================
# 5. Tightened graph, v2 relaxed across supports
# ============================================================

def support_adjacent(O, O2):
    return len(O.intersection(O2)) == SUPPORT_INTERSECTION_THRESHOLD

def build_state_graph(states):
    n = len(states)
    adj = [[] for _ in range(n)]

    support_to_indices = defaultdict(list)
    for idx, (O, S) in enumerate(states):
        support_to_indices[O].append(idx)

    supports = list(support_to_indices.keys())

    # internal sign-flip edges
    for O, idxs in support_to_indices.items():
        for a in range(len(idxs)):
            for b in range(a + 1, len(idxs)):
                ia, ib = idxs[a], idxs[b]
                Sa = states[ia][1]
                Sb = states[ib][1]
                if len(Sa.symmetric_difference(Sb)) == 2:
                    adj[ia].append(ib)
                    adj[ib].append(ia)

    # cross-support edges
    for i in range(len(supports)):
        Oi = supports[i]
        for j in range(i + 1, len(supports)):
            Oj = supports[j]
            if not support_adjacent(Oi, Oj):
                continue

            idxs_i = support_to_indices[Oi]
            idxs_j = support_to_indices[Oj]

            for ia in idxs_i:
                for ib in idxs_j:
                    Oa, Sa = states[ia]
                    Ob, Sb = states[ib]

                    if REQUIRE_EQUAL_SIGN_SIZE_ACROSS_SUPPORT and len(Sa) != len(Sb):
                        continue
                    if REQUIRE_EQUAL_TAIL_PARITY_ACROSS_SUPPORT and state_tail_parity(states[ia]) != state_tail_parity(states[ib]):
                        continue
                    if state_parity(states[ia]) != state_parity(states[ib]):
                        continue

                    adj[ia].append(ib)
                    adj[ib].append(ia)

    return adj

def connected_components(adj):
    n = len(adj)
    seen = [False] * n
    comps = []
    for i in range(n):
        if seen[i]:
            continue
        q = deque([i])
        seen[i] = True
        comp = []
        while q:
            u = q.popleft()
            comp.append(u)
            for v in adj[u]:
                if not seen[v]:
                    seen[v] = True
                    q.append(v)
        comps.append(comp)
    return comps

# ============================================================
# 6. Groupoid proxy stats
# ============================================================

def groupoid_stats(states):
    adj = build_state_graph(states)
    comps = connected_components(adj)
    comp_sizes = [len(c) for c in comps]
    degs = [len(nei) for nei in adj]

    total_edges = 0
    bad_edges = 0
    for i, nei in enumerate(adj):
        for j in nei:
            if j > i:
                total_edges += 1
                if state_parity(states[i]) != state_parity(states[j]):
                    bad_edges += 1

    parity_defect_score = bad_edges / total_edges if total_edges > 0 else 0.0

    return {
        "component_count": len(comps),
        "max_component_size": max(comp_sizes) if comp_sizes else 0,
        "min_component_size": min(comp_sizes) if comp_sizes else 0,
        "avg_component_size": sum(comp_sizes) / len(comp_sizes) if comp_sizes else 0.0,
        "max_degree": max(degs) if degs else 0,
        "avg_degree": sum(degs) / len(degs) if degs else 0.0,
        "parity_defect_score": parity_defect_score,
    }

# ============================================================
# 7. Analytic surrogate v2
#   final stage defect forced to 0
# ============================================================

def state_defect(state, i):
    O, S = state
    future_flag = flag[i:]
    missing_future = sum(1 for p in future_flag if p not in O)

    if i == 5:
        parity_penalty = 0
        sign_penalty = 0
    else:
        parity_penalty = 1 if state_parity(state) == -1 else 0
        sign_penalty = 1 if len(S) == 2 else 0

    return missing_future + parity_penalty + sign_penalty

def weighted_state_mass(state):
    O, S = state
    return 1.0 / (1.0 + len(S))

def analytic_stats(states, i):
    defects = [state_defect(st, i) for st in states]
    dmax = max(defects) if defects else 0
    weighted_principal_mass = sum(
        weighted_state_mass(st) for st, d in zip(states, defects) if d == dmax
    )
    avg_defect = sum(defects) / len(defects) if defects else 0.0
    return {
        "toy_pole_order": dmax,
        "weighted_principal_mass": weighted_principal_mass,
        "avg_defect": avg_defect,
    }

# ============================================================
# 8. Terminal alignment
# ============================================================

def alignment_stats(states):
    ints_Ostar = [len(O.intersection(O_star)) for (O, S) in states]
    ints_tail = [len(O.intersection(tail)) for (O, S) in states]
    avg_sign_size = sum(len(S) for (O, S) in states) / len(states) if states else 0.0
    return {
        "avg_intersection_with_Ostar": sum(ints_Ostar) / len(ints_Ostar) if ints_Ostar else 0.0,
        "avg_intersection_with_tail": sum(ints_tail) / len(ints_tail) if ints_tail else 0.0,
        "avg_sign_size": avg_sign_size,
    }

# ============================================================
# 9. Stage summary
# ============================================================

rows = []
stage_state_store = {}

for i in range(6):
    states = stage_states(i)
    stage_state_store[i] = states

    parities = [state_parity(st) for st in states]
    plus = sum(1 for x in parities if x == 1)
    minus = sum(1 for x in parities if x == -1)

    row = {
        "stage": i,
        "#states": len(states),
        "parity_plus": plus,
        "parity_minus": minus,
    }
    row.update(groupoid_stats(states))
    row.update(analytic_stats(states, i))
    row.update(alignment_stats(states))
    rows.append(row)

summary_df = pd.DataFrame(rows)

print("\n=== Leech/Conway-light tightened v2 stage summary ===")
print(summary_df.to_string(index=False))

# ============================================================
# 10. Deltas
# ============================================================

delta_df = summary_df.copy()
for col in [
    "#states",
    "component_count",
    "parity_defect_score",
    "toy_pole_order",
    "weighted_principal_mass",
    "avg_defect",
    "avg_intersection_with_Ostar",
    "avg_intersection_with_tail",
    "avg_sign_size",
]:
    delta_df[f"delta_{col}"] = delta_df[col].diff()

print("\n=== Leech/Conway-light tightened v2 stage differences ===")
show_cols = ["stage"] + [c for c in delta_df.columns if c.startswith("delta_")]
print(delta_df[show_cols].to_string(index=False))

# ============================================================
# 11. Export
# ============================================================

summary_df.to_csv(os.path.join(OUTDIR, "leech_conway_light_tightened_v2_stage_summary.csv"), index=False)
delta_df.to_csv(os.path.join(OUTDIR, "leech_conway_light_tightened_v2_stage_deltas.csv"), index=False)

write_latex_table(
    summary_df,
    filename="leech_conway_light_tightened_v2_stage_summary.tex",
    caption="Stage summary for the tightened Leech/Conway-light benchmark (v2).",
    label="tab:leech_conway_light_tightened_v2_stage_summary",
    index=False
)

write_latex_table(
    delta_df[show_cols],
    filename="leech_conway_light_tightened_v2_stage_deltas.tex",
    caption="Stage differences for the tightened Leech/Conway-light benchmark (v2).",
    label="tab:leech_conway_light_tightened_v2_stage_deltas",
    index=False
)

# ============================================================
# 12. Figures
# ============================================================

plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["#states"], marker="o")
plt.xlabel("Stage")
plt.ylabel("Number of signed states")
plt.title("Leech/Conway-light tightened v2: state count")
plt.grid(True)
save_current_figure("leech_conway_light_tightened_v2_state_count")
plt.show()

plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["toy_pole_order"], marker="o", label="toy pole order")
plt.plot(summary_df["stage"], summary_df["avg_defect"], marker="s", label="average defect")
plt.plot(summary_df["stage"], summary_df["weighted_principal_mass"], marker="^", label="weighted principal mass")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title("Leech/Conway-light tightened v2: analytic diagnostics")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_v2_analytic")
plt.show()

plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["component_count"], marker="o", label="component count")
plt.plot(summary_df["stage"], summary_df["avg_degree"], marker="s", label="average degree")
plt.plot(summary_df["stage"], summary_df["parity_defect_score"], marker="^", label="parity defect score")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title("Leech/Conway-light tightened v2: groupoid diagnostics")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_v2_groupoid")
plt.show()

plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["parity_plus"], marker="o", label="parity +")
plt.plot(summary_df["stage"], summary_df["parity_minus"], marker="s", label="parity -")
plt.xlabel("Stage")
plt.ylabel("Count")
plt.title("Leech/Conway-light tightened v2: parity split")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_v2_parity_split")
plt.show()

plt.figure(figsize=(7, 4))
plt.plot(summary_df["stage"], summary_df["avg_intersection_with_Ostar"], marker="o", label=r"avg $|O \cap O_*|$")
plt.plot(summary_df["stage"], summary_df["avg_intersection_with_tail"], marker="s", label=r"avg $|O \cap T|$")
plt.plot(summary_df["stage"], summary_df["avg_sign_size"], marker="^", label="avg sign size")
plt.xlabel("Stage")
plt.ylabel("Value")
plt.title("Leech/Conway-light tightened v2: terminal alignment")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_v2_terminal_alignment")
plt.show()

plt.figure(figsize=(8, 4.5))
plt.plot(delta_df["stage"], delta_df["delta_toy_pole_order"], marker="o", label=r"$\Delta$ toy pole order")
plt.plot(delta_df["stage"], delta_df["delta_avg_defect"], marker="s", label=r"$\Delta$ avg defect")
plt.plot(delta_df["stage"], delta_df["delta_avg_intersection_with_Ostar"], marker="^", label=r"$\Delta$ avg $|O\cap O_*|$")
plt.plot(delta_df["stage"], delta_df["delta_parity_defect_score"], marker="d", label=r"$\Delta$ parity defect score")
plt.plot(delta_df["stage"], delta_df["delta_component_count"], marker="x", label=r"$\Delta$ component count")
plt.xlabel("Stage")
plt.ylabel("Stage difference")
plt.title("Leech/Conway-light tightened v2: synchronised stage differences")
plt.grid(True)
plt.legend()
save_current_figure("leech_conway_light_tightened_v2_sync")
plt.show()

print("\nSaved files:")
for fn in sorted(os.listdir(OUTDIR)):
    print(" -", fn)
```
