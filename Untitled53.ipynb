```python
# make_koibuchi_baseline_figs.py
# Generate clean Fig10.pdf / Fig11.pdf by auto-cropping the white figure region
# from provided screenshot PNG/PDF assets.

import argparse
from pathlib import Path

import numpy as np
from PIL import Image

try:
    import fitz  # PyMuPDF
except Exception:
    fitz = None


def load_image_any(path: Path, pdf_zoom: float = 2.0) -> Image.Image:
    """Load an image from PNG/JPG or render first page of a PDF."""
    suffix = path.suffix.lower()
    if suffix in [".png", ".jpg", ".jpeg", ".webp"]:
        return Image.open(path).convert("RGB")

    if suffix == ".pdf":
        if fitz is None:
            raise RuntimeError("PyMuPDF (fitz) is required to render PDFs. Install pymupdf.")
        doc = fitz.open(str(path))
        page = doc.load_page(0)
        mat = fitz.Matrix(pdf_zoom, pdf_zoom)
        pix = page.get_pixmap(matrix=mat, alpha=False)
        doc.close()
        arr = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.height, pix.width, pix.n)
        return Image.fromarray(arr).convert("RGB")

    raise ValueError(f"Unsupported input type: {path}")


def find_white_bbox(img: Image.Image, thresh: int = 235, frac: float = 0.02) -> tuple[int, int, int, int]:
    """
    Find bounding box of dominant 'white' region.
    Uses row/col occupancy thresholds to be robust to noise.
    """
    gray = np.array(img.convert("L"))
    mask = gray >= thresh  # white-ish pixels

    h, w = mask.shape
    row_sum = mask.sum(axis=1)
    col_sum = mask.sum(axis=0)

    row_min = int(w * frac)
    col_min = int(h * frac)

    rows = np.where(row_sum > row_min)[0]
    cols = np.where(col_sum > col_min)[0]

    if len(rows) == 0 or len(cols) == 0:
        raise RuntimeError(
            "Auto-crop failed (no dominant white region detected). "
            "Try lowering --thresh (e.g. 220) or increase PDF zoom."
        )

    y0, y1 = int(rows.min()), int(rows.max()) + 1
    x0, x1 = int(cols.min()), int(cols.max()) + 1
    return x0, y0, x1, y1


def crop_with_margin(img: Image.Image, bbox: tuple[int, int, int, int], margin: int = 30) -> Image.Image:
    x0, y0, x1, y1 = bbox
    x0 = max(0, x0 - margin)
    y0 = max(0, y0 - margin)
    x1 = min(img.width, x1 + margin)
    y1 = min(img.height, y1 + margin)
    return img.crop((x0, y0, x1, y1))


def save_as_pdf(img: Image.Image, out_pdf: Path, dpi: int = 300) -> None:
    out_pdf.parent.mkdir(parents=True, exist_ok=True)
    img = img.convert("RGB")
    img.save(str(out_pdf), "PDF", resolution=float(dpi))


def process_one(in_path: Path, out_pdf: Path, thresh: int, frac: float, margin: int, dpi: int, pdf_zoom: float,
                debug_png: Path | None = None) -> None:
    img = load_image_any(in_path, pdf_zoom=pdf_zoom)
    bbox = find_white_bbox(img, thresh=thresh, frac=frac)
    cropped = crop_with_margin(img, bbox, margin=margin)
    save_as_pdf(cropped, out_pdf, dpi=dpi)
    if debug_png is not None:
        debug_png.parent.mkdir(parents=True, exist_ok=True)
        cropped.save(str(debug_png))


def main():
    ap = argparse.ArgumentParser()
    ap.add_argument("--fig10_in", type=str, required=True,
                    help="Input for Fig10 (recommended: Fig10_crop_82.png). PDF/PNG both ok.")
    ap.add_argument("--fig11_in", type=str, required=True,
                    help="Input for Fig11 (recommended: Fig11_crop_83.png). PDF/PNG both ok.")
    ap.add_argument("--out_dir", type=str, default=".",
                    help="Output directory for Fig10.pdf and Fig11.pdf")
    ap.add_argument("--thresh", type=int, default=235,
                    help="White threshold (lower if auto-crop fails). Typical: 220-240")
    ap.add_argument("--frac", type=float, default=0.02,
                    help="Row/col occupancy fraction for bbox detection. Typical: 0.01-0.05")
    ap.add_argument("--margin", type=int, default=30,
                    help="Margin (pixels) to expand bbox")
    ap.add_argument("--dpi", type=int, default=300,
                    help="PDF raster DPI")
    ap.add_argument("--pdf_zoom", type=float, default=2.0,
                    help="PDF render zoom if inputs are PDFs")
    ap.add_argument("--debug", action="store_true",
                    help="Also save debug PNGs (cropped) next to PDFs")
    args = ap.parse_args()

    out_dir = Path(args.out_dir)
    fig10_out = out_dir / "Fig10.pdf"
    fig11_out = out_dir / "Fig11.pdf"

    dbg10 = out_dir / "Fig10_cropped.png" if args.debug else None
    dbg11 = out_dir / "Fig11_cropped.png" if args.debug else None

    process_one(Path(args.fig10_in), fig10_out, args.thresh, args.frac, args.margin, args.dpi, args.pdf_zoom, dbg10)
    process_one(Path(args.fig11_in), fig11_out, args.thresh, args.frac, args.margin, args.dpi, args.pdf_zoom, dbg11)

    print(f"[OK] Wrote: {fig10_out}")
    print(f"[OK] Wrote: {fig11_out}")
    if args.debug:
        print(f"[DBG] Wrote: {dbg10}")
        print(f"[DBG] Wrote: {dbg11}")


if __name__ == "__main__":
    main()
```

```python
# === Colab 1-cell: generate FG_BDI_ZT_Map.pdf without argparse (safe for notebooks) ===
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# ----------------------------
# 0) Inputs (edit here)
# ----------------------------
BOUNDARY_CSV = Path("fg_boundary_points.csv")   # <-- あなたの境界点CSV（列: lambda,rho）に置換可
OUT_PDF      = Path("FG_BDI_ZT_Map.pdf")
OUT_GRID_CSV = Path("FG_BDI_ZT_Map_grid.csv")

LAM_MIN, LAM_MAX = 0.0, 4.0
RHO_MIN, RHO_MAX = 0.0, 1.0
GRID_N_LAM, GRID_N_RHO = 320, 240

DEG = 4                         # implicit poly degree
U_LAM, U_RHO = 1.0, 0.0          # protocol direction u (default: lambda sweep)
EPS_DIST   = 0.05                # distance scale
DELTA_TRANS = 0.05               # transversality scale

# ----------------------------
# 1) Prepare boundary CSV (DEMO if missing)
# ----------------------------
if not BOUNDARY_CSV.exists():
    lam = np.linspace(LAM_MIN, LAM_MAX, 140)
    # DEMO boundary curve (placeholder): replace by your digitized / computed boundary CSV later
    rho = 0.15 + 0.70/(1.0 + np.exp(-2.2*(lam-2.0)))  # sigmoid-like boundary
    rho = np.clip(rho, RHO_MIN, RHO_MAX)
    pd.DataFrame({"lambda": lam, "rho": rho}).to_csv(BOUNDARY_CSV, index=False)
    print(f"[DEMO] {BOUNDARY_CSV} not found -> wrote a demo boundary CSV. Replace with your real boundary points later.")
else:
    print(f"[OK] using boundary CSV: {BOUNDARY_CSV}")

# ----------------------------
# 2) Fit implicit polynomial f(lambda,rho)=0 to boundary points (SVD)
# ----------------------------
def monomial_powers(deg: int):
    pw = []
    for i in range(deg + 1):
        for j in range(deg + 1 - i):
            pw.append((i, j))
    return pw

def design_matrix(x, y, powers):
    A = np.zeros((len(x), len(powers)), dtype=float)
    for m, (i, j) in enumerate(powers):
        A[:, m] = (x ** i) * (y ** j)
    return A

dfb = pd.read_csv(BOUNDARY_CSV)
assert {"lambda","rho"}.issubset(dfb.columns), "CSV must contain columns: lambda, rho"
x = dfb["lambda"].to_numpy(float)
y = dfb["rho"].to_numpy(float)

powers = monomial_powers(DEG)
A = design_matrix(x, y, powers)
_, _, Vt = np.linalg.svd(A, full_matrices=False)
a = Vt[-1, :]
a = a / (np.linalg.norm(a) + 1e-12)

def eval_f_and_grad(LAM, RHO, powers, a):
    f = np.zeros_like(LAM, dtype=float)
    dlam = np.zeros_like(LAM, dtype=float)
    drho = np.zeros_like(LAM, dtype=float)
    for (i, j), c in zip(powers, a):
        f += c * (LAM**i) * (RHO**j)
        if i > 0:
            dlam += c * i * (LAM**(i-1)) * (RHO**j)
        if j > 0:
            drho += c * j * (LAM**i) * (RHO**(j-1))
    return f, dlam, drho

lam_grid = np.linspace(LAM_MIN, LAM_MAX, GRID_N_LAM)
rho_grid = np.linspace(RHO_MIN, RHO_MAX, GRID_N_RHO)
LAM, RHO = np.meshgrid(lam_grid, rho_grid)

f, dlam, drho = eval_f_and_grad(LAM, RHO, powers, a)
grad_norm = np.sqrt(dlam**2 + drho**2) + 1e-12

# implicit distance proxy
dist = np.abs(f) / grad_norm

# directional transversality magnitude |<∇f, u>|
u_norm = np.sqrt(U_LAM**2 + U_RHO**2) + 1e-12
u_l, u_r = U_LAM/u_norm, U_RHO/u_norm
trans = np.abs(dlam*u_l + drho*u_r)

# composite risk score (near boundary AND low transversality)
risk = np.exp(-dist / max(EPS_DIST, 1e-12)) * np.exp(-trans / max(DELTA_TRANS, 1e-12))

# ----------------------------
# 3) Save grid CSV + plot PDF
# ----------------------------
pd.DataFrame({
    "lambda": LAM.ravel(),
    "rho": RHO.ravel(),
    "f": f.ravel(),
    "dist_proxy": dist.ravel(),
    "trans_u": trans.ravel(),
    "risk": risk.ravel(),
}).to_csv(OUT_GRID_CSV, index=False)

fig = plt.figure(figsize=(10.5, 4.8))
ax = plt.gca()

im = ax.imshow(
    risk, origin="lower",
    extent=[LAM_MIN, LAM_MAX, RHO_MIN, RHO_MAX],
    aspect="auto", interpolation="nearest",
)
cb = plt.colorbar(im, ax=ax)
cb.set_label("Risk score: exp(-dist/eps)·exp(-trans/delta)")

ax.contour(LAM, RHO, f, levels=[0.0], linewidths=1.2)

ax.set_xlabel(r"Irradiation parameter $\lambda$")
ax.set_ylabel(r"Spot size ratio $\rho=d/D$")
ax.set_title("BDI–ZT boundary primitives on (λ, ρ) slice (fitted proxy + primitives)")
fig.tight_layout()
fig.savefig(OUT_PDF, bbox_inches="tight")
plt.show()

print(f"[OK] wrote {OUT_PDF}")
print(f"[OK] wrote {OUT_GRID_CSV}")
```

```python
# === Fig11 boundary digitizer -> fg_boundary_points.csv (ONE CELL) ===
# Usage:
# 1) Set IMG_PATH to your Fig11 image (png recommended). If you only have PDF, first convert to png.
# 2) Click 3 calibration points in this order:
#    (λ,ρ)=(λmin,ρmin) bottom-left, (λmax,ρmin) bottom-right, (λmin,ρmax) top-left
# 3) Then click boundary curve points (20-80 points).
# 4) Press "Export CSV" button -> fg_boundary_points.csv saved.

import numpy as np
import pandas as pd
from PIL import Image
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

# ----------------------------
# EDIT HERE
IMG_PATH = "Fig11_crop_83.png"   # <- set your Fig11 image path
LAM_MIN, LAM_MAX = 0.0, 4.0      # <- set axis range shown in Fig11
RHO_MIN, RHO_MAX = 0.0, 1.0      # <- set axis range shown in Fig11
OUT_CSV = "fg_boundary_points.csv"
# ----------------------------

img = np.array(Image.open(IMG_PATH).convert("RGB"))
H, W = img.shape[0], img.shape[1]

calib_px = []   # 3 points
curve_px = []   # many points

fig = go.FigureWidget()
fig.add_trace(go.Image(z=img))
# overlay markers for calibration (blue) and curve points (red)
fig.add_trace(go.Scatter(x=[], y=[], mode="markers", marker=dict(size=8, color="cyan"), name="calib"))
fig.add_trace(go.Scatter(x=[], y=[], mode="markers", marker=dict(size=5, color="red"), name="boundary"))

fig.update_layout(
    width=950,
    height=int(950 * H / W),
    margin=dict(l=0, r=0, t=40, b=0),
    title="Click 3 calibration points: BL(λmin,ρmin) -> BR(λmax,ρmin) -> TL(λmin,ρmax), then click boundary points",
)
# IMPORTANT: make y increase downward to match image pixel coordinates
fig.update_yaxes(autorange="reversed", visible=False)
fig.update_xaxes(visible=False)

status = widgets.HTML("<b>Status:</b> waiting for clicks…")
btn_clear = widgets.Button(description="Clear points", button_style="warning")
btn_export = widgets.Button(description="Export CSV", button_style="success")
out = widgets.Output()

def px_to_data(x, y):
    # Requires 3 calibration points:
    # p00: (LAM_MIN, RHO_MIN) = bottom-left
    # p10: (LAM_MAX, RHO_MIN) = bottom-right
    # p01: (LAM_MIN, RHO_MAX) = top-left
    (x00, y00), (x10, y10), (x01, y01) = calib_px
    lam = LAM_MIN + (x - x00) / (x10 - x00) * (LAM_MAX - LAM_MIN)
    rho = RHO_MIN + (y00 - y) / (y00 - y01) * (RHO_MAX - RHO_MIN)
    return lam, rho

def refresh_plot():
    fig.data[1].x = [p[0] for p in calib_px]
    fig.data[1].y = [p[1] for p in calib_px]
    fig.data[2].x = [p[0] for p in curve_px]
    fig.data[2].y = [p[1] for p in curve_px]

def onclick(trace, points, selector):
    if not points.xs:
        return
    x = float(points.xs[0])
    y = float(points.ys[0])

    if len(calib_px) < 3:
        calib_px.append((x, y))
        refresh_plot()
        status.value = f"<b>Status:</b> calib {len(calib_px)}/3 stored. px=({x:.1f},{y:.1f})"
        if len(calib_px) == 3:
            status.value = "<b>Status:</b> calibration done. Now click boundary curve points (20-80)."
    else:
        curve_px.append((x, y))
        refresh_plot()
        status.value = f"<b>Status:</b> boundary points = {len(curve_px)} (click more, then Export)."

fig.data[0].on_click(onclick)

def clear_points(_):
    calib_px.clear()
    curve_px.clear()
    refresh_plot()
    status.value = "<b>Status:</b> cleared. Click calibration points again."

def export_csv(_):
    with out:
        out.clear_output()
        if len(calib_px) < 3:
            print("ERROR: need 3 calibration clicks first.")
            return
        if len(curve_px) < 5:
            print("ERROR: boundary points too few. Click at least ~20 points ideally.")
            return

        rows = []
        for (x, y) in curve_px:
            lam, rho = px_to_data(x, y)
            rows.append((lam, rho))
        df = pd.DataFrame(rows, columns=["lam", "rho"]).dropna()

        # sort & thin a bit (optional)
        df = df.sort_values("lam").reset_index(drop=True)
        df.to_csv(OUT_CSV, index=False)

        print(f"[OK] wrote {OUT_CSV}")
        display(df.head(10))
        print("…")
        display(df.tail(10))

btn_clear.on_click(clear_points)
btn_export.on_click(export_csv)

display(widgets.HBox([btn_clear, btn_export]), status, fig, out)
```

```python
import json, numpy as np, pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

MODEL_JSON = Path("/content/boundary_model/boundary_model.json")
assert MODEL_JSON.exists(), f"not found: {MODEL_JSON}"

model = json.loads(MODEL_JSON.read_text(encoding="utf-8"))

print("keys:", list(model.keys()))
print("theta_dim:", model.get("theta_dim"))
print("theta_names:", model.get("theta_names"))
print("f_type:", model.get("f_type"))
print("has scaling:", "scaling" in model)

coeffs = model["coeffs"]  # list of dicts: {"power":[i,j,...], "c": float}
theta_dim = int(model["theta_dim"])

# ----------------------------
# (A) scaling: try common formats; fallback to identity
# We define forward map: raw x -> theta
# ----------------------------
sc = model.get("scaling", None)

def forward_scale(x_vec):
    """
    x_vec: (..., theta_dim) raw coordinates (lam,rho) etc
    returns theta in same shape.
    """
    x = np.asarray(x_vec, float)

    if sc is None:
        return x

    # common patterns
    # 1) {"shift":[...], "scale":[...]} meaning theta=(x-shift)/scale
    if isinstance(sc, dict) and "shift" in sc and "scale" in sc:
        shift = np.asarray(sc["shift"], float)
        scale = np.asarray(sc["scale"], float)
        return (x - shift) / (scale + 1e-12)

    # 2) {"mu":[...], "sigma":[...]} meaning theta=(x-mu)/sigma
    if isinstance(sc, dict) and "mu" in sc and "sigma" in sc:
        mu = np.asarray(sc["mu"], float)
        sigma = np.asarray(sc["sigma"], float)
        return (x - mu) / (sigma + 1e-12)

    # 3) {"min":[...], "max":[...]} meaning theta=(x-min)/(max-min)
    if isinstance(sc, dict) and "min" in sc and "max" in sc:
        mn = np.asarray(sc["min"], float)
        mx = np.asarray(sc["max"], float)
        return (x - mn) / (mx - mn + 1e-12)

    # if unknown format -> assume identity
    return x

# ----------------------------
# (B) build f(theta) from coeffs
# ----------------------------
def f_theta(theta_vec):
    """
    theta_vec: (..., theta_dim)
    f = sum_k c_k * prod_j theta_j^(power_k[j])
    """
    th = np.asarray(theta_vec, float)
    out = np.zeros(th.shape[:-1], float)
    for term in coeffs:
        pw = term["power"]
        c  = float(term["c"])
        t = np.ones(out.shape, float)
        for j in range(theta_dim):
            p = int(pw[j]) if j < len(pw) else 0
            if p != 0:
                t *= th[..., j] ** p
        out += c * t
    return out

def f_eval(lam, rho):
    # raw -> theta -> f
    x = np.stack([lam, rho], axis=-1)  # assume first two dims are lam,rho
    th = forward_scale(x)
    return f_theta(th)

# ----------------------------
# (C) extract contour f=0 -> boundary points CSV
# ----------------------------
LAM_MIN, LAM_MAX = 0.0, 4.0
RHO_MIN, RHO_MAX = 0.0, 1.0
N_LAM, N_RHO = 700, 500

lam = np.linspace(LAM_MIN, LAM_MAX, N_LAM)
rho = np.linspace(RHO_MIN, RHO_MAX, N_RHO)
LAM, RHO = np.meshgrid(lam, rho)

F = f_eval(LAM, RHO)

fig = plt.figure()
cs  = plt.contour(LAM, RHO, F, levels=[0.0])
plt.close(fig)

# ---------- robust contour extraction (matplotlib API-safe) ----------
# sanity: f=0 がそもそも存在するか
fmin, fmax = float(np.nanmin(F)), float(np.nanmax(F))
if not (fmin <= 0.0 <= fmax):
    raise RuntimeError(
        f"No f=0 crossing on this grid: min(F)={fmin:.3e}, max(F)={fmax:.3e}. "
        "Domain/scale may be mismatched with this boundary_model."
    )

segs = None

# 1) most robust: allsegs (list of segments per level)
if hasattr(cs, "allsegs") and len(cs.allsegs) > 0 and len(cs.allsegs[0]) > 0:
    segs = cs.allsegs[0]   # list of (N,2) arrays

# 2) fallback: old API (collections)
elif hasattr(cs, "collections") and len(getattr(cs, "collections", [])) > 0:
    segs = [p.vertices for p in cs.collections[0].get_paths() if p.vertices is not None]

# 3) fallback: some versions expose artists
elif hasattr(cs, "artists") and len(getattr(cs, "artists", [])) > 0:
    segs = [p.vertices for p in cs.artists[0].get_paths() if p.vertices is not None]

if not segs:
    raise RuntimeError("No f=0 contour segments were extracted (API/degenerate contour).")

# longest segment = main boundary
verts = max(segs, key=lambda a: a.shape[0])

out = pd.DataFrame({"lam": verts[:,0], "rho": verts[:,1]})
out.to_csv("fg_boundary_points.csv", index=False)
print("[OK] wrote fg_boundary_points.csv with", len(out), "points")


# quick preview
plt.figure(figsize=(6.2,3.2))
plt.plot(out["lam"], out["rho"], "-", lw=1.6)
plt.xlim(LAM_MIN, LAM_MAX); plt.ylim(RHO_MIN, RHO_MAX)
plt.xlabel("lam"); plt.ylabel("rho")
plt.title("Boundary extracted from boundary_model.json (f=0)")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# make_koibuchi_baseline_figs.py
# Generate clean Fig10.pdf / Fig11.pdf by auto-cropping the white figure region
# from provided screenshot PNG/PDF assets.

import argparse
from pathlib import Path

import numpy as np
from PIL import Image

try:
    import fitz  # PyMuPDF
except Exception:
    fitz = None


def load_image_any(path: Path, pdf_zoom: float = 2.0) -> Image.Image:
    """Load an image from PNG/JPG or render first page of a PDF."""
    suffix = path.suffix.lower()
    if suffix in [".png", ".jpg", ".jpeg", ".webp"]:
        return Image.open(path).convert("RGB")

    if suffix == ".pdf":
        if fitz is None:
            raise RuntimeError("PyMuPDF (fitz) is required to render PDFs. Install pymupdf.")
        doc = fitz.open(str(path))
        page = doc.load_page(0)
        mat = fitz.Matrix(pdf_zoom, pdf_zoom)
        pix = page.get_pixmap(matrix=mat, alpha=False)
        doc.close()
        arr = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.height, pix.width, pix.n)
        return Image.fromarray(arr).convert("RGB")

    raise ValueError(f"Unsupported input type: {path}")


def find_white_bbox(img: Image.Image, thresh: int = 235, frac: float = 0.02) -> tuple[int, int, int, int]:
    """
    Find bounding box of dominant 'white' region.
    Uses row/col occupancy thresholds to be robust to noise.
    """
    gray = np.array(img.convert("L"))
    mask = gray >= thresh  # white-ish pixels

    h, w = mask.shape
    row_sum = mask.sum(axis=1)
    col_sum = mask.sum(axis=0)

    row_min = int(w * frac)
    col_min = int(h * frac)

    rows = np.where(row_sum > row_min)[0]
    cols = np.where(col_sum > col_min)[0]

    if len(rows) == 0 or len(cols) == 0:
        raise RuntimeError(
            "Auto-crop failed (no dominant white region detected). "
            "Try lowering --thresh (e.g. 220) or increase PDF zoom."
        )

    y0, y1 = int(rows.min()), int(rows.max()) + 1
    x0, x1 = int(cols.min()), int(cols.max()) + 1
    return x0, y0, x1, y1


def crop_with_margin(img: Image.Image, bbox: tuple[int, int, int, int], margin: int = 30) -> Image.Image:
    x0, y0, x1, y1 = bbox
    x0 = max(0, x0 - margin)
    y0 = max(0, y0 - margin)
    x1 = min(img.width, x1 + margin)
    y1 = min(img.height, y1 + margin)
    return img.crop((x0, y0, x1, y1))


def save_as_pdf(img: Image.Image, out_pdf: Path, dpi: int = 300) -> None:
    out_pdf.parent.mkdir(parents=True, exist_ok=True)
    img = img.convert("RGB")
    img.save(str(out_pdf), "PDF", resolution=float(dpi))


def process_one(in_path: Path, out_pdf: Path, thresh: int, frac: float, margin: int, dpi: int, pdf_zoom: float,
                debug_png: Path | None = None) -> None:
    img = load_image_any(in_path, pdf_zoom=pdf_zoom)
    bbox = find_white_bbox(img, thresh=thresh, frac=frac)
    cropped = crop_with_margin(img, bbox, margin=margin)
    save_as_pdf(cropped, out_pdf, dpi=dpi)
    if debug_png is not None:
        debug_png.parent.mkdir(parents=True, exist_ok=True)
        cropped.save(str(debug_png))


def main():
    ap = argparse.ArgumentParser()
    ap.add_argument("--fig10_in", type=str, required=True,
                    help="Input for Fig10 (recommended: Fig10_crop_82.png). PDF/PNG both ok.")
    ap.add_argument("--fig11_in", type=str, required=True,
                    help="Input for Fig11 (recommended: Fig11_crop_83.png). PDF/PNG both ok.")
    ap.add_argument("--out_dir", type=str, default=".",
                    help="Output directory for Fig10.pdf and Fig11.pdf")
    ap.add_argument("--thresh", type=int, default=235,
                    help="White threshold (lower if auto-crop fails). Typical: 220-240")
    ap.add_argument("--frac", type=float, default=0.02,
                    help="Row/col occupancy fraction for bbox detection. Typical: 0.01-0.05")
    ap.add_argument("--margin", type=int, default=30,
                    help="Margin (pixels) to expand bbox")
    ap.add_argument("--dpi", type=int, default=300,
                    help="PDF raster DPI")
    ap.add_argument("--pdf_zoom", type=float, default=2.0,
                    help="PDF render zoom if inputs are PDFs")
    ap.add_argument("--debug", action="store_true",
                    help="Also save debug PNGs (cropped) next to PDFs")
    args = ap.parse_args()

    out_dir = Path(args.out_dir)
    fig10_out = out_dir / "Fig10.pdf"
    fig11_out = out_dir / "Fig11.pdf"

    dbg10 = out_dir / "Fig10_cropped.png" if args.debug else None
    dbg11 = out_dir / "Fig11_cropped.png" if args.debug else None

    process_one(Path(args.fig10_in), fig10_out, args.thresh, args.frac, args.margin, args.dpi, args.pdf_zoom, dbg10)
    process_one(Path(args.fig11_in), fig11_out, args.thresh, args.frac, args.margin, args.dpi, args.pdf_zoom, dbg11)

    print(f"[OK] Wrote: {fig10_out}")
    print(f"[OK] Wrote: {fig11_out}")
    if args.debug:
        print(f"[DBG] Wrote: {dbg10}")
        print(f"[DBG] Wrote: {dbg11}")


if __name__ == "__main__":
    main()
```

```python
# ==== Fig10-11 -> insertion-ready PDFs (auto-crop + 1-page PDF) ====
# Outputs:
#   Fig10_insert.pdf, Fig11_insert.pdf (and .png previews)
#
# Works with PNG sources (recommended: crop/preview images) OR PDFs (fallback via render).
#
# --- install if missing (Colab usually OK) ---
import sys, subprocess, importlib, glob, os

def ensure(pkg, import_name=None):
    name = import_name or pkg
    try:
        importlib.import_module(name)
    except Exception:
        subprocess.check_call([sys.executable, "-m", "pip", "-q", "install", pkg])

ensure("opencv-python", "cv2")
ensure("pymupdf", "fitz")

import cv2
import numpy as np
import fitz
import matplotlib.pyplot as plt

# ---------- (A) find candidate inputs ----------
def find_first(patterns, roots=(".", "/content")):
    hits = []
    for r in roots:
        for pat in patterns:
            hits += glob.glob(os.path.join(r, pat), recursive=True)
    hits = [h for h in hits if os.path.isfile(h)]
    return hits[0] if hits else None

# Priority: cropped/preview PNGs > PDFs
FIG10_IN = find_first([
    "**/*Fig10*crop*.png",
    "**/*preview*fig10*.png",
    "**/*fig10*preview*.png",
    "**/*Fig10*.png",
    "**/*fig10*.png",
    "**/*Fig10*.pdf",
    "**/*fig10*.pdf",
])

FIG11_IN = find_first([
    "**/*preview*fig11*.png",          # full view often here
    "**/*Fig11*crop*.png",
    "**/*Fig11*.png",
    "**/*fig11*.png",
    "**/*Fig11*.pdf",
    "**/*fig11*.pdf",
])

print("[INPUT] Fig10:", FIG10_IN)
print("[INPUT] Fig11:", FIG11_IN)
assert FIG10_IN is not None, "Fig10 input not found. Put Fig10.* (png/pdf) under /content."
assert FIG11_IN is not None, "Fig11 input not found. Put Fig11.* (png/pdf) under /content."

# ---------- (B) load image (PNG directly / PDF rendered) ----------
def load_as_bgr(path, zoom=2.0, page=0):
    ext = os.path.splitext(path.lower())[1]
    if ext in [".png", ".jpg", ".jpeg", ".webp"]:
        img = cv2.imread(path, cv2.IMREAD_COLOR)
        if img is None:
            raise RuntimeError(f"cv2.imread failed: {path}")
        return img
    if ext == ".pdf":
        doc = fitz.open(path)
        page = doc[page]
        pix = page.get_pixmap(matrix=fitz.Matrix(zoom, zoom), alpha=False)
        img = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.height, pix.width, 3)
        doc.close()
        # fitz gives RGB; convert to BGR for cv2
        return cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
    raise ValueError(f"Unsupported: {path}")

# ---------- (C) auto-crop: pick the largest "white-ish" connected component ----------
def autocrop_white_panel(bgr, pad=12):
    gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)

    # threshold for bright pixels (tune if needed)
    _, bw = cv2.threshold(gray, 242, 255, cv2.THRESH_BINARY)

    # close gaps so the panel becomes one blob
    k = cv2.getStructuringElement(cv2.MORPH_RECT, (19, 19))
    bw2 = cv2.morphologyEx(bw, cv2.MORPH_CLOSE, k, iterations=2)

    n, labels, stats, _ = cv2.connectedComponentsWithStats(bw2, connectivity=8)

    # stats: [x,y,w,h,area]
    if n <= 1:
        return bgr  # fallback: no components

    # skip background label 0; pick largest area component
    best = 1 + np.argmax(stats[1:, 4])
    x, y, w, h, area = stats[best]

    # pad and clamp
    H, W = bgr.shape[:2]
    x0 = max(0, x - pad); y0 = max(0, y - pad)
    x1 = min(W, x + w + pad); y1 = min(H, y + h + pad)

    crop = bgr[y0:y1, x0:x1].copy()

    # second-pass trim: remove extra margins by tight bounding box on non-white pixels
    g2 = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
    mask = g2 < 250
    ys, xs = np.where(mask)
    if len(xs) > 0:
        x0, x1 = xs.min(), xs.max()
        y0, y1 = ys.min(), ys.max()
        # small pad again
        pad2 = 8
        x0 = max(0, x0 - pad2); y0 = max(0, y0 - pad2)
        x1 = min(crop.shape[1]-1, x1 + pad2); y1 = min(crop.shape[0]-1, y1 + pad2)
        crop = crop[y0:y1+1, x0:x1+1].copy()

    return crop

# ---------- (D) save as PDF (1 page) via matplotlib ----------
def save_pdf_from_bgr(bgr, out_pdf, out_png=None, dpi_out=300):
    rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
    h, w = rgb.shape[:2]
    fig_w = w / dpi_out
    fig_h = h / dpi_out
    plt.figure(figsize=(fig_w, fig_h), dpi=dpi_out)
    plt.imshow(rgb)
    plt.axis("off")
    plt.tight_layout(pad=0)
    plt.savefig(out_pdf, bbox_inches="tight", pad_inches=0)
    if out_png:
        plt.savefig(out_png, bbox_inches="tight", pad_inches=0, dpi=dpi_out)
    plt.close()
    print("[OK] wrote:", out_pdf, (f"+ {out_png}" if out_png else ""))

# ---------- run ----------
img10 = load_as_bgr(FIG10_IN, zoom=2.5, page=0)
img11 = load_as_bgr(FIG11_IN, zoom=2.5, page=0)

crop10 = autocrop_white_panel(img10, pad=14)
crop11 = autocrop_white_panel(img11, pad=14)

save_pdf_from_bgr(crop10, "Fig10_insert.pdf", out_png="Fig10_insert.png", dpi_out=320)
save_pdf_from_bgr(crop11, "Fig11_insert.pdf", out_png="Fig11_insert.png", dpi_out=320)

# quick preview in notebook
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.imshow(cv2.cvtColor(crop10, cv2.COLOR_BGR2RGB)); plt.axis("off"); plt.title("Fig10_insert (cropped)")
plt.subplot(1,2,2); plt.imshow(cv2.cvtColor(crop11, cv2.COLOR_BGR2RGB)); plt.axis("off"); plt.title("Fig11_insert (cropped)")
plt.show()
```

```python
# ===== boundary_model/Fig10-11 auto-format (crop + export PDF/PNG) =====
import os, glob
import numpy as np
import matplotlib.pyplot as plt

# --- deps: PyMuPDF for PDF rendering ---
try:
    import fitz  # PyMuPDF
except Exception:
    !pip -q install pymupdf
    import fitz

try:
    from PIL import Image
except Exception:
    !pip -q install pillow
    from PIL import Image

ROOT = "/content/boundary_model"
FIG10_STEM = "Fig10"
FIG11_STEM = "Fig11"

def find_first(stem, root=ROOT):
    exts = ["pdf","png","jpg","jpeg","webp"]
    # exact stem.ext first
    for ext in exts:
        p = os.path.join(root, f"{stem}.{ext}")
        if os.path.exists(p):
            return p
    # fallback: any file containing stem
    hits = []
    for ext in exts:
        hits += glob.glob(os.path.join(root, f"*{stem}*.{ext}"))
    return sorted(hits)[0] if hits else None

def load_rgb(path, page=0, zoom=2.2):
    path = str(path)
    if path.lower().endswith(".pdf"):
        doc = fitz.open(path)
        p = doc.load_page(page)
        pix = p.get_pixmap(matrix=fitz.Matrix(zoom, zoom), alpha=False)
        img = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.height, pix.width, 3)
        doc.close()
        return img
    else:
        return np.array(Image.open(path).convert("RGB"))

def crop_by_bright_bbox(rgb, thr=210, pad=25):
    """Pick bbox of bright pixels (white-ish)"""
    gray = rgb.mean(axis=2)
    mask = gray > thr
    ys = np.where(mask.any(axis=1))[0]
    xs = np.where(mask.any(axis=0))[0]
    if ys.size == 0 or xs.size == 0:
        return rgb
    y0, y1 = int(ys[0]), int(ys[-1])
    x0, x1 = int(xs[0]), int(xs[-1])
    y0 = max(0, y0 - pad); y1 = min(rgb.shape[0]-1, y1 + pad)
    x0 = max(0, x0 - pad); x1 = min(rgb.shape[1]-1, x1 + pad)
    return rgb[y0:y1+1, x0:x1+1]

def trim_margins_by_rowcol_fraction(rgb, bright_thr=200, frac=0.28, pad=18):
    """
    Further trim top/bottom/left/right where 'bright' fraction is small.
    frac: threshold of bright-pixel fraction required to keep the row/col.
    """
    gray = rgb.mean(axis=2)
    bright = (gray > bright_thr)
    row_frac = bright.mean(axis=1)
    col_frac = bright.mean(axis=0)

    ys = np.where(row_frac > frac)[0]
    xs = np.where(col_frac > frac)[0]
    if ys.size == 0 or xs.size == 0:
        return rgb

    y0, y1 = int(ys[0]), int(ys[-1])
    x0, x1 = int(xs[0]), int(xs[-1])

    y0 = max(0, y0 - pad); y1 = min(rgb.shape[0]-1, y1 + pad)
    x0 = max(0, x0 - pad); x1 = min(rgb.shape[1]-1, x1 + pad)
    return rgb[y0:y1+1, x0:x1+1]

def save_pdf_png(rgb, out_pdf, out_png=None, dpi=320):
    h, w = rgb.shape[:2]
    fig = plt.figure(figsize=(w/dpi, h/dpi), dpi=dpi)
    plt.imshow(rgb)
    plt.axis("off")
    plt.tight_layout(pad=0)
    fig.savefig(out_pdf, bbox_inches="tight", pad_inches=0)
    if out_png:
        fig.savefig(out_png, bbox_inches="tight", pad_inches=0)
    plt.close(fig)

def process_one(stem, page=0):
    inp = find_first(stem)
    assert inp is not None, f"NOT FOUND: {stem}.* under {ROOT}"
    print(f"[INPUT] {stem}: {inp}")

    rgb = load_rgb(inp, page=page, zoom=2.2)

    # --- main crop pipeline (tune here if needed) ---
    out = crop_by_bright_bbox(rgb, thr=210, pad=28)
    out = trim_margins_by_rowcol_fraction(out, bright_thr=200, frac=0.28, pad=18)

    out_pdf = os.path.join(ROOT, f"{stem}_insert.pdf")
    out_png = os.path.join(ROOT, f"{stem}_insert.png")
    save_pdf_png(out, out_pdf, out_png=out_png, dpi=320)

    print(f"[OK] wrote: {out_pdf}")
    print(f"[OK] wrote: {out_png}")
    return out

# --- run ---
out10 = process_one(FIG10_STEM, page=0)
out11 = process_one(FIG11_STEM, page=0)

# --- preview ---
plt.figure(figsize=(10,4)); plt.imshow(out10); plt.axis("off"); plt.title("Fig10_insert (preview)"); plt.show()
plt.figure(figsize=(10,4)); plt.imshow(out11); plt.axis("off"); plt.title("Fig11_insert (preview)"); plt.show()
```

```python
# ============================================================
# Colab ONE-CELL: Koibuchi digitized curves -> inflection vs DTT prediction
# Output: Fig3_extended.pdf  (drop-in replacement for Fig.3)
# ============================================================

import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.mixture import GaussianMixture

from scipy.interpolate import UnivariateSpline

# -----------------------------
# USER CONFIG
# -----------------------------
INPUT_CSV = "FG_BDI_ZT_Map_grid.csv"   # <-- put your digitized CSV in the working dir
OUTPUT_PDF = "Fig3_extended.pdf"

# Choose the slice D/H for the (lambda,rho) map panel (a).
# If None: auto-pick the most frequent D/H in the CSV.
D_OVER_H_SLICE = None

# Protocol direction u for TI: lambda-sweep => u = (1,0) in (lambda,rho).
U_DIR = np.array([1.0, 0.0])

# Risk-score hyperparameters (stable defaults)
EPS = 1e-9
DIST_SCALE = 0.15  # controls how sharply risk rises near the boundary
TI_SCALE   = 0.15  # controls how sharply risk rises near tangency

# Example condition shown in panel (b):
# If None: auto-pick median rho within the selected D/H slice.
EXAMPLE_RHO = None

# -----------------------------
# Robust column mapping
# -----------------------------
COLMAP_CANDIDATES = {
    "lambda": ["lambda", "lam", "Lambda", "irradiation", "x"],
    "rho": ["rho", "d_over_D", "dD", "d_over_D_ratio", "spot_ratio"],
    "D_over_H": ["D_over_H", "D/H", "DbyH", "aspect_ratio", "D_H"],
    "SH_U": ["SH_U", "SH_upper", "S_H_U", "SHU", "SH (upper)", "SH_upper_surface"],
    "SH_L": ["SH_L", "SH_lower", "S_H_L", "SHL", "SH (lower)", "SH_lower_surface"],
    "SG_U": ["SG_U", "SG_upper", "S_G_U", "SGU"],
    "SG_L": ["SG_L", "SG_lower", "S_G_L", "SGL"],
}

def pick_col(df, key):
    for c in COLMAP_CANDIDATES[key]:
        if c in df.columns:
            return c
    raise KeyError(f"Missing a column for '{key}'. Candidates: {COLMAP_CANDIDATES[key]}")

df = pd.read_csv(INPUT_CSV)

c_lam = pick_col(df, "lambda")
c_rho = pick_col(df, "rho")
c_DH  = pick_col(df, "D_over_H")
c_SHU = pick_col(df, "SH_U")
c_SHL = pick_col(df, "SH_L")

has_SG = True
try:
    c_SGU = pick_col(df, "SG_U")
    c_SGL = pick_col(df, "SG_L")
except KeyError:
    has_SG = False

# sanitize
work = df[[c_lam, c_rho, c_DH, c_SHU, c_SHL] + ([c_SGU, c_SGL] if has_SG else [])].copy()
work.columns = ["lambda", "rho", "D_over_H", "SH_U", "SH_L"] + (["SG_U", "SG_L"] if has_SG else [])
work = work.replace([np.inf, -np.inf], np.nan).dropna()

# -----------------------------
# Phase labels (independent of inflection):
# Use bending-asymmetry proxy from upper/lower split.
# -----------------------------
asym_SH = np.abs(work["SH_U"] - work["SH_L"])
if has_SG:
    asym_SG = np.abs(work["SG_U"] - work["SG_L"])
    asym = 0.7 * asym_SH + 0.3 * asym_SG
else:
    asym = asym_SH

# Unsupervised 2-cluster split (low-asym vs high-asym)
gmm = GaussianMixture(n_components=2, random_state=0)
labels = gmm.fit_predict(asym.to_numpy().reshape(-1,1))
# map to {0,1} where 1 = "high-asym" regime
means = [asym[labels==k].mean() for k in [0,1]]
high_cluster = int(np.argmax(means))
y = (labels == high_cluster).astype(int)

# -----------------------------
# Fit boundary proxy f(lambda, rho) ≈ 0 from labels
# (slice-specific fit: use chosen D/H slice only for a clean (lambda,rho) boundary)
# -----------------------------
if D_OVER_H_SLICE is None:
    # pick the most frequent D/H
    counts = work["D_over_H"].value_counts()
    D_OVER_H_SLICE = float(counts.index[0])

slice_mask = np.isclose(work["D_over_H"].to_numpy(), D_OVER_H_SLICE, atol=1e-6)
W = work.loc[slice_mask].copy()
Y = y[slice_mask]

if len(W) < 50:
    raise RuntimeError(f"Too few points in D/H slice = {D_OVER_H_SLICE}. "
                       f"Either set D_OVER_H_SLICE=None or choose a slice with more digitized points.")

X = W[["lambda","rho"]].to_numpy()

clf = Pipeline([
    ("poly", PolynomialFeatures(degree=3, include_bias=False)),
    ("lr", LogisticRegression(max_iter=5000, solver="lbfgs"))
])
clf.fit(X, Y)

def f_proxy(x2):
    # signed score (decision function): f=0 is the boundary
    return clf.decision_function(np.asarray(x2).reshape(-1,2)).reshape(-1)

# numerical gradient for dist and TI
def grad_f(x, h=1e-3):
    x = np.asarray(x, dtype=float)
    fx = f_proxy(x)[0]
    # central differences
    e1 = np.array([h,0.0]); e2 = np.array([0.0,h])
    dfdlam = (f_proxy(x+e1)[0] - f_proxy(x-e1)[0])/(2*h)
    dfdrho = (f_proxy(x+e2)[0] - f_proxy(x-e2)[0])/(2*h)
    return np.array([dfdlam, dfdrho], dtype=float), fx

def dist_and_ti(x):
    g, fx = grad_f(x)
    gn = np.linalg.norm(g) + EPS
    dist = abs(fx)/gn
    ti = abs(np.dot(g, U_DIR))  # directional transversality magnitude
    return dist, ti, fx, g

def risk_score(dist, ti):
    # composite "near-boundary + low-transversality" risk
    # smooth + bounded; higher = riskier
    return math.exp(-dist/(DIST_SCALE+EPS)) * math.exp(-ti/(TI_SCALE+EPS))

# -----------------------------
# Inflection extraction from S_H(lambda):
# Define S_H as mean of upper/lower (can change if you prefer upper only)
# -----------------------------
def inflection_lambda(lam, SH, s=0.5):
    lam = np.asarray(lam, float)
    SH  = np.asarray(SH, float)
    idx = np.argsort(lam)
    lam, SH = lam[idx], SH[idx]
    # smoothing spline
    spl = UnivariateSpline(lam, SH, s=s*len(lam))
    grid = np.linspace(lam.min(), lam.max(), 600)
    d2 = spl.derivative(n=2)(grid)
    # find sign-change of second derivative
    sgn = np.sign(d2)
    zc = np.where(np.diff(sgn) != 0)[0]
    if len(zc) == 0:
        # fallback: closest to zero
        k = int(np.argmin(np.abs(d2)))
        return float(grid[k])
    # choose the most prominent sign change (largest |d2| around it)
    best = None
    best_score = -1
    for k in zc:
        win = slice(max(0,k-5), min(len(d2),k+6))
        score = float(np.max(np.abs(d2[win])))
        if score > best_score:
            best_score = score
            best = k
    return float(grid[best])

# -----------------------------
# Build per-condition predictions (rho within slice)
# -----------------------------
W["SH_mean"] = 0.5*(W["SH_U"] + W["SH_L"])

# choose example rho
if EXAMPLE_RHO is None:
    rhos = np.sort(W["rho"].unique())
    EXAMPLE_RHO = float(rhos[len(rhos)//2])

# group by rho in the slice (since D/H fixed here)
pred_rows = []
example_bundle = None

for rho_val, G in W.groupby("rho"):
    G = G.sort_values("lambda")
    lam = G["lambda"].to_numpy()
    SHm = G["SH_mean"].to_numpy()

    # inflection (independent diagnostic from curvature)
    lam_infl = inflection_lambda(lam, SHm, s=0.7)

    # DTT score along lambda sweep at this rho
    scores = []
    dists  = []
    tis    = []
    for l in lam:
        dist, ti, fx, grad = dist_and_ti([l, rho_val])
        dists.append(dist); tis.append(ti)
        scores.append(risk_score(dist, ti))
    scores = np.asarray(scores)
    lam_pred = float(lam[int(np.argmax(scores))])

    pred_rows.append([float(rho_val), lam_pred, lam_infl])

    # keep example panel data
    if abs(float(rho_val) - EXAMPLE_RHO) < 1e-12:
        example_bundle = dict(
            rho=float(rho_val), lam=lam, SHm=SHm, score=scores,
            lam_pred=lam_pred, lam_infl=lam_infl
        )

pred = pd.DataFrame(pred_rows, columns=["rho","lambda_pred","lambda_infl"])

if example_bundle is None:
    # pick closest
    i = int(np.argmin(np.abs(pred["rho"].to_numpy()-EXAMPLE_RHO)))
    rho_pick = float(pred.iloc[i]["rho"])
    G = W[W["rho"]==rho_pick].sort_values("lambda")
    lam = G["lambda"].to_numpy()
    SHm = G["SH_mean"].to_numpy()
    scores = np.array([risk_score(*dist_and_ti([l, rho_pick])[:2]) for l in lam])
    example_bundle = dict(
        rho=rho_pick, lam=lam, SHm=SHm, score=scores,
        lam_pred=float(pred.iloc[i]["lambda_pred"]),
        lam_infl=float(pred.iloc[i]["lambda_infl"])
    )

# -----------------------------
# Panel (a): risk map on (lambda,rho) for the selected D/H slice
# -----------------------------
lam_min, lam_max = W["lambda"].min(), W["lambda"].max()
rho_min, rho_max = W["rho"].min(), W["rho"].max()

Lam = np.linspace(lam_min, lam_max, 260)
Rho = np.linspace(rho_min, rho_max, 220)
LL, RR = np.meshgrid(Lam, Rho)

risk_grid = np.zeros_like(LL)
f_grid = np.zeros_like(LL)

for i in range(LL.shape[0]):
    for j in range(LL.shape[1]):
        x = [LL[i,j], RR[i,j]]
        dist, ti, fx, grad = dist_and_ti(x)
        risk_grid[i,j] = risk_score(dist, ti)
        f_grid[i,j] = fx

# -----------------------------
# Make the extended Fig.3 (single PDF)
# Layout: 2 rows x 3 cols:
#   (a) map spans left two rows
#   (b) top-middle: SH curve; bottom-middle: DTT score
#   (c) scatter spans right two rows
# -----------------------------
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(14.5, 6.2))
gs = GridSpec(2, 3, width_ratios=[1.45, 1.05, 1.05], height_ratios=[1.0, 1.0], wspace=0.32, hspace=0.28)

axA = fig.add_subplot(gs[:,0])
axB1 = fig.add_subplot(gs[0,1])
axB2 = fig.add_subplot(gs[1,1], sharex=axB1)
axC = fig.add_subplot(gs[:,2])

# (a) risk map + boundary contour
im = axA.pcolormesh(LL, RR, risk_grid, shading="auto")
cs = axA.contour(LL, RR, f_grid, levels=[0.0], linewidths=2.0)
axA.scatter(W["lambda"], W["rho"], s=8, alpha=0.35)

axA.set_xlabel(r"$\lambda$")
axA.set_ylabel(r"$\rho=d/D$")
axA.set_title(rf"(a) DTT risk map on $(\lambda,\rho)$ slice  ($D/H={D_OVER_H_SLICE:g}$)")
cbar = fig.colorbar(im, ax=axA, fraction=0.046, pad=0.02)
cbar.set_label("risk score")

# (b) representative curves + vertical lines
lam = example_bundle["lam"]
SHm = example_bundle["SHm"]
score = example_bundle["score"]
lam_pred = example_bundle["lam_pred"]
lam_infl = example_bundle["lam_infl"]

axB1.plot(lam, SHm)
axB1.axvline(lam_pred, linestyle="--", linewidth=2.0)
axB1.axvline(lam_infl, linestyle=":", linewidth=2.5)
axB1.set_ylabel(r"$S_H$ (mean of U/L)")
axB1.set_title(rf"(b) Example at $\rho={example_bundle['rho']:g}$, $D/H={D_OVER_H_SLICE:g}$")
axB1.legend([r"$S_H(\lambda)$", r"$\lambda_{\mathrm{{pred}}}$ (DTT)", r"$\lambda_{\mathrm{{infl}}}$ (inflection)"],
            loc="best", fontsize=9)

axB2.plot(lam, score)
axB2.axvline(lam_pred, linestyle="--", linewidth=2.0)
axB2.axvline(lam_infl, linestyle=":", linewidth=2.5)
axB2.set_xlabel(r"$\lambda$")
axB2.set_ylabel("DTT risk")

# (c) scatter pred vs infl
axC.scatter(pred["lambda_infl"], pred["lambda_pred"], s=18, alpha=0.75)
m1 = float(min(pred["lambda_infl"].min(), pred["lambda_pred"].min()))
m2 = float(max(pred["lambda_infl"].max(), pred["lambda_pred"].max()))
axC.plot([m1,m2],[m1,m2], linewidth=2.0)
axC.set_xlabel(r"$\lambda_{\mathrm{infl}}$ (from $\partial_\lambda^2 S_H$)")
axC.set_ylabel(r"$\lambda_{\mathrm{pred}}$ (from DTT)")
axC.set_title("(c) Agreement across conditions")
axC.set_xlim(m1, m2); axC.set_ylim(m1, m2)

fig.suptitle("Extended Fig. 3: DTT predicts Koibuchi’s post-hoc inflection points", y=1.02, fontsize=13)
fig.tight_layout()

fig.savefig(OUTPUT_PDF, bbox_inches="tight")
plt.close(fig)

print(f"[OK] wrote: {OUTPUT_PDF}")
print(f"[INFO] Use this as a drop-in replacement for Fig.3 in LaTeX.")
```

```python
# ============================================================
# ONE-CELL (REPLACEMENT):
# Use your existing grid+boundary outputs and (optional) Koibuchi digitized boundary CSVs
# Output: Fig3_extended.pdf  (drop-in for Fig.3)
# ============================================================

import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.interpolate import UnivariateSpline, griddata

# -----------------------------
# USER CONFIG
# -----------------------------
GRID_CSV = "FG_BDI_ZT_Map_grid.csv"     # exists in your folder
OUR_BD_CSV = "fg_boundary_points.csv"   # exists in your folder

# Put Koibuchi digitized boundary CSVs here (from WebPlotDigitizer).
# If you haven't created them yet, leave as [] and you'll get panel (a) only.
KOIBUCHI_CSVS = [
    "koibuchi_Fig10_boundary.csv",
    "koibuchi_Fig11_boundary.csv",
]

OUTPUT_PDF = "Fig3_extended.pdf"

# Risk construction:
# If GRID_CSV already has "risk" column, we use it.
# Else, if it has "dist" and "TI" columns, we compute risk = exp(-dist/a)*exp(-TI/b)
DIST_SCALE = 0.15
TI_SCALE   = 0.15
EPS = 1e-12

# -----------------------------
# Helper: pick columns robustly
# -----------------------------
def pick(df, cands, required=True):
    for c in cands:
        if c in df.columns:
            return c
    if required:
        raise KeyError(f"Missing column. Candidates={cands}. Available={list(df.columns)}")
    return None

LAMBDA_CANDS = ["lambda","lam","Lambda","x","X"]
RHO_CANDS    = ["rho","d_over_D","dD","y","Y"]
RISK_CANDS   = ["risk","risk_score","score","DTT_risk"]
DIST_CANDS   = ["dist","distance","d","dist_to_boundary"]
TI_CANDS     = ["TI","ti","transversality","TI_u","directional_TI"]

# -----------------------------
# Load grid
# -----------------------------
G = pd.read_csv(GRID_CSV)

c_lam = pick(G, LAMBDA_CANDS)
c_rho = pick(G, RHO_CANDS)
c_risk = pick(G, RISK_CANDS, required=False)
c_dist = pick(G, DIST_CANDS, required=False)
c_ti   = pick(G, TI_CANDS, required=False)

G = G.replace([np.inf,-np.inf], np.nan).dropna(subset=[c_lam,c_rho])

if c_risk is None:
    if c_dist is None or c_ti is None:
        raise RuntimeError(
            "GRID_CSV must contain either a risk column OR both dist and TI columns.\n"
            f"Columns found: {list(G.columns)}"
        )
    dist = G[c_dist].to_numpy(float)
    ti   = G[c_ti].to_numpy(float)
    risk = np.exp(-dist/(DIST_SCALE+EPS)) * np.exp(-ti/(TI_SCALE+EPS))
    G["_risk_"] = risk
    c_risk = "_risk_"

lam_all = G[c_lam].to_numpy(float)
rho_all = G[c_rho].to_numpy(float)
risk_all = G[c_risk].to_numpy(float)

# For plotting panel (a), we create a grid by pivoting (if rectangular),
# otherwise we fall back to scattered pcolormesh-like drawing.
lam_u = np.unique(lam_all)
rho_u = np.unique(rho_all)

# try rectangular grid
rectangular = (len(lam_u)*len(rho_u) == len(G))
risk_grid = None
LL = RR = None
if rectangular:
    pivot = G.pivot_table(index=c_rho, columns=c_lam, values=c_risk, aggfunc="mean")
    pivot = pivot.sort_index().sort_index(axis=1)
    RR, LL = np.meshgrid(pivot.index.to_numpy(float), pivot.columns.to_numpy(float), indexing="ij")
    risk_grid = pivot.to_numpy(float)

# risk interpolator for arbitrary points (Koibuchi path)
# always available via griddata on scattered points
pts = np.column_stack([lam_all, rho_all])

def risk_at(lam, rho):
    q = np.column_stack([np.asarray(lam,float), np.asarray(rho,float)])
    return griddata(pts, risk_all, q, method="linear")

# -----------------------------
# Load our boundary points (overlay)
# -----------------------------
B = pd.read_csv(OUR_BD_CSV)
b_lam = pick(B, LAMBDA_CANDS)
b_rho = pick(B, RHO_CANDS)
B = B.replace([np.inf,-np.inf], np.nan).dropna(subset=[b_lam,b_rho])

# -----------------------------
# Koibuchi inflection (boundary-curve inflection on rho(lambda))
# -----------------------------
def boundary_inflection_lambda(df_xy):
    # expects columns x=lambda, y=rho (any names accepted via pick)
    xcol = pick(df_xy, LAMBDA_CANDS)
    ycol = pick(df_xy, RHO_CANDS)
    x = df_xy[xcol].to_numpy(float)
    y = df_xy[ycol].to_numpy(float)
    idx = np.argsort(x)
    x, y = x[idx], y[idx]
    # smooth spline rho(lambda)
    # s controls smoothing; scale with n
    spl = UnivariateSpline(x, y, s=0.5*len(x))
    grid = np.linspace(x.min(), x.max(), 800)
    d2 = spl.derivative(n=2)(grid)
    sgn = np.sign(d2)
    zc = np.where(np.diff(sgn)!=0)[0]
    if len(zc)==0:
        k = int(np.argmin(np.abs(d2)))
        return float(grid[k]), grid, spl(grid), d2
    # pick most prominent sign change
    best, best_score = zc[0], -1.0
    for k in zc:
        win = slice(max(0,k-6), min(len(d2),k+7))
        score = float(np.max(np.abs(d2[win])))
        if score > best_score:
            best_score = score
            best = k
    return float(grid[best]), grid, spl(grid), d2

koibuchi_results = []
for fn in KOIBUCHI_CSVS:
    if os.path.exists(fn):
        K = pd.read_csv(fn)
        K = K.replace([np.inf,-np.inf], np.nan).dropna()
        lam_infl, gridx, griddy, d2 = boundary_inflection_lambda(K)

        # DTT prediction along the SAME Koibuchi curve: choose argmax risk along path
        k_lam = pick(K, LAMBDA_CANDS)
        k_rho = pick(K, RHO_CANDS)
        lam_path = K[k_lam].to_numpy(float)
        rho_path = K[k_rho].to_numpy(float)
        r_path = risk_at(lam_path, rho_path)
        # drop NaN from interpolation (outside convex hull)
        ok = np.isfinite(r_path)
        if ok.sum() >= 5:
            lam_path2 = lam_path[ok]; rho_path2 = rho_path[ok]; r_path2 = r_path[ok]
            lam_pred = float(lam_path2[int(np.nanargmax(r_path2))])
        else:
            lam_pred = np.nan

        koibuchi_results.append(dict(
            file=fn, K=K, lam_infl=lam_infl, lam_pred=lam_pred,
            lam_path=lam_path, rho_path=rho_path, risk_path=r_path
        ))

# -----------------------------
# Build the figure (a,b,c). If no Koibuchi CSVs, output panel (a) only.
# -----------------------------
from matplotlib.gridspec import GridSpec

hasK = (len(koibuchi_results) > 0)

fig = plt.figure(figsize=(14.5, 6.2))
gs = GridSpec(2, 3, width_ratios=[1.45, 1.05, 1.05], height_ratios=[1.0, 1.0], wspace=0.32, hspace=0.28)

axA = fig.add_subplot(gs[:,0])
axB1 = fig.add_subplot(gs[0,1]) if hasK else None
axB2 = fig.add_subplot(gs[1,1], sharex=axB1) if hasK else None
axC = fig.add_subplot(gs[:,2]) if hasK else None

# (a) risk map + our boundary + Koibuchi boundary overlay
if risk_grid is not None:
    # RR is rho-grid (rows), LL is lambda-grid (cols) due to indexing="ij"
    im = axA.pcolormesh(LL, RR, risk_grid, shading="auto")
else:
    im = axA.scatter(lam_all, rho_all, c=risk_all, s=8, alpha=0.9)

axA.scatter(B[b_lam], B[b_rho], s=12, alpha=0.75, label="Our boundary points")

for kr in koibuchi_results:
    K = kr["K"]
    k_lam = pick(K, LAMBDA_CANDS)
    k_rho = pick(K, RHO_CANDS)
    axA.plot(K[k_lam], K[k_rho], linewidth=2.0, label=f"Koibuchi digitized ({kr['file']})")

axA.set_xlabel(r"$\lambda$")
axA.set_ylabel(r"$\rho=d/D$")
axA.set_title("(a) DTT risk map + boundaries (our vs Koibuchi)")
cb = fig.colorbar(im, ax=axA, fraction=0.046, pad=0.02)
cb.set_label("risk score")
axA.legend(loc="best", fontsize=8)

if hasK:
    # (b) show one representative (first) file: rho(lambda) and risk along path
    kr0 = koibuchi_results[0]
    lam_path = kr0["lam_path"]
    rho_path = kr0["rho_path"]
    risk_path = kr0["risk_path"]

    # rho(lambda) curve
    axB1.plot(lam_path, rho_path)
    axB1.axvline(kr0["lam_infl"], linestyle=":", linewidth=2.5)
    if np.isfinite(kr0["lam_pred"]):
        axB1.axvline(kr0["lam_pred"], linestyle="--", linewidth=2.0)
    axB1.set_ylabel(r"$\rho(\lambda)$")
    axB1.set_title(f"(b) Example: {kr0['file']}")
    axB1.legend([r"Koibuchi boundary $\rho(\lambda)$",
                 r"$\lambda_{\mathrm{infl}}$ (boundary inflection)",
                 r"$\lambda_{\mathrm{pred}}$ (max DTT risk)"],
                loc="best", fontsize=8)

    # risk along path
    axB2.plot(lam_path, risk_path)
    axB2.axvline(kr0["lam_infl"], linestyle=":", linewidth=2.5)
    if np.isfinite(kr0["lam_pred"]):
        axB2.axvline(kr0["lam_pred"], linestyle="--", linewidth=2.0)
    axB2.set_xlabel(r"$\lambda$")
    axB2.set_ylabel("risk along Koibuchi boundary")

    # (c) scatter across all Koibuchi digitized boundaries
    infl = np.array([k["lam_infl"] for k in koibuchi_results], float)
    pred = np.array([k["lam_pred"] for k in koibuchi_results], float)
    ok = np.isfinite(infl) & np.isfinite(pred)
    axC.scatter(infl[ok], pred[ok], s=40, alpha=0.85)
    if ok.sum() > 0:
        m1 = float(min(infl[ok].min(), pred[ok].min()))
        m2 = float(max(infl[ok].max(), pred[ok].max()))
        axC.plot([m1,m2],[m1,m2], linewidth=2.0)
        axC.set_xlim(m1,m2); axC.set_ylim(m1,m2)
    axC.set_xlabel(r"$\lambda_{\mathrm{infl}}$ (Koibuchi boundary inflection)")
    axC.set_ylabel(r"$\lambda_{\mathrm{pred}}$ (max DTT risk)")
    axC.set_title("(c) Agreement across digitized baselines")

fig.suptitle("Extended Fig. 3: DTT predicts Koibuchi’s post-hoc boundary inflection points", y=1.02, fontsize=13)
fig.tight_layout()
fig.savefig(OUTPUT_PDF, bbox_inches="tight")
plt.close(fig)

print(f"[OK] wrote: {OUTPUT_PDF}")
if not hasK:
    print("[NOTE] Koibuchi CSVs not found. I produced panel (a)-only style figure.")
    print("       Create Koibuchi digitized boundary CSVs (lambda,rho) and set KOIBUCHI_CSVS, then rerun.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

GRID_CSV = "FG_BDI_ZT_Map_grid.csv"
OUT_CSV  = "fg_boundary_points.csv"
RESAMPLE_N = 400   # 各枝を何点に揃えるか（不要なら None に）

g = pd.read_csv(GRID_CSV)

# f-grid を (rho x lambda) で作る
pivot_f = (g.pivot_table(index="rho", columns="lambda", values="f", aggfunc="mean")
             .sort_index()
             .sort_index(axis=1))

rho = pivot_f.index.to_numpy(float)      # y
lam = pivot_f.columns.to_numpy(float)    # x
LL, RR = np.meshgrid(lam, rho)           # shapes: (len(rho), len(lam))
f_grid = pivot_f.to_numpy(float)

# contour は描画せずに「等高線座標だけ」抜く
fig, ax = plt.subplots(figsize=(6,4))
cs = ax.contour(LL, RR, f_grid, levels=[0.0])
plt.close(fig)

segs = cs.allsegs[0]   # level=0 の全コンポーネント
rows = []

for k, seg in enumerate(segs):
    seg = seg[np.isfinite(seg).all(axis=1)]
    if len(seg) < 2:
        continue

    # 弧長パラメータ
    d = np.sqrt(((np.diff(seg, axis=0))**2).sum(axis=1))
    s = np.concatenate([[0.0], np.cumsum(d)])
    L = float(s[-1])

    if (RESAMPLE_N is None) or (len(seg) == RESAMPLE_N) or (L == 0.0):
        lam_k = seg[:,0]
        rho_k = seg[:,1]
        s_k   = s
    else:
        t = np.linspace(0.0, L, RESAMPLE_N)
        lam_k = np.interp(t, s, seg[:,0])
        rho_k = np.interp(t, s, seg[:,1])
        s_k   = t

    for i in range(len(lam_k)):
        rows.append((k, i, s_k[i], lam_k[i], rho_k[i]))

B = pd.DataFrame(rows, columns=["component","idx","s","lam","rho"])
B.to_csv(OUT_CSV, index=False)

print("[OK] wrote", OUT_CSV)
print("components:", B["component"].nunique(), " | n=", len(B))
print("lam range:", B["lam"].min(), B["lam"].max())
print("rho range:", B["rho"].min(), B["rho"].max())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------- inputs -----------------
GRID_CSV = "FG_BDI_ZT_Map_grid.csv"
BND_CSV  = "fg_boundary_points.csv"
OUT_PDF  = "Fig3_fixed_final.pdf"

# ----------------- load grids -----------------
g = pd.read_csv(GRID_CSV)

pivot_risk = (g.pivot_table(index="rho", columns="lambda", values="risk", aggfunc="mean")
                .sort_index().sort_index(axis=1))
rho = pivot_risk.index.to_numpy(float)
lam = pivot_risk.columns.to_numpy(float)
LL, RR = np.meshgrid(lam, rho)
risk_grid = pivot_risk.to_numpy(float)

# (optional) f-grid: draw true boundary as contour (robust even if CSV missing)
pivot_f = (g.pivot_table(index="rho", columns="lambda", values="f", aggfunc="mean")
             .sort_index().sort_index(axis=1))
f_grid = pivot_f.to_numpy(float)

# ----------------- load boundary points (component-separated) -----------------
B = pd.read_csv(BND_CSV)

# ensure numeric
for c in ["component", "idx"]:
    if c in B.columns:
        B[c] = pd.to_numeric(B[c], errors="coerce").astype("Int64")
for c in ["lam", "rho"]:
    B[c] = pd.to_numeric(B[c], errors="coerce")

B = B.dropna(subset=["lam","rho"])
if "component" not in B.columns:
    # fallback: treat as single component
    B["component"] = 0

# ----------------- helper: detect an approximate crossing between two components -----------------
def nearest_cross_point(Axy, Bxy, max_pairs=2000):
    # downsample to keep O(n^2) manageable
    A = Axy
    C = Bxy
    if len(A) > max_pairs:
        idx = np.linspace(0, len(A)-1, max_pairs).astype(int)
        A = A[idx]
    if len(C) > max_pairs:
        idx = np.linspace(0, len(C)-1, max_pairs).astype(int)
        C = C[idx]
    # pairwise distances
    # (len(A), len(C), 2) is too big; do blockwise
    best = (np.inf, None, None)
    block = 512
    for i0 in range(0, len(A), block):
        Ablk = A[i0:i0+block]
        # compute squared distances to all C
        # Ablk[:,None,:] - C[None,:,:]
        d2 = ((Ablk[:,None,0]-C[None,:,0])**2 + (Ablk[:,None,1]-C[None,:,1])**2)
        j = np.argmin(d2)
        ii, jj = np.unravel_index(j, d2.shape)
        if d2[ii, jj] < best[0]:
            best = (d2[ii, jj], Ablk[ii], C[jj])
    if best[1] is None:
        return None
    # midpoint of nearest pair
    return 0.5*(best[1] + best[2])

# ----------------- plot -----------------
fig = plt.figure(figsize=(7.6, 5.2))
ax = fig.add_subplot(111)

im = ax.pcolormesh(LL, RR, risk_grid, shading="auto")

# draw boundary contour from f-grid (thin) for robustness
ax.contour(LL, RR, f_grid, levels=[0.0], linewidths=1.2)

# draw component-separated polylines with legend
comps = sorted([int(x) for x in B["component"].dropna().unique()])
line_handles = []
for k in comps:
    dfc = B[B["component"] == k].sort_values(by="idx" if "idx" in B.columns else "lam")
    h, = ax.plot(dfc["lam"].to_numpy(float),
                 dfc["rho"].to_numpy(float),
                 linewidth=2.8,
                 label=fr"boundary component {k}")
    line_handles.append(h)

# crossing marker if exactly 2 components
if len(comps) >= 2:
    A = B[B["component"] == comps[0]].sort_values(by="idx" if "idx" in B.columns else "lam")
    C = B[B["component"] == comps[1]].sort_values(by="idx" if "idx" in B.columns else "lam")
    xpt = nearest_cross_point(A[["lam","rho"]].to_numpy(float),
                              C[["lam","rho"]].to_numpy(float))
    if xpt is not None:
        ax.scatter([xpt[0]], [xpt[1]], s=60, marker="x")
        ax.annotate("crossing",
                    xy=(xpt[0], xpt[1]),
                    xytext=(8, 8),
                    textcoords="offset points")

# labels / cosmetics
ax.set_xlabel(r"$\lambda$")
ax.set_ylabel(r"$\rho$")
ax.set_title(r"Boundary-risk map with component-separated boundary overlay")

cb = fig.colorbar(im, ax=ax, fraction=0.046, pad=0.02)
cb.set_label("risk")

# legend (only boundary components; keep it compact)
ax.legend(loc="best", frameon=True)

fig.tight_layout()
fig.savefig(OUT_PDF, bbox_inches="tight")
plt.close(fig)

import os, time
print("[OK] wrote", OUT_PDF, "| mtime:", time.ctime(os.path.getmtime(OUT_PDF)))
!ls -lh "$OUT_PDF"
```

```python
import os, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =======================
# Inputs (edit if needed)
# =======================
GRID_CSV = "FG_BDI_ZT_Map_grid.csv"     # must contain columns: lambda, rho, f, risk (risk optional)
BND_CSV  = "fg_boundary_points.csv"     # component-separated boundary points (optional)
OUT_PDF  = "Fig3_abc_final.pdf"

# --- Koibuchi digitized curve (optional but needed to show λ_infl in Fig.3b/c) ---
# Expected minimal columns: lambda and some curvature column (e.g., SH_U or SH)
KOIBUCHI_CSV = "koibuchi_SH_digitized.csv"   # <-- if you have it, put it here
KOIBUCHI_YCOL_CANDIDATES = ["SH_U", "SH", "S_H", "mean_curvature", "curvature"]
SMOOTH_WIN = 9  # odd recommended; moving-average window for the inflection estimate

# --- choose the sweep slice rho = const (should match the Koibuchi configuration you're validating) ---
RHO_SLICE_TARGET = 0.50   # change to match the external setting (ρ=d/D)

# =======================
# Load grid
# =======================
g = pd.read_csv(GRID_CSV)

# pivot to grids
pivot_f = (g.pivot_table(index="rho", columns="lambda", values="f", aggfunc="mean")
             .sort_index().sort_index(axis=1))
rho = pivot_f.index.to_numpy(float)
lam = pivot_f.columns.to_numpy(float)
LL, RR = np.meshgrid(lam, rho)
f_grid = pivot_f.to_numpy(float)

has_risk = ("risk" in g.columns)
if has_risk:
    pivot_risk = (g.pivot_table(index="rho", columns="lambda", values="risk", aggfunc="mean")
                    .sort_index().sort_index(axis=1))
    risk_grid = pivot_risk.to_numpy(float)
else:
    risk_grid = None

# choose rho slice index
j = int(np.argmin(np.abs(rho - float(RHO_SLICE_TARGET))))
rho0 = float(rho[j])

# =======================
# Compute primitives along the rho-slice: dist ≈ |f|/||∇f|| and TI = |<∇f, (1,0)>| = |∂_λ f|
# =======================
# finite differences (grid spacing possibly non-uniform -> use coordinate arrays)
df_drho, df_dlam = np.gradient(f_grid, rho, lam, edge_order=1)  # shapes match f_grid

f_slice   = f_grid[j, :].copy()
dlam_slice = df_dlam[j, :].copy()
drho_slice = df_drho[j, :].copy()

grad_norm = np.sqrt(dlam_slice**2 + drho_slice**2) + 1e-12
dist_slice = np.abs(f_slice) / grad_norm                      # Eq.(13)-style proxy
TI_slice   = np.abs(dlam_slice)                               # Eq.(12) with sweep direction (1,0)

# risk score along slice
if has_risk:
    risk_slice = risk_grid[j, :].copy()
    score_name = "risk score (grid)"
else:
    # fallback: simple "near boundary + near tangency" score (not identical to your paper risk)
    risk_slice = (1.0/(dist_slice + 1e-6)) * (1.0/(TI_slice + 1e-6))
    score_name = "risk score (fallback)"

# predicted lambda: peak of score
k_pred = int(np.nanargmax(risk_slice))
lam_pred = float(lam[k_pred])

# =======================
# Optional: compute Koibuchi inflection λ_infl from digitized curvature curve
# definition used in your PDF text: sign change of ∂^2_λ S_H after smoothing
# =======================
lam_infl = None
infl_info = ""
if os.path.exists(KOIBUCHI_CSV):
    K = pd.read_csv(KOIBUCHI_CSV)
    # pick y-column
    ycol = None
    for c in KOIBUCHI_YCOL_CANDIDATES:
        if c in K.columns:
            ycol = c
            break
    if ycol is None:
        infl_info = f"[WARN] {KOIBUCHI_CSV} found but no curvature column in {KOIBUCHI_YCOL_CANDIDATES}."
    else:
        x = pd.to_numeric(K["lambda"], errors="coerce").to_numpy(float)
        y = pd.to_numeric(K[ycol], errors="coerce").to_numpy(float)
        ok = np.isfinite(x) & np.isfinite(y)
        x, y = x[ok], y[ok]
        # sort by lambda
        srt = np.argsort(x)
        x, y = x[srt], y[srt]

        # simple moving-average smoothing
        win = int(SMOOTH_WIN)
        if win < 3: win = 3
        if win % 2 == 0: win += 1
        if len(y) >= win:
            kernel = np.ones(win)/win
            y_s = np.convolve(y, kernel, mode="same")
        else:
            y_s = y.copy()

        # second derivative
        d1 = np.gradient(y_s, x, edge_order=1)
        d2 = np.gradient(d1,  x, edge_order=1)

        # find sign-change indices of d2
        sgn = np.sign(d2)
        # treat zeros: replace 0 with previous nonzero sign (simple)
        for t in range(1, len(sgn)):
            if sgn[t] == 0:
                sgn[t] = sgn[t-1]
        cross = np.where(np.diff(sgn) != 0)[0]

        if len(cross) == 0:
            infl_info = "[WARN] no sign change found in second derivative (check digitization / smoothing)."
        else:
            # choose the most pronounced crossing (max |d2| near the crossing)
            idx = cross[np.argmax(np.abs(d2[cross]))]
            lam_infl = float(x[idx])
            infl_info = f"[OK] λ_infl ≈ {lam_infl:.4f} from column '{ycol}' (win={win})."
else:
    infl_info = f"[NOTE] Koibuchi digitized CSV not found: {KOIBUCHI_CSV} (λ_infl line will be omitted)."

# =======================
# Panel (a): map + boundary overlay
# =======================
fig = plt.figure(figsize=(11.5, 3.9))
axA = fig.add_subplot(1,3,1)

if has_risk:
    im = axA.pcolormesh(LL, RR, risk_grid, shading="auto")
else:
    # if risk not present, visualize |f| as a proxy
    im = axA.pcolormesh(LL, RR, np.abs(f_grid), shading="auto")

# true boundary f=0
axA.contour(LL, RR, f_grid, levels=[0.0], linewidths=1.2)

# overlay component-separated boundary points if available
if os.path.exists(BND_CSV):
    B = pd.read_csv(BND_CSV)
    if "component" in B.columns and "lam" in B.columns and "rho" in B.columns:
        for comp, dfc in B.groupby("component"):
            axA.plot(dfc["lam"].to_numpy(float), dfc["rho"].to_numpy(float), linewidth=1.6)

# mark the rho-slice and predicted/inflection points
axA.axhline(rho0, linewidth=1.0, linestyle="--")
axA.scatter([lam_pred], [rho0], s=25, marker="x")
if lam_infl is not None:
    axA.scatter([lam_infl], [rho0], s=25, marker="o")

axA.set_xlabel(r"$\lambda$")
axA.set_ylabel(r"$\rho$")
axA.set_title("(a) Boundary-risk map + $f=0$")

cb = fig.colorbar(im, ax=axA, fraction=0.046, pad=0.02)
cb.set_label("risk" if has_risk else r"$|f|$")

# =======================
# Panel (b): score along λ-sweep
# =======================
axB = fig.add_subplot(1,3,2)
axB.plot(lam, risk_slice, linewidth=2.0)
axB.axvline(lam_pred, linewidth=1.2, linestyle="--")
if lam_infl is not None:
    axB.axvline(lam_infl, linewidth=1.2, linestyle=":")
axB.set_xlabel(r"$\lambda$")
axB.set_title(f"(b) {score_name} at $\\rho={rho0:.3f}$")
axB.set_ylabel("score")
axB.text(0.02, 0.98, f"$\\lambda_{{pred}}={lam_pred:.3f}$", transform=axB.transAxes, va="top")
if lam_infl is not None:
    axB.text(0.02, 0.90, f"$\\lambda_{{infl}}={lam_infl:.3f}$", transform=axB.transAxes, va="top")

# =======================
# Panel (c): primitives dist and TI
# =======================
axC = fig.add_subplot(1,3,3)
axC.plot(lam, dist_slice, linewidth=2.0)
axC.set_xlabel(r"$\lambda$")
axC.set_ylabel(r"dist proxy $|f|/\|\nabla f\|$")
axC.set_title(f"(c) dist & transversality at $\\rho={rho0:.3f}$")

axC2 = axC.twinx()
axC2.plot(lam, TI_slice, linewidth=2.0)
axC2.set_ylabel(r"TI magnitude $|\partial_\lambda f|$")

# vertical markers
axC.axvline(lam_pred, linewidth=1.2, linestyle="--")
if lam_infl is not None:
    axC.axvline(lam_infl, linewidth=1.2, linestyle=":")

fig.tight_layout()
fig.savefig(OUT_PDF, bbox_inches="tight")
plt.close(fig)

print("[OK] wrote", OUT_PDF, "| mtime:", time.ctime(os.path.getmtime(OUT_PDF)))
print("rho_slice used:", rho0, "| λ_pred:", lam_pred)
print(infl_info)
if lam_infl is not None:
    print("abs error |λ_pred-λ_infl| =", abs(lam_pred - lam_infl))
!ls -lh "$OUT_PDF"
```

```python
# --- install (Colabならだいたい不要だが念のため) ---
!pip -q install pymupdf opencv-python pillow

import fitz, re, os, math
import numpy as np
import cv2
from PIL import Image

FIG10_PDF = "Fig10.pdf"
FIG11_PDF = "Fig11.pdf"

def render_page(pdf_path, zoom=3.0):
    doc = fitz.open(pdf_path)
    page = doc.load_page(0)
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat, alpha=False)
    img = np.frombuffer(pix.samples, dtype=np.uint8).reshape(pix.height, pix.width, 3)
    return img, page, zoom

def find_y_of_text(page, pattern):
    # pattern例: r"Figure\s*10"
    hits = []
    for b in page.get_text("blocks"):
        x0,y0,x1,y1,txt,*_ = b
        if re.search(pattern, txt, re.IGNORECASE):
            hits.append((x0,y0,x1,y1,txt))
    if not hits:
        return None
    # 最初の出現を採用
    return hits[0]

def crop_band_around_y(img, y_center_px, band_h_px=2200):
    h, w = img.shape[:2]
    y0 = max(0, int(y_center_px - band_h_px//2))
    y1 = min(h, int(y_center_px + band_h_px//2))
    return img[y0:y1, :, :], y0, y1

def detect_panels(band_img, expect_n):
    # 暗背景→明るい矩形(パネル)を拾う
    gray = cv2.cvtColor(band_img, cv2.COLOR_BGR2GRAY)
    # 明部抽出（閾値はPDF次第で微調整可）
    _, th = cv2.threshold(gray, 80, 255, cv2.THRESH_BINARY)
    th = cv2.morphologyEx(th, cv2.MORPH_CLOSE, np.ones((7,7),np.uint8), iterations=2)

    cnts, _ = cv2.findContours(th, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    boxes = []
    H, W = gray.shape[:2]
    for c in cnts:
        x,y,w,h = cv2.boundingRect(c)
        area = w*h
        if area < 0.01*H*W:  # 小さすぎるもの除外
            continue
        ar = w / max(h,1)
        # 図パネルっぽい縦横比を緩く許容
        if 0.6 < ar < 2.2:
            boxes.append((area,x,y,w,h))
    boxes.sort(reverse=True)

    # 上位から採用（足りない場合は閾値調整）
    picked = [(x,y,w,h) for (area,x,y,w,h) in boxes[:expect_n]]
    # 左上→右下順に並べ替え
    picked.sort(key=lambda b: (b[1]//200, b[0]))
    return picked

def save_panels(pdf_path, fig_no, expect_n, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    img, page, zoom = render_page(pdf_path, zoom=3.0)

    hit = find_y_of_text(page, rf"Figure\s*{fig_no}")
    if hit is None:
        raise RuntimeError(f"Could not find 'Figure {fig_no}' text in {pdf_path}.")
    x0,y0,x1,y1,_ = hit
    y_center_px = int(((y0+y1)/2.0) * zoom)

    band, band_y0, band_y1 = crop_band_around_y(img, y_center_px, band_h_px=2600)

    boxes = detect_panels(band, expect_n=expect_n)
    if len(boxes) != expect_n:
        print(f"[WARN] detected panels = {len(boxes)} (expected {expect_n}). Try changing threshold/band_h.")
        print("boxes:", boxes)

    paths=[]
    for i,(x,y,w,h) in enumerate(boxes, start=1):
        pad = 10
        x0p = max(0, x-pad); y0p = max(0, y-pad)
        x1p = min(band.shape[1], x+w+pad); y1p = min(band.shape[0], y+h+pad)
        panel = band[y0p:y1p, x0p:x1p, :]
        out = os.path.join(out_dir, f"Fig{fig_no}_panel_{i:02d}.png")
        cv2.imwrite(out, panel)
        paths.append(out)

    # debug保存（band確認用）
    cv2.imwrite(os.path.join(out_dir, f"Fig{fig_no}_band.png"), band)
    return paths

p10 = save_panels(FIG10_PDF, fig_no=10, expect_n=2, out_dir="digitize_panels")
p11 = save_panels(FIG11_PDF, fig_no=11, expect_n=6, out_dir="digitize_panels")

print("[OK] saved panels:")
print("Fig10:", *p10, sep="\n  ")
print("Fig11:", *p11, sep="\n  ")
```

```python
import os, re, math
import numpy as np
import pandas as pd
import cv2

# --------- 手でここだけ埋める（図の軸目盛りを見て） ----------
# x は λ、y は 縦軸量（例: S_G/N_T^U など）
AXIS = {
    # Fig10 (2 panels)
    "Fig10_panel_01": dict(xmin=0.0, xmax=4.0, ymin=0.0, ymax=1.0),
    "Fig10_panel_02": dict(xmin=0.0, xmax=4.0, ymin=0.0, ymax=1.0),

    # Fig11 (6 panels)
    "Fig11_panel_01": dict(xmin=0.0, xmax=4.0, ymin=0.0, ymax=1.0),
    "Fig11_panel_02": dict(xmin=0.0, xmax=4.0, ymin=0.0, ymax=1.0),
    "Fig11_panel_03": dict(xmin=0.0, xmax=4.0, ymin=0.0, ymax=1.0),
    "Fig11_panel_04": dict(xmin=0.0, xmax=4.0, ymin=0.0, ymax=1.0),
    "Fig11_panel_05": dict(xmin=0.0, xmax=4.0, ymin=0.0, ymax=1.0),
    "Fig11_panel_06": dict(xmin=0.0, xmax=4.0, ymin=0.0, ymax=1.0),
}

# --------- 抽出設定（色点を拾う） ----------
# 彩度が高い画素を「点/曲線」とみなす（暗背景PDFでも効きやすい）
SAT_MIN = 60     # 低いと背景ノイズ拾う、高いと点を取りこぼす
VAL_MIN = 40
MIN_BLOB_AREA = 6  # 小さすぎる連結成分を捨てる

def detect_plot_box(img):
    """
    パネル内の『プロット枠（軸がある矩形）』を推定。
    うまくいかなければ pad をいじる or 手で bbox 指定に変える。
    """
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # エッジ→輪郭→最大矩形っぽいもの
    edges = cv2.Canny(gray, 50, 150)
    edges = cv2.dilate(edges, np.ones((3,3),np.uint8), iterations=1)
    cnts, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    H,W = gray.shape[:2]
    best=None
    best_score=-1
    for c in cnts:
        x,y,w,h = cv2.boundingRect(c)
        area=w*h
        if area < 0.15*H*W:
            continue
        ar=w/max(h,1)
        if 0.8 < ar < 1.8:
            score=area
            if score>best_score:
                best_score=score
                best=(x,y,w,h)
    if best is None:
        # fallback: パネル全体を使う
        return (0,0,W,H)
    # 少し内側に寄せる
    x,y,w,h=best
    pad=12
    return (max(0,x+pad), max(0,y+pad), max(10,w-2*pad), max(10,h-2*pad))

def extract_points(panel_path):
    key = os.path.splitext(os.path.basename(panel_path))[0]  # Fig10_panel_01
    img = cv2.imread(panel_path)
    H,W = img.shape[:2]

    x,y,w,h = detect_plot_box(img)
    plot = img[y:y+h, x:x+w, :]

    hsv = cv2.cvtColor(plot, cv2.COLOR_BGR2HSV)
    sat = hsv[:,:,1]
    val = hsv[:,:,2]
    mask = (sat>=SAT_MIN) & (val>=VAL_MIN)
    mask = mask.astype(np.uint8)*255
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, np.ones((3,3),np.uint8), iterations=1)

    # 連結成分→重心
    num, labels, stats, centroids = cv2.connectedComponentsWithStats(mask, connectivity=8)
    pts=[]
    for i in range(1, num):
        area = stats[i, cv2.CC_STAT_AREA]
        if area < MIN_BLOB_AREA:
            continue
        cx, cy = centroids[i]
        # plot座標系→panel座標系→全体座標
        px = x + cx
        py = y + cy
        pts.append((px, py, area))

    if key not in AXIS:
        raise RuntimeError(f"AXIS['{key}'] is not defined.")
    ax = AXIS[key]

    # 座標変換：plot box 左→右が xmin→xmax、下→上が ymin→ymax
    xmin,xmax,ymin,ymax = ax["xmin"],ax["xmax"],ax["ymin"],ax["ymax"]
    # 注意：画像yは下向き
    out=[]
    for (px,py,area) in pts:
        xd = xmin + (px - x)/max(w,1) * (xmax-xmin)
        yd = ymax - (py - y)/max(h,1) * (ymax-ymin)
        out.append((key, xd, yd, area))
    return out

# ---- run all panels ----
panel_dir="digitize_panels"
panels=sorted([os.path.join(panel_dir,f) for f in os.listdir(panel_dir) if re.match(r"Fig(10|11)_panel_\\d+\\.png", f)])
rows=[]
for p in panels:
    rows += extract_points(p)

df = pd.DataFrame(rows, columns=["panel","lambda","y","blob_area"])
df.to_csv("koibuchi_fig10_fig11_digitized_raw.csv", index=False)
print("[OK] wrote koibuchi_fig10_fig11_digitized_raw.csv  n=", len(df))
df.head()
```

```python
import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# 0) 入力CSVの自動選択（必要なら固定してOK）
# =========================
CAND = [
    "risk_vs_lambda_rho_0.30_max.csv",
    "risk_vs_lambda_rho_0.30_mean.csv",
    "risk_vs_lambda_rho_0.30.csv",
    "risk_vs_lambda_agg_over_rho.csv",
]
IN_CSV = None
for c in CAND:
    if os.path.exists(c):
        IN_CSV = c
        break
if IN_CSV is None:
    hits = glob.glob("risk_vs_lambda*.csv")
    IN_CSV = hits[0] if hits else None

print("IN_CSV =", IN_CSV)
assert IN_CSV is not None, "risk_vs_lambda*.csv が見つかりません"

df = pd.read_csv(IN_CSV)

# 列名の推定（lambda と risk を含む列を探す）
def pick_col(cols, keys):
    cols_l = [c.lower() for c in cols]
    for k in keys:
        for c, cl in zip(cols, cols_l):
            if k in cl:
                return c
    return None

col_x = pick_col(df.columns, ["lambda","lam"])
col_y = pick_col(df.columns, ["risk"])
assert col_x is not None and col_y is not None, f"列名が見つかりません: {df.columns}"

df = df[[col_x, col_y]].rename(columns={col_x:"lambda", col_y:"risk"}).dropna()
df = df.sort_values("lambda").reset_index(drop=True)

x = df["lambda"].to_numpy(dtype=float)
y = df["risk"].to_numpy(dtype=float)

# =========================
# 1) スムージング（Savitzky-Golay があれば使用、なければ移動平均）
# =========================
y_s = y.copy()

use_savgol = True
try:
    from scipy.signal import savgol_filter
except Exception as e:
    use_savgol = False
    print("[WARN] scipy.signal.savgol_filter が使えないので移動平均にフォールバック:", e)

N = len(y)
if use_savgol:
    # window は奇数・polyorderより大きい必要
    w = int(max(11, (0.03*N)//2*2 + 1))     # データ長の約3%（最低11）
    w = min(w, N-1 if (N-1)%2==1 else N-2)  # N未満の奇数に調整
    w = max(w, 7)                           # 最低7
    if w <= 3:
        w = 5
    poly = 3 if w >= 7 else 2
    y_s = savgol_filter(y, window_length=w, polyorder=poly, mode="interp")
    print(f"[INFO] Savitzky–Golay: window={w}, poly={poly}")
else:
    # 移動平均（奇数窓）
    w = int(max(9, (0.03*N)//2*2 + 1))
    w = min(w, N-1 if (N-1)%2==1 else N-2)
    w = max(w, 5)
    k = w//2
    y_pad = np.pad(y, (k,k), mode="edge")
    y_s = np.convolve(y_pad, np.ones(w)/w, mode="valid")
    print(f"[INFO] Moving average: window={w}")

# =========================
# 2) 1階/2階微分（非等間隔にも対応：np.gradientにxを渡す）
# =========================
d1 = np.gradient(y_s, x)
d2 = np.gradient(d1, x)
d3 = np.gradient(d2, x)

# ピーク（参考）
i_peak = int(np.argmax(y_s))
x_peak = float(x[i_peak])
y_peak = float(y_s[i_peak])

# =========================
# 3) 変曲点候補：d2 の符号反転（ゼロ交差）を検出し、線形補間で位置推定
# =========================
sgn = np.sign(d2)
# 0が混ざると不安定なので、0は直近の符号で埋める
for i in range(1, len(sgn)):
    if sgn[i] == 0:
        sgn[i] = sgn[i-1] if sgn[i-1] != 0 else 0
# まだ0があるなら前後で埋める
for i in range(len(sgn)-2, -1, -1):
    if sgn[i] == 0:
        sgn[i] = sgn[i+1] if sgn[i+1] != 0 else 0

cross_idx = np.where(sgn[:-1]*sgn[1:] < 0)[0]  # i と i+1 の間で符号が変わる

# フィルタ閾値（曲率が十分大きいところだけ）
curv = np.abs(d2)
curv_thr = 0.05 * np.max(curv)  # 5%（必要なら 0.02 などに下げる）
margin = 3                      # 端の数点は除外

cands = []
for i in cross_idx:
    if i < margin or i > len(x)-margin-2:
        continue
    # 曲率が小さすぎる交差は除外（ノイズ対策）
    if max(abs(d2[i]), abs(d2[i+1])) < curv_thr:
        continue

    # 線形補間で d2=0 の位置
    x0, x1 = x[i], x[i+1]
    f0, f1 = d2[i], d2[i+1]
    t = -f0 / (f1 - f0)  # f0 + t*(f1-f0)=0
    lam0 = x0 + t*(x1 - x0)

    # risk 値も補間（スムーズ系列上）
    y0 = y_s[i] + t*(y_s[i+1] - y_s[i])
    # 参考：d3（変曲点の「鋭さ」）
    d3_0 = d3[i] + t*(d3[i+1] - d3[i])

    side = "left" if lam0 < x_peak else "right"
    cands.append((lam0, y0, d3_0, side))

cand_df = pd.DataFrame(cands, columns=["lambda_inflect","risk_smooth","d3","side"])
cand_df = cand_df.sort_values("lambda_inflect").reset_index(drop=True)

print("candidates =", len(cand_df))
display(cand_df.head(20))

# =========================
# 4) 最終採用：ピークの左右で「ピークに最も近い」変曲点を1つずつ選ぶ
# =========================
chosen = []
if len(cand_df) > 0:
    left_df  = cand_df[cand_df["side"]=="left"].copy()
    right_df = cand_df[cand_df["side"]=="right"].copy()

    if len(left_df) > 0:
        left_df["dist_to_peak"] = np.abs(left_df["lambda_inflect"] - x_peak)
        chosen.append(left_df.sort_values("dist_to_peak").iloc[0])
    if len(right_df) > 0:
        right_df["dist_to_peak"] = np.abs(right_df["lambda_inflect"] - x_peak)
        chosen.append(right_df.sort_values("dist_to_peak").iloc[0])

chosen_df = pd.DataFrame(chosen).drop(columns=["dist_to_peak"], errors="ignore")
print("CHOSEN (left/right near peak):")
display(chosen_df)

# 保存（候補一覧＋採用点）
cand_out = IN_CSV.replace(".csv", "_inflection_candidates.csv")
chosen_out = IN_CSV.replace(".csv", "_inflection_chosen.csv")
cand_df.to_csv(cand_out, index=False)
chosen_df.to_csv(chosen_out, index=False)
print("saved:", cand_out)
print("saved:", chosen_out)

# =========================
# 5) 図（risk曲線 + 変曲点マーク + 2階微分も参考で描く）
# =========================
fig_pdf = IN_CSV.replace(".csv", "_inflection.pdf")

plt.figure()
plt.plot(x, y_s, label="risk (smoothed)")
plt.plot(x, y, alpha=0.25, label="risk (raw)")
plt.axvline(x_peak, linestyle="--", label=f"peak λ={x_peak:.3f}")

# 候補点（薄い点）
if len(cand_df)>0:
    plt.scatter(cand_df["lambda_inflect"], cand_df["risk_smooth"], s=25, marker="o", label="inflection candidates")

# 採用点（強調）
if len(chosen_df)>0:
    plt.scatter(chosen_df["lambda_inflect"], chosen_df["risk_smooth"], s=80, marker="x", label="CHOSEN inflection")

plt.xlabel("lambda")
plt.ylabel("risk")
plt.title("Inflection point detection on risk vs lambda")
plt.grid(True)
plt.legend()
plt.savefig(fig_pdf, bbox_inches="tight")
plt.show()
print("saved:", fig_pdf)

# 参考：2階微分のゼロ交差を確認したい場合（任意）
fig2_pdf = IN_CSV.replace(".csv", "_d2.pdf")
plt.figure()
plt.plot(x, d2, label="d2 (second derivative)")
plt.axhline(0.0)
plt.axvline(x_peak, linestyle="--", label="peak")
if len(chosen_df)>0:
    for _,r in chosen_df.iterrows():
        plt.axvline(r["lambda_inflect"], linestyle=":", label=f"inflect λ={r['lambda_inflect']:.3f}")
plt.xlabel("lambda")
plt.ylabel("d2")
plt.title("Second derivative (zero crossings = inflections)")
plt.grid(True)
plt.legend()
plt.savefig(fig2_pdf, bbox_inches="tight")
plt.show()
print("saved:", fig2_pdf)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

# ===== 入力（今あるファイル名に合わせる）=====
RISK_CSV   = "risk_vs_lambda_rho_0.30_max.csv"                  # あなたのρ固定 sweep（集約済み）
DTT_CHOSEN = "risk_vs_lambda_rho_0.30_max_inflection_chosen.csv" # さっき生成した chosen
KOIB_CSV   = "koibuchi_fig11_digitized_boundary.csv"             # Koibuchi境界（digitize）

assert os.path.exists(RISK_CSV),   f"not found: {RISK_CSV}"
assert os.path.exists(DTT_CHOSEN), f"not found: {DTT_CHOSEN}"
assert os.path.exists(KOIB_CSV),   f"not found: {KOIB_CSV}"

# ===== ローダ（列名を自動推定）=====
def pick_col(cols, keys):
    cols_l = [c.lower() for c in cols]
    for k in keys:
        for c, cl in zip(cols, cols_l):
            if k in cl:
                return c
    return None

# ===== 1) DTT側（risk曲線 & 予測変曲点）=====
risk_df = pd.read_csv(RISK_CSV)
xcol = pick_col(risk_df.columns, ["lambda","lam"])
ycol = pick_col(risk_df.columns, ["risk"])
risk_df = risk_df[[xcol,ycol]].rename(columns={xcol:"lambda", ycol:"risk"}).dropna()
risk_df = risk_df.sort_values("lambda").reset_index(drop=True)

x = risk_df["lambda"].to_numpy(float)
y = risk_df["risk"].to_numpy(float)
lam_peak = float(x[np.argmax(y)])

dtt = pd.read_csv(DTT_CHOSEN)
dtt_lams = dtt["lambda_inflect"].to_numpy(float) if "lambda_inflect" in dtt.columns else np.array([])
print("[DTT] peak lambda =", lam_peak)
print("[DTT] chosen inflections =", dtt_lams)

# ===== 2) Koibuchi境界の変曲点（rho(lambda) の d2=0 を推定）=====
kb = pd.read_csv(KOIB_CSV)
kb_lam = pick_col(kb.columns, ["lambda","lam"])
kb_rho = pick_col(kb.columns, ["rho","d/d","ratio","y"])  # digitizeだとx,yの可能性もある
assert kb_lam is not None and kb_rho is not None, f"Koibuchi列名が推定できない: {kb.columns}"

kb = kb[[kb_lam, kb_rho]].rename(columns={kb_lam:"lambda", kb_rho:"rho"}).dropna()
kb = kb.sort_values("lambda").reset_index(drop=True)

# 重複λを潰す（digitizeの散らばり対策：中央値）
kb2 = kb.groupby("lambda")["rho"].median().reset_index().sort_values("lambda").reset_index(drop=True)

xl = kb2["lambda"].to_numpy(float)
rl = kb2["rho"].to_numpy(float)

# スムージング（scipyがあればSavitzky、なければ移動平均）
use_savgol = True
try:
    from scipy.signal import savgol_filter
except Exception:
    use_savgol = False

N = len(rl)
if use_savgol and N >= 9:
    w = int(max(11, (0.07*N)//2*2 + 1))  # 境界は粗いので7%くらいから
    w = min(w, N-1 if (N-1)%2==1 else N-2)
    w = max(w, 7)
    poly = 3 if w >= 7 else 2
    r_s = savgol_filter(rl, window_length=w, polyorder=poly, mode="interp")
    print(f"[Koibuchi] Savitzky–Golay: window={w}, poly={poly}")
else:
    w = int(max(9, (0.07*N)//2*2 + 1))
    w = min(w, N-1 if (N-1)%2==1 else N-2)
    w = max(w, 5)
    k = w//2
    r_pad = np.pad(rl, (k,k), mode="edge")
    r_s = np.convolve(r_pad, np.ones(w)/w, mode="valid")
    print(f"[Koibuchi] Moving average: window={w}")

# 2階微分→ゼロ交差
d1 = np.gradient(r_s, xl)
d2 = np.gradient(d1, xl)

sgn = np.sign(d2)
# 0埋め
for i in range(1,len(sgn)):
    if sgn[i]==0: sgn[i]=sgn[i-1]
for i in range(len(sgn)-2,-1,-1):
    if sgn[i]==0: sgn[i]=sgn[i+1]

cross = np.where(sgn[:-1]*sgn[1:] < 0)[0]

# 曲率が小さすぎる交差は除外
curv = np.abs(d2)
thr = 0.05*np.max(curv) if np.max(curv)>0 else 0.0

cands = []
for i in cross:
    if max(abs(d2[i]),abs(d2[i+1])) < thr:
        continue
    x0,x1 = xl[i], xl[i+1]
    f0,f1 = d2[i], d2[i+1]
    t = -f0/(f1-f0)
    lam0 = x0 + t*(x1-x0)
    rho0 = r_s[i] + t*(r_s[i+1]-r_s[i])
    cands.append((lam0,rho0))

cand_df = pd.DataFrame(cands, columns=["lambda_inflect","rho_smooth"]).sort_values("lambda_inflect").reset_index(drop=True)
print("[Koibuchi] inflection candidates:", len(cand_df))
display(cand_df.head(20))

# DTTピークに最も近いKoibuchi変曲点を代表として採用
if len(cand_df)>0:
    cand_df["dist_to_DTT_peak"] = np.abs(cand_df["lambda_inflect"] - lam_peak)
    kb_pick = cand_df.sort_values("dist_to_DTT_peak").iloc[0]
    kb_lam_star = float(kb_pick["lambda_inflect"])
    kb_rho_star = float(kb_pick["rho_smooth"])
else:
    kb_lam_star = None
    kb_rho_star = None

print("[Koibuchi] chosen (nearest to DTT peak):", kb_lam_star, kb_rho_star)

# 保存
cand_df.to_csv("koibuchi_inflection_candidates.csv", index=False)
pd.DataFrame([{"lambda_inflect":kb_lam_star, "rho_smooth":kb_rho_star}]).to_csv("koibuchi_inflection_chosen.csv", index=False)

# ===== 3) 重ね図 (A) risk vs lambda + DTT inflection + Koibuchi inflection =====
pdfA = "Fig_overlay_inflection_risk.pdf"
plt.figure()
plt.plot(x, y, label="DTT risk (rho≈0.30)")
plt.axvline(lam_peak, linestyle="--", label=f"DTT peak λ={lam_peak:.3f}")

for lam in dtt_lams:
    plt.axvline(float(lam), linestyle=":", label=f"DTT inflect λ={float(lam):.3f}")

if kb_lam_star is not None:
    plt.axvline(kb_lam_star, linewidth=2, label=f"Koibuchi inflect λ={kb_lam_star:.3f}")

plt.xlabel("lambda")
plt.ylabel("risk")
plt.title("DTT-predicted inflections vs Koibuchi boundary inflection (lambda-axis)")
plt.grid(True)
plt.legend()
plt.savefig(pdfA, bbox_inches="tight")
plt.show()
print("saved:", pdfA)

# ===== 4) 重ね図 (B) Koibuchi boundary rho(lambda) + inflection marker =====
pdfB = "Fig_overlay_inflection_koibuchi_boundary.pdf"
plt.figure()
plt.plot(xl, rl, alpha=0.25, label="Koibuchi digitized (raw)")
plt.plot(xl, r_s, label="Koibuchi boundary (smoothed)")
plt.axvline(lam_peak, linestyle="--", label="DTT peak λ")
if kb_lam_star is not None:
    plt.scatter([kb_lam_star],[kb_rho_star], s=80, marker="x", label="Koibuchi inflection")
plt.xlabel("lambda")
plt.ylabel("rho (boundary)")
plt.title("Koibuchi boundary and its inflection (digitized)")
plt.grid(True)
plt.legend()
plt.savefig(pdfB, bbox_inches="tight")
plt.show()
print("saved:", pdfB)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os, glob

# ===== 入力（DTT側）=====
RISK_CSV   = "risk_vs_lambda_rho_0.30_max.csv"
DTT_CHOSEN = "risk_vs_lambda_rho_0.30_max_inflection_chosen.csv"

# ===== Koibuchi側：自動発見 =====
def find_first(patterns):
    for pat in patterns:
        hits = sorted(glob.glob(pat))
        if hits:
            return hits[0]
    return None

KOIB_CSV = find_first([
    "koibuchi*_digitized_boundary.csv",
    "koibuchi*Fig11*_digitized_boundary.csv",
    "koibuchi*fig11*_digitized_boundary.csv",
    "koibuchi*digitized*boundary*.csv",
    # 最悪これ（rawからでも列推定で行けるが、まずはboundaryを優先）
    "koibuchi*digitized*.csv",
])

print("RISK_CSV =", RISK_CSV)
print("DTT_CHOSEN =", DTT_CHOSEN)
print("KOIB_CSV =", KOIB_CSV)

assert os.path.exists(RISK_CSV),   f"not found: {RISK_CSV}"
assert os.path.exists(DTT_CHOSEN), f"not found: {DTT_CHOSEN}"
assert KOIB_CSV is not None and os.path.exists(KOIB_CSV), f"not found: {KOIB_CSV}"

# ===== ローダ（列名を自動推定）=====
def pick_col(cols, keys):
    cols_l = [c.lower() for c in cols]
    for k in keys:
        for c, cl in zip(cols, cols_l):
            if k in cl:
                return c
    return None

# ===== 1) DTT側（risk曲線 & 予測変曲点）=====
risk_df = pd.read_csv(RISK_CSV)
xcol = pick_col(risk_df.columns, ["lambda","lam"])
ycol = pick_col(risk_df.columns, ["risk"])
risk_df = risk_df[[xcol,ycol]].rename(columns={xcol:"lambda", ycol:"risk"}).dropna()
risk_df = risk_df.sort_values("lambda").reset_index(drop=True)

x = risk_df["lambda"].to_numpy(float)
y = risk_df["risk"].to_numpy(float)
lam_peak = float(x[np.argmax(y)])

dtt = pd.read_csv(DTT_CHOSEN)
dtt_lams = dtt["lambda_inflect"].to_numpy(float)

print("[DTT] peak lambda =", lam_peak)
print("[DTT] chosen inflections =", dtt_lams)

# ===== 2) Koibuchi境界の変曲点（rho(lambda) の d2=0 を推定）=====
kb = pd.read_csv(KOIB_CSV)
kb_lam = pick_col(kb.columns, ["lambda","lam","x"])
kb_rho = pick_col(kb.columns, ["rho","d/d","ratio","y"])
assert kb_lam is not None and kb_rho is not None, f"Koibuchi列名が推定できない: {kb.columns}"

kb = kb[[kb_lam, kb_rho]].rename(columns={kb_lam:"lambda", kb_rho:"rho"}).dropna()
kb = kb.sort_values("lambda").reset_index(drop=True)

# 重複λは中央値で潰す（digitizeの散らばり対策）
kb2 = kb.groupby("lambda")["rho"].median().reset_index().sort_values("lambda").reset_index(drop=True)

xl = kb2["lambda"].to_numpy(float)
rl = kb2["rho"].to_numpy(float)

# スムージング（scipyがあればSavitzky）
use_savgol = True
try:
    from scipy.signal import savgol_filter
except Exception:
    use_savgol = False

N = len(rl)
if use_savgol and N >= 9:
    w = int(max(11, (0.07*N)//2*2 + 1))
    w = min(w, N-1 if (N-1)%2==1 else N-2)
    w = max(w, 7)
    poly = 3 if w >= 7 else 2
    r_s = savgol_filter(rl, window_length=w, polyorder=poly, mode="interp")
else:
    w = int(max(9, (0.07*N)//2*2 + 1))
    w = min(w, N-1 if (N-1)%2==1 else N-2)
    w = max(w, 5)
    k = w//2
    r_pad = np.pad(rl, (k,k), mode="edge")
    r_s = np.convolve(r_pad, np.ones(w)/w, mode="valid")

d1 = np.gradient(r_s, xl)
d2 = np.gradient(d1, xl)

sgn = np.sign(d2)
for i in range(1,len(sgn)):
    if sgn[i]==0: sgn[i]=sgn[i-1]
for i in range(len(sgn)-2,-1,-1):
    if sgn[i]==0: sgn[i]=sgn[i+1]

cross = np.where(sgn[:-1]*sgn[1:] < 0)[0]

curv = np.abs(d2)
thr = 0.05*np.max(curv) if np.max(curv)>0 else 0.0

cands = []
for i in cross:
    if max(abs(d2[i]),abs(d2[i+1])) < thr:
        continue
    x0,x1 = xl[i], xl[i+1]
    f0,f1 = d2[i], d2[i+1]
    t = -f0/(f1-f0)
    lam0 = x0 + t*(x1-x0)
    rho0 = r_s[i] + t*(r_s[i+1]-r_s[i])
    cands.append((lam0,rho0))

cand_df = pd.DataFrame(cands, columns=["lambda_inflect","rho_smooth"]).sort_values("lambda_inflect").reset_index(drop=True)
cand_df.to_csv("koibuchi_inflection_candidates.csv", index=False)

if len(cand_df)>0:
    cand_df["dist_to_DTT_peak"] = np.abs(cand_df["lambda_inflect"] - lam_peak)
    kb_pick = cand_df.sort_values("dist_to_DTT_peak").iloc[0]
    kb_lam_star = float(kb_pick["lambda_inflect"])
    kb_rho_star = float(kb_pick["rho_smooth"])
else:
    kb_lam_star, kb_rho_star = None, None

pd.DataFrame([{"lambda_inflect":kb_lam_star, "rho_smooth":kb_rho_star}]).to_csv("koibuchi_inflection_chosen.csv", index=False)
print("[Koibuchi] chosen inflection λ =", kb_lam_star)

# ===== 3) 重ね図 (A) risk vs lambda + DTT inflection + Koibuchi inflection =====
pdfA = "Fig_overlay_inflection_risk.pdf"
plt.figure()
plt.plot(x, y, label="DTT risk (rho≈0.30)")
plt.axvline(lam_peak, linestyle="--", label=f"DTT peak λ={lam_peak:.3f}")
for lam in dtt_lams:
    plt.axvline(float(lam), linestyle=":", label=f"DTT inflect λ={float(lam):.3f}")
if kb_lam_star is not None:
    plt.axvline(kb_lam_star, linewidth=2, label=f"Koibuchi inflect λ={kb_lam_star:.3f}")
plt.xlabel("lambda"); plt.ylabel("risk")
plt.title("DTT inflections vs Koibuchi boundary inflection (lambda-axis)")
plt.grid(True); plt.legend()
plt.savefig(pdfA, bbox_inches="tight")
plt.show()
print("saved:", pdfA)

# ===== 4) 重ね図 (B) Koibuchi boundary rho(lambda) + inflection marker =====
pdfB = "Fig_overlay_inflection_koibuchi_boundary.pdf"
plt.figure()
plt.plot(xl, rl, alpha=0.25, label="Koibuchi digitized (raw)")
plt.plot(xl, r_s, label="Koibuchi boundary (smoothed)")
plt.axvline(lam_peak, linestyle="--", label="DTT peak λ")
if kb_lam_star is not None:
    plt.scatter([kb_lam_star],[kb_rho_star], s=80, marker="x", label="Koibuchi inflection")
plt.xlabel("lambda"); plt.ylabel("rho (boundary)")
plt.title("Koibuchi boundary and its inflection (digitized)")
plt.grid(True); plt.legend()
plt.savefig(pdfB, bbox_inches="tight")
plt.show()
print("saved:", pdfB)
```

```python
# ============================================================
# ONE-SHOT COMPLETE FIXED PIPELINE
#   - robust de-dup (dx=0 回避)
#   - smoothing (Savitzky–Golay 優先 / fallback あり)
#   - inflection detection (2階微分の符号反転 + 端点除外 + score)
#   - outputs: CSVs + PDF figure
#
# 使い方（Colabでこのセルをそのまま実行）:
#   1) INPUT_CSV を raw の境界点CSVに合わせる（例: koibuchi_fig11_digitized_boundary.csv）
#   2) X_COL/Y_COL が不明なら None のままでOK（自動推定）
#   3) 実行 → candidates/chosen CSV と PDF が生成される
#
# 追加（任意）: RISK_GRID_CSV を指定すると、境界上の risk(lambda) も同様に処理して図/CSV出力する
# ============================================================

import os, glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# USER SETTINGS (edit here)
# ---------------------------
INPUT_CSV = "koibuchi_fig11_digitized_boundary.csv"   # raw boundary points (lambda, rho) の想定
X_COL = None                                          # 例: "lambda" （不明なら None）
Y_COL = None                                          # 例: "rho"    （不明なら None）

# smoothing params
SMOOTH_METHOD = "savgol"   # "savgol" or "ma" (moving-average)
SAVGOL_POLYORDER = 3
WINDOW_FRACTION = 0.20     # データ長の何割を窓に使うか（0.1〜0.3推奨）
WINDOW_MIN = 11            # 最小窓（奇数）
MARGIN_FRAC = 0.03         # 端点除外割合（微分不安定域を除外）

# outputs prefix
OUT_PREFIX = "koibuchi_boundary"

# ---- OPTIONAL: evaluate risk on boundary (needs a grid CSV) ----
# 想定: grid CSV に列 {lambda, rho, risk} がある（列名は自動推定）
RISK_GRID_CSV = None   # 例: "FG_BDI_ZT_map_grid.csv" / "fg_BDI_ZT_map_grid.csv" など
RISK_COL = None        # grid 内の risk 列（不明なら None）
DO_RISK_PIPELINE = (RISK_GRID_CSV is not None)

# ---------------------------
# Helpers
# ---------------------------
def _auto_pick_xy_cols(df: pd.DataFrame):
    cols = list(df.columns)
    low = [c.lower() for c in cols]

    def pick(keys, avoid=()):
        for k in keys:
            for c, cl in zip(cols, low):
                if k in cl and not any(a in cl for a in avoid):
                    return c
        return None

    x = pick(["lambda", "lam", "x"])
    y = pick(["rho", "y"])
    return x, y

def _auto_pick_risk_cols(df: pd.DataFrame):
    cols = list(df.columns)
    low = [c.lower() for c in cols]
    x = None
    r = None
    for c, cl in zip(cols, low):
        if x is None and ("lambda" in cl or "lam" in cl):
            x = c
        if r is None and ("risk" in cl or "score" in cl):
            r = c
    return x, r

def _robust_unique_by_x(x, y, agg="median"):
    """
    dx=0 を根絶するため、x 重複を確実に潰す。
    """
    d = pd.DataFrame({"x": x, "y": y}).replace([np.inf, -np.inf], np.nan).dropna()
    d = d.sort_values("x")
    if agg == "median":
        u = d.groupby("x", as_index=False)["y"].median()
    elif agg == "mean":
        u = d.groupby("x", as_index=False)["y"].mean()
    else:
        raise ValueError("agg must be median or mean")
    return u["x"].to_numpy(), u["y"].to_numpy(), len(d), len(u)

def _choose_window(n, frac, wmin):
    w = int(max(wmin, (n * frac)))
    if w % 2 == 0:
        w += 1
    w = min(w, n if n % 2 == 1 else n - 1)
    if w < 5:
        w = 5 if n >= 5 else n
    if w % 2 == 0:
        w = max(3, w - 1)
    return w

def _smooth(y, method="savgol", window=31, polyorder=3):
    y = np.asarray(y, dtype=float)
    if len(y) < 7:
        return y.copy()

    if method == "savgol":
        try:
            from scipy.signal import savgol_filter
            w = min(window, len(y) if len(y) % 2 == 1 else len(y) - 1)
            if w < 5:
                return y.copy()
            po = min(polyorder, w - 2)
            return savgol_filter(y, window_length=w, polyorder=po, mode="interp")
        except Exception:
            # fallback to moving average
            method = "ma"

    if method == "ma":
        w = min(window, len(y) if len(y) % 2 == 1 else len(y) - 1)
        if w < 3:
            return y.copy()
        if w % 2 == 0:
            w += 1
        pad = w // 2
        y_pad = np.r_[np.full(pad, y[0]), y, np.full(pad, y[-1])]
        return np.convolve(y_pad, np.ones(w) / w, mode="valid")

    raise ValueError("Unknown smoothing method")

def _inflection_candidates(x, y_s, margin_frac=0.03):
    """
    2階微分の符号反転を候補にする。
    score は「符号反転点近傍での曲率ジャンプ量」を基本に定義。
    """
    x = np.asarray(x, dtype=float)
    y_s = np.asarray(y_s, dtype=float)

    # safety: x strictly increasing
    if np.any(np.diff(x) <= 0):
        raise ValueError("x must be strictly increasing after de-dup. Check sorting/aggregation.")

    d1 = np.gradient(y_s, x, edge_order=2)
    d2 = np.gradient(d1,  x, edge_order=2)

    sgn = np.sign(d2)
    idx = np.where(sgn[:-1] * sgn[1:] < 0)[0]  # sign flip between i and i+1

    margin = max(5, int(margin_frac * len(x)))
    idx = idx[(idx > margin) & (idx < len(x) - margin - 1)]

    # candidate x at i (or interpolate root approx)
    # root approx: linear interpolation of d2 between i and i+1
    cand_x = []
    cand_y = []
    cand_score = []
    for i in idx:
        x0, x1 = x[i], x[i + 1]
        f0, f1 = d2[i], d2[i + 1]
        if f1 == f0:
            xr = x0
        else:
            xr = x0 - f0 * (x1 - x0) / (f1 - f0)
        yr = np.interp(xr, x, y_s)
        # score: curvature jump magnitude + slope magnitude（ノイズ耐性のため軽く足す）
        jump = abs(f1 - f0)
        slope = abs(d1[i])
        cand_x.append(float(xr))
        cand_y.append(float(yr))
        cand_score.append(float(jump + 0.1 * slope))

    out = pd.DataFrame({"x_inflect": cand_x, "y_smooth": cand_y, "score": cand_score})
    out = out.sort_values("score", ascending=False).reset_index(drop=True)
    return out, d1, d2

def _find_file_or_glob(path):
    if path is None:
        return None
    if os.path.exists(path):
        return path
    hits = glob.glob(path)
    if hits:
        return hits[0]
    hits = glob.glob(f"**/{path}", recursive=True)
    if hits:
        return hits[0]
    return None

# ---------------------------
# Main: boundary inflection
# ---------------------------
in_path = _find_file_or_glob(INPUT_CSV)
if in_path is None:
    raise FileNotFoundError(f"INPUT_CSV not found: {INPUT_CSV}")

raw = pd.read_csv(in_path)
if X_COL is None or Y_COL is None:
    ax, ay = _auto_pick_xy_cols(raw)
    X_COL = ax if X_COL is None else X_COL
    Y_COL = ay if Y_COL is None else Y_COL
if X_COL is None or Y_COL is None:
    raise ValueError(f"Could not auto-detect X_COL/Y_COL from columns: {list(raw.columns)}")

x0 = raw[X_COL].to_numpy()
y0 = raw[Y_COL].to_numpy()

x, y, n_raw, n_uni = _robust_unique_by_x(x0, y0, agg="median")
win = _choose_window(len(y), WINDOW_FRACTION, WINDOW_MIN)
y_s = _smooth(y, method=SMOOTH_METHOD, window=win, polyorder=SAVGOL_POLYORDER)

cands, d1, d2 = _inflection_candidates(x, y_s, margin_frac=MARGIN_FRAC)

# save candidates / chosen
cand_csv = f"{OUT_PREFIX}_inflection_candidates.csv"
chosen_csv = f"{OUT_PREFIX}_inflection_chosen.csv"
cands.to_csv(cand_csv, index=False)
chosen = cands.head(1).copy()
chosen.to_csv(chosen_csv, index=False)

print("=== Boundary inflection pipeline (fixed) ===")
print(f"INPUT   : {in_path}")
print(f"X_COL,Y_COL : {X_COL}, {Y_COL}")
print(f"raw points   : {n_raw}")
print(f"unique x     : {n_uni}  (duplicates collapsed -> dx=0 eliminated)")
print(f"smooth window: {win}  method={SMOOTH_METHOD}")
print(f"candidates   : {len(cands)}")
if len(cands) > 0:
    print("top candidate:", chosen.iloc[0].to_dict())
print(f"saved: {cand_csv}")
print(f"saved: {chosen_csv}")

# figure
fig_pdf = f"Fig_{OUT_PREFIX}_inflection.pdf"
plt.figure(figsize=(7,4.5))
plt.scatter(x0, y0, s=8, alpha=0.25, label="raw points")
plt.plot(x, y, alpha=0.6, label="collapsed (median by x)")
plt.plot(x, y_s, linewidth=2.0, label="smoothed")

# mark candidates (top K)
K = min(5, len(cands))
for j in range(K):
    xx = cands.loc[j, "x_inflect"]
    yy = cands.loc[j, "y_smooth"]
    plt.axvline(xx, linestyle="--", alpha=0.8)
    plt.plot([xx],[yy], marker="o", markersize=6)

# emphasize chosen
if len(chosen) == 1:
    xx = chosen.loc[0, "x_inflect"]
    yy = chosen.loc[0, "y_smooth"]
    plt.axvline(xx, linestyle="-.", linewidth=2.0, alpha=0.9, label=f"chosen x≈{xx:.3f}")

plt.xlabel(X_COL)
plt.ylabel(Y_COL)
plt.title("Inflection detection (dedup + smoothing + sign-change of 2nd derivative)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig(fig_pdf, bbox_inches="tight")
plt.show()
print("saved:", fig_pdf)

# ---------------------------
# Optional: risk(lambda) on Koibuchi boundary
# ---------------------------
if DO_RISK_PIPELINE:
    grid_path = _find_file_or_glob(RISK_GRID_CSV)
    if grid_path is None:
        raise FileNotFoundError(f"RISK_GRID_CSV not found: {RISK_GRID_CSV}")

    grid = pd.read_csv(grid_path)

    # auto columns
    gx = None
    grho = None
    grisk = RISK_COL
    cols = list(grid.columns)
    low = [c.lower() for c in cols]
    for c, cl in zip(cols, low):
        if gx is None and ("lambda" in cl or "lam" in cl):
            gx = c
        if grho is None and ("rho" in cl):
            grho = c
        if grisk is None and ("risk" in cl or "score" in cl):
            grisk = c
    if gx is None or grho is None or grisk is None:
        raise ValueError(f"Could not detect (lambda,rho,risk) cols in grid: {cols}")

    # interpolate risk at boundary points (requires scipy)
    try:
        from scipy.interpolate import griddata
        pts = grid[[gx, grho]].to_numpy(dtype=float)
        vals = grid[grisk].to_numpy(dtype=float)

        # evaluate on raw boundary points
        bpts = raw[[X_COL, Y_COL]].to_numpy(dtype=float)
        risk_raw = griddata(pts, vals, bpts, method="linear")

        # fallback nearest for NaNs
        mask = np.isnan(risk_raw)
        if np.any(mask):
            risk_raw[mask] = griddata(pts, vals, bpts[mask], method="nearest")

        # collapse duplicates by lambda and compute risk(lambda)
        xr, yr, n_raw2, n_uni2 = _robust_unique_by_x(raw[X_COL].to_numpy(), risk_raw, agg="median")
        win2 = _choose_window(len(yr), WINDOW_FRACTION, WINDOW_MIN)
        yr_s = _smooth(yr, method=SMOOTH_METHOD, window=win2, polyorder=SAVGOL_POLYORDER)
        rcands, rd1, rd2 = _inflection_candidates(xr, yr_s, margin_frac=MARGIN_FRAC)

        # outputs
        risk_series_csv = f"{OUT_PREFIX}_risk_on_boundary_series.csv"
        pd.DataFrame({"lambda": xr, "risk": yr, "risk_smooth": yr_s}).to_csv(risk_series_csv, index=False)

        risk_cand_csv = f"{OUT_PREFIX}_risk_inflection_candidates.csv"
        risk_chosen_csv = f"{OUT_PREFIX}_risk_inflection_chosen.csv"
        rcands.to_csv(risk_cand_csv, index=False)
        rchosen = rcands.head(1).copy()
        rchosen.to_csv(risk_chosen_csv, index=False)

        print("\n=== Risk-on-boundary pipeline ===")
        print(f"GRID    : {grid_path}")
        print(f"cols    : lambda={gx}, rho={grho}, risk={grisk}")
        print(f"unique lambda in risk series: {n_uni2}")
        print(f"smooth window: {win2}")
        print(f"candidates: {len(rcands)}")
        if len(rcands) > 0:
            print("top candidate:", rchosen.iloc[0].to_dict())
        print(f"saved: {risk_series_csv}")
        print(f"saved: {risk_cand_csv}")
        print(f"saved: {risk_chosen_csv}")

        # figure
        rfig_pdf = f"Fig_{OUT_PREFIX}_risk_inflection.pdf"
        plt.figure(figsize=(7,4.5))
        plt.plot(xr, yr, alpha=0.6, label="risk (collapsed median)")
        plt.plot(xr, yr_s, linewidth=2.0, label="risk (smoothed)")

        KK = min(5, len(rcands))
        for j in range(KK):
            xx = rcands.loc[j, "x_inflect"]
            yy = rcands.loc[j, "y_smooth"]
            plt.axvline(xx, linestyle="--", alpha=0.8)
            plt.plot([xx],[yy], marker="o", markersize=6)

        if len(rchosen) == 1:
            xx = rchosen.loc[0, "x_inflect"]
            plt.axvline(xx, linestyle="-.", linewidth=2.0, alpha=0.9, label=f"chosen λ≈{xx:.3f}")

        plt.xlabel("lambda")
        plt.ylabel("risk")
        plt.title("Inflections of risk evaluated on Koibuchi boundary (fixed)")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.savefig(rfig_pdf, bbox_inches="tight")
        plt.show()
        print("saved:", rfig_pdf)

    except Exception as e:
        print("\n[WARN] risk-on-boundary skipped (SciPy griddata unavailable or error).")
        print("       Set RISK_GRID_CSV=None or install SciPy.")
        print("       Error:", repr(e))
```

```python
# ============================================================
# ONE-SHOT COMPLETE (COLAB SAFE) PIPELINE
#  - auto-mount Google Drive (if in Colab)
#  - auto-find INPUT_CSV (boundary points) if missing
#  - de-dup x to eliminate dx=0 warnings
#  - smoothing (Savitzky–Golay if available, else MA)
#  - inflection detection (sign change of 2nd derivative, endpoint excluded)
#  - outputs: candidates/chosen CSV + PDF figure
# ============================================================

import os, glob, difflib, time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ---------------------------
# USER SETTINGS
# ---------------------------
INPUT_CSV = "AUTO"   # "AUTO" recommended. Or set exact filename/path.
X_COL = None         # auto-detect if None (lambda-like)
Y_COL = None         # auto-detect if None (rho-like)

OUT_PREFIX = "koibuchi_boundary"
SMOOTH_METHOD = "savgol"   # "savgol" or "ma"
SAVGOL_POLYORDER = 3
WINDOW_FRACTION = 0.20
WINDOW_MIN = 11
MARGIN_FRAC = 0.03

# Optional: if you want risk-on-boundary, set to "AUTO" or a path; otherwise None
RISK_GRID_CSV = None   # e.g., "AUTO" or "FG_BDI_ZT_map_grid.csv"
RISK_COL = None        # auto-detect if None


# ---------------------------
# Colab Drive mount (safe)
# ---------------------------
def _maybe_mount_drive():
    try:
        import google.colab  # noqa
        from google.colab import drive
        if not os.path.exists("/content/drive"):
            os.makedirs("/content/drive", exist_ok=True)
        # If not mounted, this will prompt auth; if already mounted, it returns quickly.
        drive.mount("/content/drive", force_remount=False)
        return True
    except Exception:
        return False

_maybe_mount_drive()

# ---------------------------
# File search utilities
# ---------------------------
SEARCH_ROOTS = ["/content", "/content/drive"]  # drive may not exist; handled below

def _existing_roots():
    return [r for r in SEARCH_ROOTS if os.path.exists(r)]

def _glob_all(pattern):
    hits = []
    for root in _existing_roots():
        hits += glob.glob(os.path.join(root, pattern), recursive=True)
    return sorted(set(hits))

def _best_match(target_name, candidates):
    if not candidates:
        return None
    if target_name in ("AUTO", None, ""):
        # Heuristic preference order for boundary CSVs
        prefs = [
            "koibuchi_fig11_digitized_boundary",
            "koibuchi_fig10_fig11_digitized_raw",
            "fg_boundary_points",
            "boundary_points",
            "digitized_boundary",
            "digitized",
        ]
        for p in prefs:
            for c in candidates:
                if p in os.path.basename(c).lower():
                    return c
        # fallback: most recently modified
        return max(candidates, key=lambda p: os.path.getmtime(p))
    # If user gave a name, choose closest by similarity
    scored = []
    for c in candidates:
        base = os.path.basename(c)
        s = difflib.SequenceMatcher(None, target_name.lower(), base.lower()).ratio()
        scored.append((s, c))
    scored.sort(reverse=True)
    return scored[0][1]

def _resolve_csv(user_value, patterns):
    # 1) direct path
    if user_value not in ("AUTO", None, "") and os.path.exists(user_value):
        return user_value

    # 2) search by exact filename within roots
    if user_value not in ("AUTO", None, ""):
        hits = []
        for root in _existing_roots():
            hits += glob.glob(os.path.join(root, "**", user_value), recursive=True)
        if hits:
            return _best_match(user_value, hits)

    # 3) pattern search
    hits = []
    for pat in patterns:
        hits += _glob_all(f"**/{pat}")
    hits = sorted(set(hits))
    if hits:
        return _best_match(user_value, hits)

    return None

# ---------------------------
# Column auto-detect
# ---------------------------
def _auto_pick_xy_cols(df: pd.DataFrame):
    cols = list(df.columns)
    low = [c.lower() for c in cols]

    def pick(keys):
        for k in keys:
            for c, cl in zip(cols, low):
                if k in cl:
                    return c
        return None

    x = pick(["lambda", "lam"])
    y = pick(["rho"])
    # fallback: first two numeric columns
    if x is None or y is None:
        num_cols = [c for c in cols if np.issubdtype(df[c].dtype, np.number)]
        if len(num_cols) >= 2:
            x = x or num_cols[0]
            y = y or num_cols[1]
    return x, y

def _auto_pick_risk_cols(df: pd.DataFrame):
    cols = list(df.columns)
    low = [c.lower() for c in cols]
    lam = None
    rho = None
    risk = None
    for c, cl in zip(cols, low):
        if lam is None and ("lambda" in cl or "lam" in cl):
            lam = c
        if rho is None and ("rho" in cl):
            rho = c
        if risk is None and ("risk" in cl or "score" in cl):
            risk = c
    return lam, rho, risk

# ---------------------------
# Math helpers
# ---------------------------
def _robust_unique_by_x(x, y, agg="median"):
    d = pd.DataFrame({"x": x, "y": y}).replace([np.inf, -np.inf], np.nan).dropna()
    d = d.sort_values("x")
    if agg == "median":
        u = d.groupby("x", as_index=False)["y"].median()
    else:
        u = d.groupby("x", as_index=False)["y"].mean()
    return u["x"].to_numpy(), u["y"].to_numpy(), len(d), len(u)

def _choose_window(n, frac, wmin):
    w = int(max(wmin, n * frac))
    if w % 2 == 0:
        w += 1
    w = min(w, n if n % 2 == 1 else n - 1)
    if w < 5:
        w = 5 if n >= 5 else n
    if w % 2 == 0:
        w = max(3, w - 1)
    return w

def _smooth(y, method="savgol", window=31, polyorder=3):
    y = np.asarray(y, dtype=float)
    if len(y) < 7:
        return y.copy()
    if method == "savgol":
        try:
            from scipy.signal import savgol_filter
            w = min(window, len(y) if len(y) % 2 == 1 else len(y) - 1)
            if w < 5:
                return y.copy()
            po = min(polyorder, w - 2)
            return savgol_filter(y, window_length=w, polyorder=po, mode="interp")
        except Exception:
            method = "ma"
    if method == "ma":
        w = min(window, len(y) if len(y) % 2 == 1 else len(y) - 1)
        if w < 3:
            return y.copy()
        if w % 2 == 0:
            w += 1
        pad = w // 2
        y_pad = np.r_[np.full(pad, y[0]), y, np.full(pad, y[-1])]
        return np.convolve(y_pad, np.ones(w) / w, mode="valid")
    raise ValueError("Unknown smoothing method")

def _inflection_candidates(x, y_s, margin_frac=0.03):
    x = np.asarray(x, dtype=float)
    y_s = np.asarray(y_s, dtype=float)
    if np.any(np.diff(x) <= 0):
        raise ValueError("x must be strictly increasing after de-dup.")
    d1 = np.gradient(y_s, x, edge_order=2)
    d2 = np.gradient(d1,  x, edge_order=2)
    sgn = np.sign(d2)
    idx = np.where(sgn[:-1] * sgn[1:] < 0)[0]
    margin = max(5, int(margin_frac * len(x)))
    idx = idx[(idx > margin) & (idx < len(x) - margin - 1)]

    rows = []
    for i in idx:
        x0, x1 = x[i], x[i + 1]
        f0, f1 = d2[i], d2[i + 1]
        xr = x0 if (f1 == f0) else (x0 - f0 * (x1 - x0) / (f1 - f0))
        yr = float(np.interp(xr, x, y_s))
        jump = abs(f1 - f0)
        slope = abs(d1[i])
        score = float(jump + 0.1 * slope)
        rows.append((float(xr), yr, score))

    out = pd.DataFrame(rows, columns=["x_inflect", "y_smooth", "score"]).sort_values("score", ascending=False)
    out = out.reset_index(drop=True)
    return out, d1, d2

# ============================================================
# 1) Resolve INPUT_CSV (AUTO)
# ============================================================
boundary_patterns = [
    "koibuchi*digitized*boundary*.csv",
    "koibuchi*boundary*.csv",
    "*digitized*boundary*.csv",
    "*boundary_points*.csv",
    "fg_boundary_points.csv",
    "*.csv",
]

in_path = _resolve_csv(INPUT_CSV, boundary_patterns)

if in_path is None:
    print("=== FILE NOT FOUND ===")
    print("INPUT_CSV:", INPUT_CSV)
    print("Searched roots:", _existing_roots())
    print("Try uploading the CSV to /content, or mount Drive and ensure it exists under /content/drive.")
    print("\nCSV candidates currently visible:")
    cands = _glob_all("**/*.csv")
    for p in cands[:80]:
        print(" -", p)
    raise FileNotFoundError(f"INPUT_CSV not found: {INPUT_CSV}")

print("Using INPUT_CSV:", in_path)

raw = pd.read_csv(in_path)

if X_COL is None or Y_COL is None:
    ax, ay = _auto_pick_xy_cols(raw)
    X_COL = ax if X_COL is None else X_COL
    Y_COL = ay if Y_COL is None else Y_COL
if X_COL is None or Y_COL is None:
    raise ValueError(f"Could not detect X_COL/Y_COL from columns: {list(raw.columns)}")

x0 = raw[X_COL].to_numpy()
y0 = raw[Y_COL].to_numpy()

# de-dup (dx=0 eradication)
x, y, n_raw, n_uni = _robust_unique_by_x(x0, y0, agg="median")
win = _choose_window(len(y), WINDOW_FRACTION, WINDOW_MIN)
y_s = _smooth(y, method=SMOOTH_METHOD, window=win, polyorder=SAVGOL_POLYORDER)
cands, d1, d2 = _inflection_candidates(x, y_s, margin_frac=MARGIN_FRAC)

cand_csv = f"{OUT_PREFIX}_inflection_candidates.csv"
chosen_csv = f"{OUT_PREFIX}_inflection_chosen.csv"
cands.to_csv(cand_csv, index=False)
chosen = cands.head(1).copy()
chosen.to_csv(chosen_csv, index=False)

print("\n=== Boundary inflection (fixed) ===")
print(f"X_COL,Y_COL      : {X_COL}, {Y_COL}")
print(f"raw points       : {n_raw}")
print(f"unique x (dedup) : {n_uni}  -> dx=0 eliminated")
print(f"smooth window    : {win}  method={SMOOTH_METHOD}")
print(f"candidates       : {len(cands)}")
if len(chosen) == 1:
    print("top candidate:", chosen.iloc[0].to_dict())
print("saved:", cand_csv)
print("saved:", chosen_csv)

fig_pdf = f"Fig_{OUT_PREFIX}_inflection.pdf"
plt.figure(figsize=(7,4.5))
plt.scatter(x0, y0, s=8, alpha=0.25, label="raw")
plt.plot(x, y, alpha=0.6, label="collapsed (median by x)")
plt.plot(x, y_s, linewidth=2.0, label="smoothed")
K = min(5, len(cands))
for j in range(K):
    xx = cands.loc[j, "x_inflect"]
    yy = cands.loc[j, "y_smooth"]
    plt.axvline(xx, linestyle="--", alpha=0.8)
    plt.plot([xx], [yy], marker="o", markersize=6)
if len(chosen) == 1:
    xx = chosen.loc[0, "x_inflect"]
    plt.axvline(xx, linestyle="-.", linewidth=2.0, alpha=0.9, label=f"chosen x≈{xx:.3f}")
plt.xlabel(X_COL)
plt.ylabel(Y_COL)
plt.title("Inflection detection (dedup + smoothing + sign-change of 2nd derivative)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.savefig(fig_pdf, bbox_inches="tight")
plt.show()
print("saved:", fig_pdf)

# ============================================================
# 2) Optional: risk-on-boundary (AUTO supported)
# ============================================================
if RISK_GRID_CSV is not None:
    grid_patterns = [
        "FG*grid*.csv",
        "*BDI*grid*.csv",
        "*map*grid*.csv",
        "*grid*.csv",
    ]
    grid_path = _resolve_csv(RISK_GRID_CSV, grid_patterns)
    if grid_path is None:
        print("\n[WARN] RISK_GRID_CSV not found; skipping risk-on-boundary.")
    else:
        print("\nUsing RISK_GRID_CSV:", grid_path)
        grid = pd.read_csv(grid_path)
        gx, grho, grisk = _auto_pick_risk_cols(grid)
        if RISK_COL is not None:
            grisk = RISK_COL
        if gx is None or grho is None or grisk is None:
            raise ValueError(f"Could not detect (lambda,rho,risk) cols in grid: {list(grid.columns)}")

        try:
            from scipy.interpolate import griddata
            pts = grid[[gx, grho]].to_numpy(dtype=float)
            vals = grid[grisk].to_numpy(dtype=float)

            bpts = raw[[X_COL, Y_COL]].to_numpy(dtype=float)
            risk_raw = griddata(pts, vals, bpts, method="linear")
            mask = np.isnan(risk_raw)
            if np.any(mask):
                risk_raw[mask] = griddata(pts, vals, bpts[mask], method="nearest")

            xr, yr, n_raw2, n_uni2 = _robust_unique_by_x(raw[X_COL].to_numpy(), risk_raw, agg="median")
            win2 = _choose_window(len(yr), WINDOW_FRACTION, WINDOW_MIN)
            yr_s = _smooth(yr, method=SMOOTH_METHOD, window=win2, polyorder=SAVGOL_POLYORDER)
            rcands, rd1, rd2 = _inflection_candidates(xr, yr_s, margin_frac=MARGIN_FRAC)

            risk_series_csv = f"{OUT_PREFIX}_risk_on_boundary_series.csv"
            pd.DataFrame({"lambda": xr, "risk": yr, "risk_smooth": yr_s}).to_csv(risk_series_csv, index=False)

            risk_cand_csv = f"{OUT_PREFIX}_risk_inflection_candidates.csv"
            risk_chosen_csv = f"{OUT_PREFIX}_risk_inflection_chosen.csv"
            rcands.to_csv(risk_cand_csv, index=False)
            rchosen = rcands.head(1).copy()
            rchosen.to_csv(risk_chosen_csv, index=False)

            print("\n=== Risk-on-boundary (fixed) ===")
            print(f"cols: lambda={gx}, rho={grho}, risk={grisk}")
            print(f"unique lambda: {n_uni2}")
            print(f"smooth window: {win2}")
            print(f"candidates  : {len(rcands)}")
            if len(rchosen) == 1:
                print("top candidate:", rchosen.iloc[0].to_dict())
            print("saved:", risk_series_csv)
            print("saved:", risk_cand_csv)
            print("saved:", risk_chosen_csv)

            rfig_pdf = f"Fig_{OUT_PREFIX}_risk_inflection.pdf"
            plt.figure(figsize=(7,4.5))
            plt.plot(xr, yr, alpha=0.6, label="risk (collapsed median)")
            plt.plot(xr, yr_s, linewidth=2.0, label="risk (smoothed)")
            KK = min(5, len(rcands))
            for j in range(KK):
                xx = rcands.loc[j, "x_inflect"]
                yy = rcands.loc[j, "y_smooth"]
                plt.axvline(xx, linestyle="--", alpha=0.8)
                plt.plot([xx], [yy], marker="o", markersize=6)
            if len(rchosen) == 1:
                xx = rchosen.loc[0, "x_inflect"]
                plt.axvline(xx, linestyle="-.", linewidth=2.0, alpha=0.9, label=f"chosen λ≈{xx:.3f}")
            plt.xlabel("lambda")
            plt.ylabel("risk")
            plt.title("Inflections of risk evaluated on Koibuchi boundary (fixed)")
            plt.grid(True)
            plt.legend()
            plt.tight_layout()
            plt.savefig(rfig_pdf, bbox_inches="tight")
            plt.show()
            print("saved:", rfig_pdf)

        except Exception as e:
            print("\n[WARN] risk-on-boundary skipped (SciPy/griddata unavailable or error).")
            print("Error:", repr(e))
```
