```python
import sympy as sp

def get_boundary_ideal(incidence_equations, state_vars, param_vars):
    """
    消去理論に基づきBoundary Idealを計算する [cite: 82, 85, 318]
    """
    # 状態変数を消去してパラメータのみの多項式を得る
    ideal = sp.groebner(incidence_equations, *state_vars, *param_vars, order='lex')
    boundary_poly = [f for f in ideal if not any(s in f.free_symbols for s in state_vars)]
    return boundary_poly

def calculate_ti(path_func, boundary_poly, s, s0=0):
    """
    Transversality Index (TI) の計算 [cite: 100, 101]
    """
    ti_values = []
    for poly in boundary_poly:
        # パスを境界多項式に代入
        phi_s = poly.subs(path_func)
        # 非ゼロとなる最小の微分次数を探す
        q = 1
        while q < 10:
            diff_val = sp.diff(phi_s, s, q).subs(s, s0)
            if diff_val != 0:
                ti_values.append(q)
                break
            q += 1
    return min(ti_values) if ti_values else None
```

```python
# 変数定義
theta1, theta2 = sp.symbols('theta1 theta2') # パラメータ（例：不平衡率）
# 係数行列 C(theta) の簡略版（3x3行列の行列式が0 = 平面）
# 論文のDefinition 7に基づき、3x3マイナーの消失を境界とする
C_matrix = sp.Matrix([
    [1, 0, theta1],
    [0, 1, theta2],
    [theta1, theta2, 1]
])

I_bd_plane = [C_matrix.det()] # Planarity Boundary Ideal [cite: 148]
print(f"Circuit Boundary Ideal: {I_bd_plane}")
```

```python
# 変数定義
w1, w2 = sp.symbols('w1 w2') # 瞬時周波数
t, a, b = sp.symbols('t a b') # 時間とパラメータ

# リッジ交差のIncidence Ideal [cite: 179, 180]
# 例：w1(t) = a*t, w2(t) = b
I_inc_ridge = [a*t - b]

# 時間変数 t を消去してパラメータ空間の境界を得る
I_bd_ridge = get_boundary_ideal(I_inc_ridge, [t], [a, b])
print(f"Ridge Crossing Boundary: {I_bd_ridge}") # a=0 または b=0 などの特異点
```

```python
# 変数定義
dw, K = sp.symbols('dw K') # デチューニング, ゲイン
s, c = sp.symbols('s c')    # sin(phi), cos(phi) [cite: 193]

# Incidence Ideal [cite: 195]
I_inc_pll = [
    dw - K*s,      # ロック条件 dw = K*sin(e) [cite: 190, 194]
    s**2 + c**2 - 1 # 円の制約 [cite: 195]
]

# 状態変数 s, c を消去
I_bd_pll = get_boundary_ideal(I_inc_pll, [s, c], [dw, K])
print(f"PLL Pull-out Boundary: {I_bd_pll}") # (dw)^2 - K^2 = 0 [cite: 200, 201]
```

```python
# 変数定義
tau1, tau2 = sp.symbols('tau1 tau2') # 遅延
w = sp.symbols('w')                  # 虚軸上の根 lambda = i*w [cite: 207]
u1, v1 = sp.symbols('u1 v1')         # cos(w*tau1), sin(w*tau1) [cite: 208]

# 簡略化した特性方程式の多項式表現（Re, Im部） [cite: 210]
# 例: dy/dt = -y(t) - y(t-tau1)
f_re = v1 # sin(w*tau)
f_im = w + u1 # w + cos(w*tau)

I_inc_stab = [f_re, f_im, u1**2 + v1**2 - 1]

# w, u1, v1 を消去して遅延空間の境界を得る [cite: 211]
I_bd_stab = get_boundary_ideal(I_inc_stab, [w, u1, v1], [tau1])
print(f"Stability Boundary Ideal: {I_bd_stab}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# パラメータ範囲の定義
t1 = np.linspace(-1.5, 1.5, 100)
t2 = np.linspace(-1.5, 1.5, 100)
T1, T2 = np.meshgrid(t1, t2)

# Boundary Ideal の出力結果: theta1^2 + theta2^2 - 1 = 0
# これを Z 値として計算
Z = T1**2 + T2**2 - 1

plt.figure(figsize=(8, 7))

# 境界 (V(I_bd)) を等高線として描画
cp = plt.contour(T1, T2, Z, levels=[0], colors='black', linewidths=2)
plt.clabel(cp, inline=True, fmt='V(I_bd): Boundary', fontsize=12)

# 領域の色分け
# 内側 (Z < 0): 平面レジーム (Rank <= 2)
# 外側 (Z > 0): 空間レジーム (Rank = 3)
plt.contourf(T1, T2, Z, levels=[-2, 0], colors=['#e6f3ff'], alpha=0.8) # 安定・平面
plt.contourf(T1, T2, Z, levels=[0, 3], colors=['#ffe6e6'], alpha=0.8)  # 不安定・空間

# ラベルと説明
plt.title("Case Study I: Parameter Space Boundary Ideal", fontsize=14)
plt.xlabel(r"Parameter $\theta_1$", fontsize=12)
plt.ylabel(r"Parameter $\theta_2$", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)

# 凡例代わりのテキスト
plt.text(-0.5, 0, "Planar Regime\n(Rank $\leq$ 2)\n$\\xi \\equiv 0$",
         ha='center', color='blue', fontweight='bold')
plt.text(1.1, 1.1, "Spatial Regime\n(Rank 3)\n$\\xi \\neq 0$",
         ha='center', color='red', fontweight='bold')

plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)
plt.show()
```

```python
# パラメータパスの定義: (1 + s, 0) すなわち境界 (1, 0) を横切る直線
s = sp.symbols('s')
path = {theta1: 1 + s, theta2: 0}

# 1. 境界多項式 phi(theta)
phi = I_bd_plane[0] # -theta1**2 - theta2**2 + 1

# 2. パスに沿った境界の変化 phi(s)
phi_s = phi.subs(path)
print(f"Path along boundary phi(s): {phi_s}")

# 3. TI (Transversality Index) の算出
# s=0 で非ゼロになる最小の微分次数
ti = calculate_ti(path, I_bd_plane, s)

print(f"--- Diagnostic Results ---")
print(f"Boundary Point: (theta1, theta2) = (1, 0)")
print(f"Transversality Index (TI): {ti}")
```
