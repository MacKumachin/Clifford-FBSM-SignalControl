```python
import os
import re
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# ---------------------------------------------------------
# 0. File Verification
# ---------------------------------------------------------
required_files = ["df_lp.csv", "gap.csv"]
missing_files = [f for f in required_files if not os.path.exists(f)]

if missing_files:
    print("【Error】The following files are missing:", missing_files)
else:
    print("File verification complete. Commencing estimations...")

    # ---------------------------------------------------------
    # 1. Aesthetics and Global Settings for Academic Output
    # ---------------------------------------------------------
    sns.set_theme(style="whitegrid")
    # Setting a standard sans-serif font suitable for journal publication
    plt.rcParams["font.family"] = "sans-serif"
    plt.rcParams["pdf.fonttype"] = 42 # Ensures fonts are embedded in PDF
    plt.rcParams["ps.fonttype"] = 42

    max_h = 12
    controls = ["d_US_10Y", "d_Commodity", "Shock_lag1"]

    # ---------------------------------------------------------
    # 2. Data Ingestion and Preprocessing
    # ---------------------------------------------------------
    df = pd.read_csv("df_lp.csv", encoding="utf-8", encoding_errors="replace")
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
    df = df.dropna(subset=["Date"]).copy()

    # Robust parsing for gap.csv
    gap_data = []
    with open("gap.csv", "r", encoding="cp932", errors="replace") as f:
        for line in f:
            parts = line.split(",")
            if len(parts) >= 2:
                q_str = parts[0].strip().replace('"', '')
                val_str = parts[1].strip().replace('"', '')

                match = re.search(r'(\d{4})\D*([1-4])', q_str)
                if match:
                    year = match.group(1)
                    q = match.group(2)
                    month = int(q) * 3 - 2
                    date_val = f"{year}-{month:02d}-01"
                    gap_data.append([date_val, val_str])

    df_gap_raw = pd.DataFrame(gap_data, columns=["Date", "Output_Gap"])
    df_gap_raw["Date"] = pd.to_datetime(df_gap_raw["Date"], errors="coerce")
    df_gap_raw["Output_Gap"] = pd.to_numeric(df_gap_raw["Output_Gap"], errors="coerce")
    df_gap_raw = df_gap_raw.dropna().copy()

    df_gap = pd.DataFrame()
    if len(df_gap_raw) > 0:
        df_gap_indexed = df_gap_raw.drop_duplicates(subset=["Date"]).set_index("Date")
        df_gap_resampled = df_gap_indexed.resample("MS").interpolate(method="linear")
        df_gap = pd.DataFrame({
            "Date": df_gap_resampled.index,
            "Output_Gap": df_gap_resampled["Output_Gap"].values
        })
        df = pd.merge(df, df_gap, on="Date", how="left")
    else:
        df["Output_Gap"] = np.nan

    # Alt_Shock Generation (with fail-safe)
    alt_proxy_name = ""
    alt_proxy_valid = False

    if "Output_Gap" in df.columns and df["Output_Gap"].notna().sum() > 20:
        temp_alt = df[["Date", "Output_Gap", "d_US_10Y", "d_Commodity"]].copy()
        temp_alt["d_Output_Gap"] = temp_alt["Output_Gap"].diff()
        temp_alt = temp_alt.dropna().copy()

        if len(temp_alt) > 15:
            X_alt = sm.add_constant(temp_alt[["d_US_10Y", "d_Commodity"]])
            y_alt = temp_alt["d_Output_Gap"]
            model_alt = sm.OLS(y_alt, X_alt).fit()
            temp_alt["Alt_Shock"] = model_alt.resid
            df = pd.merge(df, temp_alt[["Date", "Alt_Shock"]], on="Date", how="left")
            alt_proxy_name = "Output Gap"
            alt_proxy_valid = True

    if not alt_proxy_valid:
        print("[Notification] Employing the first difference of 'Expectation_Gap' as the alternative proxy due to limited 'gap.csv' overlap.")
        temp_alt = df[["Date", "Expectation_Gap", "d_US_10Y", "d_Commodity"]].copy()
        temp_alt["d_Expectation_Gap_Alt"] = temp_alt["Expectation_Gap"].diff()
        temp_alt = temp_alt.dropna().copy()

        X_alt = sm.add_constant(temp_alt[["d_US_10Y", "d_Commodity"]])
        y_alt = temp_alt["d_Expectation_Gap_Alt"]
        model_alt = sm.OLS(y_alt, X_alt).fit()
        temp_alt["Alt_Shock"] = model_alt.resid * -1

        df = pd.merge(df, temp_alt[["Date", "Alt_Shock"]], on="Date", how="left")
        alt_proxy_name = "Alt. Expectation Gap (1st Diff)"

    # ---------------------------------------------------------
    # 3. Local Projections Estimation Function
    # ---------------------------------------------------------
    def estimate_lp(data, shock_col="Shock", state_col="I_t_minus_1"):
        results_punct = []
        for h in range(max_h + 1):
            target = f"d_ln_C_forward_{h}"
            if target not in data.columns or shock_col not in data.columns:
                continue

            req_cols = [target, shock_col, state_col] + [c for c in controls if c in data.columns]
            temp = data[req_cols].dropna().copy()
            if len(temp) < 5:
                continue

            temp["Shock_Smooth"] = temp[shock_col] * (1 - temp[state_col])
            temp["Shock_Punct"] = temp[shock_col] * temp[state_col]

            X_cols = ["Shock_Smooth", "Shock_Punct"] + [c for c in controls if c in data.columns]
            X = sm.add_constant(temp[X_cols])
            y = temp[target]

            model = sm.OLS(y, X).fit(cov_type="HAC", cov_kwds={"maxlags": h + 1})
            results_punct.append({
                "h": h,
                "coef": model.params["Shock_Punct"],
                "lower": model.conf_int(alpha=0.1).loc["Shock_Punct", 0],
                "upper": model.conf_int(alpha=0.1).loc["Shock_Punct", 1],
            })
        return pd.DataFrame(results_punct)

    # ---------------------------------------------------------
    # 4. Robustness Checks Execution
    # ---------------------------------------------------------
    print("Executing 1/4: Leave-one-break-out...")
    drop_dates = pd.to_datetime(["2022-12-01", "2023-01-01", "2023-07-01"])
    df_lobo = df[~df["Date"].isin(drop_dates)].copy()
    res_lobo = estimate_lp(df_lobo)

    print("Executing 2/4: Placebo Test...")
    np.random.seed(42)
    df_placebo = df.copy()
    if len(df_placebo) > 0 and "I_t_minus_1" in df_placebo.columns:
        df_placebo["I_t_minus_1"] = np.random.permutation(df_placebo["I_t_minus_1"].values)
    res_placebo = estimate_lp(df_placebo)

    print("Executing 3/4: Wild Bootstrap...")
    def wild_bootstrap_lp(data, h, B=300):
        target = f"d_ln_C_forward_{h}"
        if target not in data.columns:
            return 0, 0, 0

        req_cols = [target, "Shock", "I_t_minus_1"] + [c for c in controls if c in data.columns]
        temp = data[req_cols].dropna().copy()
        if len(temp) < 5:
            return 0, 0, 0

        temp["Shock_Smooth"] = temp["Shock"] * (1 - temp["I_t_minus_1"])
        temp["Shock_Punct"] = temp["Shock"] * temp["I_t_minus_1"]

        X_cols = ["Shock_Smooth", "Shock_Punct"] + [c for c in controls if c in data.columns]
        X = sm.add_constant(temp[X_cols])
        y = temp[target]

        model = sm.OLS(y, X).fit()
        y_hat = model.fittedvalues
        resids = model.resid
        boot_coefs = []

        for _ in range(B):
            weights = np.random.choice([-1, 1], size=len(resids))
            y_boot = y_hat + weights * resids
            boot_model = sm.OLS(y_boot, X).fit()
            boot_coefs.append(boot_model.params["Shock_Punct"])

        return np.percentile(boot_coefs, 5), np.percentile(boot_coefs, 95), model.params["Shock_Punct"]

    wb_lower, wb_upper, wb_coef = [], [], []
    for h in range(max_h + 1):
        l, u, c = wild_bootstrap_lp(df, h, B=300)
        wb_lower.append(l)
        wb_upper.append(u)
        wb_coef.append(c)

    print("Executing 4/4: Alternative Epistemic Proxy...")
    res_alt = estimate_lp(df, shock_col="Alt_Shock")

    # ---------------------------------------------------------
    # 5. PDF Export Generation
    # ---------------------------------------------------------
    fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharey=True)

    def plot_ax(ax, x, y, lower, upper, title):
        ax.plot(x, y, color="maroon", label="Point Estimate (Punctured)")
        ax.fill_between(x, lower, upper, color="red", alpha=0.2, label="90% CI")
        ax.axhline(0, color="black", linestyle="--", linewidth=1)
        ax.set_title(title, fontsize=12, fontweight="bold")
        ax.set_xlabel("Horizon (Months)")
        ax.set_ylabel("Cumulative Growth of JGB Holdings")

    if not res_lobo.empty:
        plot_ax(axes[0, 0], res_lobo["h"], res_lobo["coef"], res_lobo["lower"], res_lobo["upper"],
                "A. Leave-one-break-out (Excl. Dec 22, Jan/Jul 23)")

    if not res_placebo.empty:
        plot_ax(axes[0, 1], res_placebo["h"], res_placebo["coef"], res_placebo["lower"], res_placebo["upper"],
                "B. Placebo Test (Randomized State Indicator)")

    if any(c != 0 for c in wb_coef):
        plot_ax(axes[1, 0], range(max_h + 1), wb_coef, wb_lower, wb_upper,
                "C. Small-Sample Wild Bootstrap (90% CI)")

    if not res_alt.empty:
        plot_ax(axes[1, 1], res_alt["h"], res_alt["coef"], res_alt["lower"], res_alt["upper"],
                f"D. Alternative Epistemic-Gap Proxy ({alt_proxy_name})")
    else:
        axes[1, 1].axhline(0, color="black", linestyle="--", linewidth=1)
        axes[1, 1].set_title("D. Alternative Epistemic-Gap Proxy", fontsize=12, fontweight="bold")
        axes[1, 1].text(0.5, 0.5, "Data insufficient for Alt Proxy", ha="center")

    plt.tight_layout()

    # PDFとして出力 (ジャーナル提出用)
    pdf_filename = "Appendix_Robustness_Panel.pdf"
    plt.savefig(pdf_filename, format="pdf", bbox_inches="tight")
    plt.show()

    # CSV Summaryの出力
    res_combined = pd.DataFrame({"Horizon": range(max_h + 1)})
    if not res_lobo.empty:
        res_combined["LOBO_Coef"] = res_lobo["coef"]
        res_combined["LOBO_SE_Lower"] = res_lobo["lower"]
        res_combined["LOBO_SE_Upper"] = res_lobo["upper"]
    if not res_alt.empty:
        res_combined["AltProxy_Coef"] = res_alt["coef"]
        res_combined["AltProxy_SE_Lower"] = res_alt["lower"]
        res_combined["AltProxy_SE_Upper"] = res_alt["upper"]

    csv_filename = "Robustness_Results_Summary.csv"
    res_combined.to_csv(csv_filename, index=False)

    print(f"\n✅ Procedure completed successfully.")
    print(f"✅ The vector graphics figure has been saved as '{pdf_filename}'.")
    print(f"✅ The summary data table has been saved as '{csv_filename}'.")
```
